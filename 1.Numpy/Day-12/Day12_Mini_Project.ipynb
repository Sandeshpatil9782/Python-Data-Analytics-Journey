### Replace using Conditions


```python
a=np.array([10,20,30,40,50])

a[a > 20] = 999

print(a)
```

### Masking
    Boolean arrays used for filtering

```python
a=np.array([10,20,30,40,50])
mask=a >20
print(mask)

print(a[mask])
```

### Modify using nditer

```python
a=np.array([1,2,3])
for i in np.nditer(a,op_flags=['readwrite']):
    i[...]=i * 10
print(a)
```

## Mini Project
    Student Marks

```python
import numpy as np

marks = np.array([
    [78,85,90],
    [45,50,55],
    [88,92,95],
    [35,40,30]
])

print(marks)
```

### Total Marks

```python
total = np.sum(marks, axis=1)

print(total)
```

### Average Marks

```python
average = np.mean(marks, axis=1)

print(average,"\n")

average = np.round(np.mean(marks, axis=1), 2)

print("Only 2 decimal places:\n",average)
```

### Highest Marks

```python
print(np.max(marks))
```

#### Lowest Marks

```python
print(np.min(marks))
```

#### Passed Student
    Average is > 40

```python
passed = average >= 40

print(passed)
```

### Bonus Marks = 5

```python
bonus = marks + 5

print(bonus)
```

#### Topper Student

```python
topper = np.argmax(total)

print(topper) #Index 2 student is topper Because total=275
```

#### Sort Totals

```python
print(np.sort(total)) #Ascending
```

```python
print(np.sort(total)[: : -1]) #Descending
```

### SUMMURY

```python
print("Totals:", total)

print("Average:", np.round(np.mean(marks, axis=1), 2))


print("Highest:", np.max(marks))

print("Lowest:", np.min(marks))
```
