## Numpy Statistical and Other Functions

```python
import numpy as np
```

```python
a=np.array([10,20,30,40,50])
a
```

### Statistical Functions

#### 1.Mean()

```python
print(np.mean(a))
```

#### 2.Median()

```python
print(np.median(a))
```

#### 3.Sum()

```python
print(np.sum(a))
```

#### 4.Min()

```python
print(np.min(a))
```

#### 5.Max()

```python
print(np.max(a))
```

#### 6.Standard Deviation- Std()
    Mesures spread of data from mean

```python
print(np.std(a))
```

#### 7.Variance-Var()
    Mesuares data variability

```python
print(np.var(a))
```

### Practice Task

```python
import numpy as np

marks = np.array([78,85,90,67,88,92])

print("Mean:", np.mean(marks))

print("Median:", np.median(marks))

print("Sum:", np.sum(marks))

print("Minimum:", np.min(marks))

print("Maximum:", np.max(marks))

print("Standard Deviation:", np.std(marks))

print("Variance:", np.var(marks))
```

### Other Functions

#### 1.reshape()
    Changes shape of array without changing data

```python
a6=np.array([1,2,3,4,5,6])
a6
```

##### 2 Rows And 3 Columns

```python
b=a6.reshape(2,3)
b
```

##### 3 Rows And 2 Columns

```python
c=a6.reshape(3,2)
c
```

#### 2.flatten()
    Converts multidimensional to 1D array

```python
print(b.flatten())
```

#### 2.1 ravel()
    Same as flatten,But it is faster and efficient

```python
print(b.ravel())
```

#### 3.transpose()
    Converts Rows into Columns and Viseversa

```python
np.transpose(b)
```

##### Using T keyword

```python
print(b.T)
```

### Practice Task

```python
import numpy as np

a = np.arange(1,13)

print("Original Array:")
print(a)

b = a.reshape(3,4)

print("Reshaped Array:")
print(b)

print("Flatten:")
print(b.flatten())

print("Transpose:")
print(b.T)
```

### Create array from 1 to 9 and reshape into 3x3.

```python
a = np.arange(1,10)

print("Original Array:")
print(a)

b = a.reshape(3,3)

print("Reshaped Array 3x3:")
print(b)
```

### Convert 2D array into 1D array.

```python
a=np.arange(1,13).reshape(4,3)
print("2D Array",a)

b=a.flatten()

print("1D Array:",b)
```
