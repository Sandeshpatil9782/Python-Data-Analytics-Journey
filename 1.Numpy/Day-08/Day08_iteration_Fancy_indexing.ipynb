## Iteration,Looping and Fancy,Advanced Indexing

```python
import numpy as np
```

### Iterate 1D Arrays

```python
a=np.array([10,20,30,40,50])
a
```

#### Loop runs through every element

```python
for i in a:
    print(i)
```

### Iterate 2D arrays

```python
b=np.array([ [1,2],
             [3,4] 
           ] )
for i in b:
    print(i)
 #Each Row is printed          
```

```python
#Individual Elements
for row in b:
    for element in row:
        print(element)
```

### Iterate 3D Arrays

```python
a=np.array([ [ [1,2],
               [3,4]
             ],
            [  [5,6],
               [7,8]
            ]
           ] )
#Gives one 2D arrays
for layer in a:
    print(layer)
```

#### Iterate Rows

```python

for layer in a:
    for row in layer:
        print(row)
```

#### Iterate Individual Elements

```python
for layer in a:
    for row in layer:
        for element in row:
            print(element)
```

### nditer()
    Iterates every element easily

```python
for i in np.nditer(a):
    print(i)
```

### enumerate()
    Gives index and value together

```python
a=np.array([10,20,30,40,50])
for index,value in enumerate(a):
    print(index,value)
```

### Pracice Task

```python
import numpy as np

marks = np.array([78,85,90,67])

print("Student Marks:")

for i in marks:
    print(i)

print("Index and Marks:")

for index,value in enumerate(marks):
    print(index,value)
```

### Fancy indexing
    Selecting multiple elements using index list

```python
a8=np.array([10,20,30,40,50,60])
print(a[ [0,2,4] ]) # 0,2,4 here index numbers
```

### Boolean Indexing
    Returns values satisfy conditions

```python
print(a8[a8>15])
```

### Multiple conditions

```python
print( (a8>20) & (a8<60))
```

### Fancy in 2D 

```python
aa8=np.arange(1,10).reshape(3,3)
aa8
```

```python
print(aa8[ [0,2] ]) #selected row 0 and row 2
```

### Where
    Gives position where value satisfy condition

```python
c=np.array([10,20,30,40,50,60])
print(np.where(c > 30 ))
```

### Fancy in 3D

```python
a = np.array([
    [
        [1,2],
        [3,4]
    ],

    [
        [5,6],
        [7,8]
    ]
])
print(a[1,0,1]) #Access single element
```

#### Access 0 and 1 Layers

```python
print(a[[0,1]]) 
```

#### Where in 3D

```python
print(np.where(a > 4)) 
```

#### Select specific element

```python
print(a[[0,1],[1,0],[0,1]])
```

### Pracice Task

```python
import numpy as np

marks = np.array([45,67,89,90,34,76,55])

print("Top Scorers:")
print(marks[marks > 80])

print("Average Students:")
print(marks[(marks >= 50) & (marks <= 80)])

print("Indexes Above 70:")
print(np.where(marks > 70))
```
