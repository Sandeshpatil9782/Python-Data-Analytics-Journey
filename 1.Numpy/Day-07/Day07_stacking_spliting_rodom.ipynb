## Stacking || Splitting || Random Module

```python
import numpy as np
```

```python
a=np.array([1,2,3])
a
```

```python
b=np.array([4,5,6])
b
```

### Stacking Functions

#### 1.hstack()
    Joins array Horizontly

```python
print(np.hstack((a,b)))
```

#### 2.vstack()
    Joins array Vertically

```python
print(np.vstack((a,b)))
```

#### 3.Concatenate()
    Combines arrays

```python
print(np.concatenate((a,b)))
```

### Splitting Functions
   

#### 1.split()
     Divides arrays into parts

```python
c=np.array([1,2,3,4,5,6])
print(np.split(c,3))
```

#### 2.hsplit()
    Horizontly split

```python
d=np.arange(1,9).reshape(2,4)
print(d)
print("Horizontly:",np.hsplit(d,2))
```

#### 3.vsplit()
    Vertically

```python
e=np.arange(1,9).reshape(4,2)
print(e)

print("Vertically:",np.vsplit(e,2))
```

### Practice Task

```python
import numpy as np

salary = np.array([25000,30000,35000])

bonus = np.array([2000,3000,4000])

print("Horizontal Stack:")
print(np.hstack((salary,bonus)))

print("Vertical Stack:")
print(np.vstack((salary,bonus)))

data = np.array([10,20,30,40,50,60])

print("Split Data:")
print(np.split(data,3))
```

### Random Module

#### 1.rand()
    Random decimal numbers from 0 and 1
    

```python
print(np.random.rand(5))
```

#### 2.randint()
    Generates random integers

```python
print(np.random.randint(1,10,5))
```

#### 3.randn()
    Random values from normal distribution

```python
print(np.random.randn(5))
```

#### 4.choice()
    Select random values from list

```python
colors=["Red","Blue","Green"]
print(np.random.choice(colors))
```

#### 5.seed()
    Fixes Random Output or same output every time

```python
np.random.seed(1)
print(np.random.randint(1,10,4))
```

```python
import numpy as np

marks = np.random.randint(35,100,10)

print("Student Marks:")
print(marks)

print("Average Marks:")
print(np.mean(marks))

print("Highest Marks:")
print(np.max(marks))
```
