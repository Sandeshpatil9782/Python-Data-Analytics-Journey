## Case Study: Airline Passenger Trend Analysis

    Business Problem
        An airline company wants to analyze passenger trends over time to improve capacity planning and seasonal pricing.

    Business Questions
        How have passenger numbers changed over the years?
        Which months have the highest passenger traffic?
        Are there seasonal patterns in air travel?
        Which year had the highest passenger volume?

```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

```

```python
sns.set_style("whitegrid")

df = sns.load_dataset("flights")

df.head()
```

### How have passenger numbers changed over time?
    Trend Analysis

    Insights:-
        Passenger trafic increases steadily over the years.

```python
yearly = df.groupby("year", as_index=False)["passengers"].sum()

plt.figure(figsize=(8,6))

sns.lineplot(

    data=yearly,

    x="year",

    y="passengers",

    marker="o",

    linewidth=2.5,

    color="darkblue"

)

plt.title("Yearly Airline Passenger Trend", fontsize=14, fontweight="bold")

plt.xlabel("Year")

plt.ylabel("Total Passengers")

plt.show()
```

### Which months have the highest average passenger traffic?
    Monthly Comparison

    insight:-
        Summer months usually show higher passenger traffic.

```python
sns.barplot(

data=df,

x="month",

y="passengers",

hue='month',
    
estimator="mean",

palette="viridis"

)

plt.title("Average Passengers by Month", fontsize=14, fontweight="bold")

plt.xlabel("Month")

plt.ylabel("Average Passengers")

plt.xticks(rotation=45)

plt.show()
```

### Are there seasonal patterns in air travel?
    Seasonality Analysis

    Insights:-
        Passenger counts consistently peak in July And August.

```python
pivot = df.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(8,6))

sns.heatmap(

pivot,

annot=True,

fmt="d",

cmap="YlOrRd"

)

plt.title("Monthly Passenger Heatmap", fontsize=14, fontweight="bold")

plt.xlabel("Year")

plt.ylabel("Month")

plt.show()
```

### How are passenger counts distributed?
    Distribution Analysis

    INSIGHT:-
        Passenger counts are right-skewed,with more observations in the lower-to-mid range

```python
sns.histplot(

data=df,

x="passengers",

bins=20,

kde=True,

color="blue"

)

plt.title("Distribution of Passenger Counts", fontsize=14, fontweight="bold")

plt.xlabel("Passengers")

plt.ylabel("Frequency")

plt.show()
```

### Which months shows the highest variation in passenger traffic?

```python
sns.boxplot(
    data=df,
    x="month",
    y="passengers",
    hue='month',
    palette="Set3"
)

plt.title("Passenger Distribution by Month", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")

plt.xticks(rotation=45)

plt.show()
```

### How is passenger traffic distributed across all observations?

```python
sns.violinplot(
    data=df,
    y="passengers",
    color="lightgreen"
)

plt.title("Distribution of Airline Passengers", fontsize=14, fontweight="bold")
plt.ylabel("Passengers")

plt.show()
```

### Is there relationship between Year and Passenger Count?

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=df,
    x="year",
    y="passengers",
    scatter_kws={"color":"steelblue","s":50},
    line_kws={"color":"red","linewidth":2}
)

plt.title("Trend of Passengers Over the Years", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Passengers")

plt.show()
```

### Which months have similar passenger traffic patterns across different years

    Insight
        -Passenger traffic generally increases during the middle of the year.
        -Months with a wider spread indicate greater variation in passenger counts.

```python
sns.stripplot(
    data=df,
    x="month",
    y="passengers",
    color="darkorange",
    size=6
)

plt.title("Passenger Distribution Across Months", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")

plt.xticks(rotation=45)

plt.show()
```

### How are passenger counts distributed by month while avoiding overlapping points?

    Insight
        -Each point represents one month's passenger count for a particular year.
        -Swarm plots prevent overlapping points, making it easier to observe the distribution and density across months.

```python
sns.swarmplot(
    data=df,
    x="month",
    y="passengers",
    palette="Set2",
    hue="month",
    legend=False
)

plt.title("Passenger Distribution by Month", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")

plt.xticks(rotation=45)

plt.show()
```

### Conclusion

This analysis identified clear trends and seasonal patterns in airline passenger traffic using Seaborn visualizations. Passenger numbers increased consistently over the years, with July and August recording the highest demand. The visualizations helped uncover distribution patterns, monthly variations, and overall growth trends. These insights can support capacity planning, demand forecasting, and pricing strategies for airline businesses.
