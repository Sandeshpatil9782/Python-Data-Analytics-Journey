### Seaborn Cat Plot (sns.catplot()) (Previously Factor Plot)

    Definition
    
    sns.catplot() is a figure-level function used to visualize relationships between categorical and numerical variables. 
    It can generate multiple categorical plot types using the kind parameter.
    
    Real-World Use Cases
        Compare average sales across product categories.
        Analyze customer spending by gender.
        Study employee salaries across departments.
        Compare loan amounts by loan status.
        Visualize customer counts by region.

    NOTE:- factorplot() has been deprecated in Seaborn. It has been replaced by catplot().

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
tips.shape
```

```python
tips.info()
```

```python
sns.catplot( data=tips,
                x='size',
                y='tip',
                hue='size',
                legend=False
              )

plt.show()
```

```python
sns.catplot( data=tips,
                x='size',
                y='tip',
                hue='sex',
                palette='dark:r'
              )

plt.show()
```

```python
sns.catplot( data=tips,
                x='size',
                y='tip',
                hue='sex',
                palette='deep',
                height=7
              )

plt.show()
```

### Kind='' Parameter

    kind="bar"
    kind="count"
    kind="box"
    kind="violin"
    kind="strip"
    kind="swarm"
    kind="point"

```python
sns.catplot( data=tips,
                x='day',
                y='tip',
                hue='sex',
                palette='Oranges',
                height=5,
                kind='bar'
              )
plt.title(
    "Average Customer Tip by Day",
    fontsize=15,
    fontweight="bold"
)


plt.show()
```

```python
g=sns.catplot( data=tips,
                x='size',
                y='total_bill',
                hue='sex',
                palette='viridis',
                height=5,
                kind='point'
              )
g.fig.suptitle(
    "Average Customer Bill by Day",
    fontsize=15,
    fontweight="bold"
)

plt.show()
```

```python
sns.catplot(
    data=tips,
    x="day",
    hue="sex",
    kind="count",
    palette="Set1",
    height=5
)

plt.show()
```

```python
sns.catplot(
    data=tips,
    x="day",
    y='total_bill',
    hue="sex",
    kind="box",
    palette="Set1",
    height=5
)

plt.show()
```

```python
sns.catplot(
    data=tips,
    x="day",
    y='total_bill',
    hue="sex",
    kind="violin",
    palette="Set1",
    height=5
)

plt.show()
```

```python
sns.catplot(
    data=tips,
    x="day",
    y='total_bill',
    hue="smoker",
    kind="strip",
    palette="Set1",
    height=5
)

plt.show()
```

```python
sns.catplot(
    data=tips,
    x="day",
    y='total_bill',
    hue="sex",
    kind="swarm",
    palette="Set1",
    height=5
)

plt.show()
```

### Faceting With Col Creates Separate Plot For Each Category

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    kind="box",
    col="sex",
    palette="Set2",
    height=5
)
plt.show()
```

### Faceting With Row Creates Separate Plot For Each Category With Row-wise

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    kind="box",
    
    row="smoker",
    
    palette="Set2",
    height=5
)

plt.show()
```

```python
g = sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    kind="box",
    col="smoker",
    palette="Set1",
    height=5,
    aspect=1.2
)

g.fig.suptitle(
    "Restaurant Spending Analysis",
    fontsize=16,
    fontweight="bold",
    color="navy"
)

g.fig.subplots_adjust(top=0.88)
```
