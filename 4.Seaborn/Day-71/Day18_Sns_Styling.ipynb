### Seaborn Plot Styling

    Plot styling makes your visualizations professional, readable, and presentation-ready. It is widely used in dashboards, reports, and business presentations.

    Definition

    Plot styling refers to customizing the appearance of visualizations using themes, colors, fonts, grids, labels, legends, and other aesthetic elements to improve readability and storytelling.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

## Themes (set_theme())

    Definition
        Sets the overall appearance of all Seaborn plots.

        sns.set_theme()
    Theme Styles
        sns.set_theme(style="darkgrid")
        sns.set_theme(style="whitegrid")
        sns.set_theme(style="dark")
        sns.set_theme(style="white")
        sns.set_theme(style="ticks")

## set_style()

    Definition
        Changes only the background style of the plot.

        sns.set_style("whitegrid")

        Available styles
            whitegrid
            darkgrid
            white
            dark
            ticks

## set_context()

    Definition
        Controls the scaling of plot elements such as fonts, lines, and markers for different display purposes.

        sns.set_context("paper")
        sns.set_context("notebook")
        sns.set_context("talk")
        sns.set_context("poster")

            Context	          Use Case
            paper	      Reports
            notebook      Jupyter Notebook
            talk	      Presentations
            poster	      Posters & Large Screens

## Color Palettes

    Definition
        A palette defines the colors used for categorical data.

        palette='deep'
        palette='Set2'
        palette='viridis'
        palette='rocket'
        palette='crest'

```python
sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(10,6), facecolor="whitesmoke")

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    palette="Set2"
)

plt.title(
    "Average Customer Bill by Day",
    fontsize=16,
    fontweight="bold",
    color="navy"
)

plt.xlabel("Day", fontsize=12, fontweight="bold")
plt.ylabel("Average Bill ($)", fontsize=12, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.legend(title="Gender", loc="best")

sns.despine()

plt.tight_layout()

plt.show()
```
