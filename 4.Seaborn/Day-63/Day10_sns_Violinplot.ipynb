## Seaborn Violin Plot (sns.violinplot())
        The Violin Plot is one of the most important plots in Exploratory Data Analysis (EDA) because it combines the information of a Box Plot and a KDE Plot.

    📌 Definition
        A Violin Plot is used to visualize the distribution of numerical data across different categories by combining a box plot with a kernel density estimation (KDE).

        Real-World Use Cases
            Compare employee salaries across departments.
            Analyze customer spending by gender.
            Compare loan amounts across loan categories.
            Study exam score distributions among classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset('tips')
var
```

```python
var.isnull().sum()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='day'
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='time'
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='day',
                linewidth=3
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='day',
                linewidth=2,
                palette='cool'
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='time',
                y='total_bill',
                hue='time',
                linewidth=2,
                order=['Dinner','Lunch']
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='day',
                linewidth=2,
                palette='cool',
                saturation=0.4
)
plt.show()
```

```python
sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='day',
                linewidth=2,
                saturation=0.4,
                #color='magenta',  # Setting a gradient palette using color= is deprecated and will be removed
                palette='dark:magenta'
)
plt.show()
```

```python
 sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='sex',
                linewidth=1,
                split=True,
             
)
plt.show()
```

### scale='width'||'area'||'count'
    Controls how the width of each violin is scaled 
    


### split=True 
    -displays both hue categories within the same violin for easy comparison.
    -Note: Works only when the hue variable has exactly two categories (e.g., Male/Female, Yes/No).


```python
 sns.violinplot(
                data=var,
                x='day',
                y='total_bill',
                hue='sex',
                linewidth=1,
                #split=True,
                #scale='area', The `scale` parameter has been renamed to density_norm
                density_norm='width'
)
plt.show()
```

### Horizontal:-
    Change X-axis Value to Y-axis And Viceversa.

```python
 sns.violinplot(
                data=var,
                y='day',
                x='total_bill',
                hue='sex',
                linewidth=1,
                split=True,
                density_norm='count'
 )
plt.legend(loc='best')
plt.show()
```

### parameter inner='quarts' | 'box' | 'quartile' | 'stick' | 'point' | None

    Controls what is displayed inside the violin.

```python
sns.violinplot(
                data=var,
                x='time',
                y='total_bill',
                hue='time',
                linewidth=2,
                order=['Dinner','Lunch'],

                inner='quart'
)
plt.show()
```

### Cut Parameter
    Restricts the KDE to the observed data range.
    -Default Range 2

```python
sns.violinplot(
    data=var,
    x='day',
    y='total_bill',
    cut=0
)
```

### Single Violin Plot Only For Numerical Values

```python
sns.violinplot(x=var['total_bill'])
               
plt.show()
```

```python
sns.violinplot(y=var['total_bill'])
               
plt.show()
```

```python
plt.figure(figsize=(10,6), facecolor='whitesmoke')

sns.violinplot(
    data=var,
    x='day',
    y='total_bill',
    hue='smoker',
    split=True,
    inner='quart',
    palette='Set2',
    linewidth=1.5
)

plt.title(
    'Restaurant Bill Distribution by Day and Smoking Status',
    fontsize=16,
    fontweight='bold',
    color='navy'
)

plt.xlabel('Day', fontweight='bold')
plt.ylabel('Total Bill ($)', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()

plt.show()
```
