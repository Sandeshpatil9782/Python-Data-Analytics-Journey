### Seaborn Strip Plot (sns.stripplot())

    A Strip Plot is commonly used in Exploratory Data Analysis (EDA) to visualize the distribution of individual data points across different categories.

     Definition

        A Strip Plot displays individual observations for a numerical variable across one or more categorical variables. It uses jitter to prevent overlapping points.

     Real-World Use Cases
        Compare employee salaries across departments.
        Analyze customer spending by payment method.
        Compare loan amounts across loan statuses.
        Study exam scores across different classes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day'
)

plt.show()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='sex'
)
plt.legend(loc='best')
plt.show()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='sex',
    palette='rocket_r',
    linewidth=1,
    edgecolor='red'
)
plt.legend(loc='best')
plt.show()
```

### jitter=True adds small random movement to points, reducing overlap and making dense data easier to interpret.

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='sex',
    palette='rocket_r',
    linewidth=1,
    edgecolor='red',
    jitter=1,
    size=10  # size controls the size of the markers.
    
)
plt.legend(loc='best')
plt.show()
```

```python
ma={'Male':'o','Female':'<'}

sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='sex',
    palette='rocket_r',
    linewidth=1,
    edgecolor='red',
    jitter=1,
    size=10,
    marker='>',
    alpha=0.6
    
    
)
plt.legend(loc='best')
plt.show()
```

```python
sns.stripplot(y=tips['total_bill'])

plt.show()
```

### dodge=True separates points of different hue categories instead of overlapping them.

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    dodge=True,
    palette="Set1"
)
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(8,6), facecolor="whitesmoke")

sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    dodge=True,
    jitter=True,
    size=7,
    alpha=0.9,
    palette="dark:g"
)

plt.title(
    "Restaurant Bill Distribution by Day and Smoking Status",
    fontsize=15,
    fontweight="bold",
    color="navy"
)

plt.xlabel("Day", fontweight="bold")
plt.ylabel("Total Bill ($)", fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```
