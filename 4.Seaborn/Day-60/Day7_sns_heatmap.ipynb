## Seaborn Heatmap (sns.heatmap)

    -A heatmap is one of the most important visualizations for Data Analysts because it helps identify relationships, patterns, and correlations quickly.
    -A Heatmap is a graphical representation of data where values are represented using colors.
    
    Real-World Use Cases
        Analyze correlation between business metrics.
        Identify highly related features before building ML models.
        Detect sales performance across regions and months.
        Analyze website traffic patterns.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)

plt.show()
```

```python
var = sns.load_dataset('anagrams')
x=var.drop(columns=['attnr']).head(10)
x
```

### vmin And vmax Parameter 

```python
sns.heatmap(x,vmin=0,vmax=20,annot=True,fmt='.2f')  # fmt='.2f' Add decimal point

plt.show()
```

### cmap='PuOr' Is foe color on the map  And  annot=True Is for labeling values in map

```python
sns.heatmap(x,cmap='PuOr',annot=True)   

plt.show()
```

```python
va=np.linspace(1,10,10).reshape(2,5)
va
sns.heatmap(va,vmin=0,vmax=20,cmap='Dark2_r')
```

```python
a=np.array([['a0','a1','a2','a3','a4'],
              ['b0','b1','b2','b3','b4']  ]
          )
a
```

```python
sns.heatmap(va,
            vmin=0,
            vmax=20,
            cmap='Dark2_r',
           annot=a,
           fmt='s')
```

```python
y={'fontsize':16,'color':'white'}

sns.heatmap(va,
            vmin=0,
            vmax=20,
            annot=True,
            annot_kws=y,
           linewidth=5,     # Line yellow Here
            linecolor='y',    # Gives Line Color Here Yellow
            cbar=False,      # Remove ColorBar

           xticklabels=False,  # Remove X-axis Labels
           yticklabels=False,  # Remove y-axis labels
           )

plt.show()
```

```python
y={'fontsize':16,'color':'white'}

lbl=sns.heatmap(va,
            vmin=0,
            vmax=20,
            annot=True,
            annot_kws=y,
           linewidth=5,     # Line yellow Here
            linecolor='y',    # Gives Line Color Here Yellow
            cbar=False,      # Remove ColorBar

           xticklabels=False,  # Remove X-axis Labels
           yticklabels=False,  # Remove y-axis labels
           )

lbl.set(xlabel='Matplotlib',ylabel='Seaborn',title='Data Visualization')   # Gives Labels into map
#sns.set(font_scale=2)  # Gives size to all labels ad titles
plt.show()
```

```python
flights = sns.load_dataset('flights')

pivot_data = flights.pivot(
    index='month',
    columns='year',
    values='passengers'
    
)

plt.figure(figsize=(10,6))

sns.heatmap(pivot_data,cmap='crest')

plt.title('Flight Passenger Heatmap')
plt.show()
```

```python
tips = sns.load_dataset('tips')

corr = tips.corr(numeric_only=True)

plt.figure(figsize=(8,6), facecolor='whitesmoke')

sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=1,
    linecolor='white',
    cbar=True
)

plt.title(
    'Correlation Heatmap of Restaurant Dataset',
    fontsize=16,
    fontweight='bold',
    color='navy'
)

plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.tight_layout()

plt.show()
```
