## Seaborn Count Plot (sns.countplot())
    A Count Plot is used to display the frequency or count of observations in each category.

    Real-World Use Cases
        Count customers by gender.
        Count loan applicants by loan status.
        Count employees by department.
        Count orders by product category.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

### Counts the Number of Male and female

```python

sns.countplot(
    data=tips,
    x='sex',
    hue='sex'
   
)

plt.show()
```

### Different Parameters

### How may Smokers Are Present  In Male And Female?

```python
sns.countplot(y='sex',
            data=tips,
            hue="smoker",

            #color='r',
            saturation=0.6,  # Transperency Numer Betwwen 0 And 1
            palette='dark:g'
             )

plt.show()
```

### Which day receives the highest number of customers?

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=tips,
    x='day',
    hue='sex',
    palette='dark:magenta',
    saturation=0.3,
    hue_order=['Female','Male']
)

plt.title('Customer Visits by Day and Gender')
plt.show()
```

### How does customer gender distribution vary across different days?

```python
sns.countplot(
    data=tips,
    y='day',
    hue='day',
    palette='viridis'
)

plt.title('Customer Visits by Day')
plt.show()
```

### Custom Category Order


```python
orders = ['Thur', 'Fri', 'Sat', 'Sun']

sns.countplot(
    data=tips,
    x='day',
    hue='day',
    order=orders
)

plt.show()
```

### Adding Data Labels

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=tips,
    x='day',
    hue='day',
    palette='deep'
)

for container in ax.containers:
    ax.bar_label(container)

plt.title('Customer Visits by Day')
plt.show()
```

```python
plt.figure(figsize=(9,5), facecolor='whitesmoke')

ax = sns.countplot(
    data=tips,
    x='day',
    hue='smoker',
    palette='Set2'
)

for container in ax.containers:
    ax.bar_label(container)

plt.title(
    'Restaurant Customer Count by Day and Smoking Status',
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.xlabel('Day', fontweight='bold')
plt.ylabel('Number of Customers', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
