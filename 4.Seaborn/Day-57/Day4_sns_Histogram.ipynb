## Seaborn Histogram (histplot)
    A Histogram is used to visualize the distribution of a numerical variable by grouping values into intervals called bins.

    Real-World Use Cases

    - Analyze customer age distribution.
    - Understand employee salary distribution.
    - Study sales amount distribution.
    - Detect skewness in business data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic= sns.load_dataset('titanic')

titanic.head()
```

### What is the age distribution of Titanic passengers?

```python

sns.displot(titanic['age'],bins=[10,20,30,40,50,60,70,80],rug=True,kde=True)  # Using displot

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=titanic,
    x='age'
)

plt.title('Distribution of Passenger Age')
plt.show()
```

## With kde

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=titanic,
    x='age',
    kde=True,  # Add density curve
    color='navy',
    
)

plt.title('Passenger Age Distribution with KDE')
plt.show()
```

## How does age distribution differ between male and female passengers?

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=titanic,
    x='age',
    bins=20,
    kde=True,  
    hue='sex',  # Group by category
    
    palette='viridis', # Change color scheme

    color='navy',
    line_kws={
        'color': 'red',
        'linewidth': 3,
        'linestyle': '--'
    }
   
)


plt.title('Passenger Age Distribution')
plt.show()
```

### stat
    The stat parameter controls what is displayed on the y-axis of the histogram.

        Available Options
        Value	                       Meaning
       'count'          Shows the frequency/count of observations (Default).
       'frequency'	    Shows count divided by bin width.
       'probability'    Shows the probability of observations in each bin.
       'percent'	    Shows percentage of observations in each bin.
       'density'	    Shows probability density; total area under histogram equals 1.

```python
sns.histplot(
    data=titanic,
    x='age',
    color='g',
    stat='probability'   
)


plt.show()
```

### multiple
    The multiple parameter controls how multiple distributions (created using hue) are displayed.

    Available Options
        Value	          Meaning
       'layer'	    Overlays histograms on top of each other (Default).
       'stack'	    Stacks histograms vertically.
       'dodge'	    Places bars side-by-side.
       'fill'	    Normalizes bars so each bin sums to 100%.

```python
sns.histplot(
    data=titanic,
    x='age',
    hue='sex',
    multiple='dodge'
)

plt.show()
```

### element
    The element parameter controls the visual style of histogram bars.

    Available Options
    Value	             Meaning
    'bars'	       Traditional histogram bars (Default).
    'step'	       Displays histogram as outlined steps.
    'poly'	       Displays histogram as connected polygons.

```python
sns.histplot(
    data=titanic,
    x='age',
    element='poly'
)

plt.show
```

```python
sns.set_theme(style='whitegrid')

# Create figure
plt.figure(figsize=(10, 6), facecolor='whitesmoke')

# Histogram
ax = sns.histplot(
    data=titanic,
    x='age',
    hue='sex',
    bins=20,
    kde=True,
    stat='percent',
    multiple='dodge',
    element='bars',
    palette='Set2',
    alpha=0.8,
    
)

# Title
plt.title(
    'Passenger Age Distribution by Gender',
    fontsize=18,
    fontweight='bold',
    color='navy',
    pad=20
)

# Axis labels
plt.xlabel(
    'Age (Years)',
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    'Percentage of Passengers (%)',
    fontsize=13,
    fontweight='bold'
)

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Customize legend


# Remove top and right borders
sns.despine()

# Adjust layout
plt.tight_layout()

plt.show()
```
