### Seaborn Box Plot (sns.boxplot())

    A Box Plot is one of the most important visualizations in Exploratory Data Analysis (EDA). It summarizes the distribution of data and helps identify outliers, spread, and central tendency.

    Definition

        A Box Plot is used to visualize the distribution of a numerical variable across different categories by displaying the median, quartiles, whiskers, and outliers.

    Real-World Use Cases
        Compare employee salaries across departments.
        Analyze customer spending by payment method.
        Compare loan amounts across loan statuses.
        Identify regions with unusual sales.
        Detect fraudulent or abnormal transactions.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### InBuilt Dataset

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.boxplot(data=tips,
           x='day',
           y='total_bill',
            hue='day'
           )

plt.show()
```

```python
sns.boxplot(data=tips,
           x='day',
           y='total_bill',
            hue='time'
           )

plt.legend(loc="best")
plt.show()
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    width=0.8
)
plt.show()
```

```python
sns.boxplot(data=tips,
           x='day',
           y='total_bill',
            hue='sex',
            #color='g'
            palette='dark:g'
           )

plt.legend(loc="best")
plt.show()
```

```python
sns.boxplot(data=tips,
           x='day',
           y='total_bill',
            hue='sex',
            #color='g'
            palette='Set1',
            order=['Fri','Sun','Thur','Sat']
           )

plt.legend(loc="best")
plt.show()
```

### Mean Markers

```python
sns.boxplot(data=tips,
           x='day',
           y='total_bill',
            hue='smoker',
            palette='plasma',
            order=['Fri','Sun','Thur','Sat'],
            showmeans=True,
            linewidth=2,
            meanprops={
                        "marker":'^',
                        'markeredgecolor':'black',
                        'markersize':10
            },
            
            orient='v',
           )

plt.legend(loc="best")
plt.show()
```

```python
sns.boxplot(data=tips,
           y='day',
           x='total_bill',
            hue='sex',
        
            palette='plasma',
            order=['Fri','Sun','Thur','Sat'],
            showmeans=True,
            linewidth=2,
            meanprops={
                        "marker":'^',
                        'markeredgecolor':'black',
                        'markersize':10
            },
            
            orient='h' # Horizonatl orientaion requires numerical x-axis Column
           )

plt.legend(loc="best")
plt.show()
```

```python
sns.boxplot(data=tips,
            palette='plasma',
            showmeans=True,
            linewidth=2,
            meanprops={
                        "marker":'^',
                        'markeredgecolor':'black',
                        'markersize':10
            },
            orient='h'
           )


plt.show()
```

### Change Outliers Marker

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    flierprops={
        "marker":"D",
        "markerfacecolor":"red",
        "markersize":8
    }
)
plt.show()
```

### Customize Median Line

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    medianprops={
        "color":"red",
        "linewidth":2
    }
)

plt.show()
```

### Customize Whisker Line

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day',
    whiskerprops={
        "color":"blue",
        "linewidth":2
    }
)
plt.show()
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    boxprops={
        "facecolor": "lightblue",
        "edgecolor": "navy",
        "linewidth": 2
    }
)

plt.title("Box Plot with Custom Box")
plt.show()
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    capprops={
        "color": "red",
        "linewidth": 3
    }
)

plt.title("Box Plot with Custom Caps")
plt.show()
```

```python
sns.set_style("darkgrid")

sns.boxplot(y=tips['total_bill'])

plt.show()

```

```python
plt.figure(figsize=(10,6), facecolor="whitesmoke")

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    palette="Set2",
    showmeans=True,
    width=0.6,
    medianprops={"color":"red","linewidth":2},
    flierprops={
        "marker":"D",
        "markerfacecolor":"black",
        "markersize":6
    }
)

plt.title(
    "Customer Bill Distribution by Day and Smoking Status",
    fontsize=16,
    fontweight="bold",
    color="navy"
)

plt.xlabel("Day", fontweight="bold")
plt.ylabel("Total Bill ($)", fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.legend(loc=2)
plt.show()
```
