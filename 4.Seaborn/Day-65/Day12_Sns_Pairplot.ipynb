## Seaborn Pair Plot (sns.pairplot())

    pairplot() is one of the most powerful plots for Exploratory Data Analysis (EDA) because it helps analyze relationships between multiple numerical variables at once.

            A Pair Plot is used to visualize pairwise relationships and the distribution of numerical variables in a dataset.

     Real-World Use Cases
        Analyze relationships between sales, profit, and discount.
        Identify correlations among financial metrics.
        Explore customer behavior across multiple features.
        Detect clusters, trends, and outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips").head(20)
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue='sex')

plt.show()
```

### Patrticular Columns Can Be Used

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex')

plt.show()
```

```python
sns.pairplot(var,
             vars=['total_bill','tip'],
             hue='sex',
             hue_order=['Female','Male'],   # Gives Order
             palette='BuGn'
            
            )

plt.show() 
```

```python
sns.pairplot(var,
            y_vars=['total_bill','tip'],
            x_vars=['total_bill','tip'],
             hue='sex',
             hue_order=['Female','Male'], 
           
            
            )
plt.show()
```

### kind='reg'
    kind='reg'|'kde'|'scatter'|'hist'|'auto'->Default

```python
sns.pairplot(var,
             hue='sex',
             hue_order=['Female','Male'], 
           kind='reg'
           
            )
plt.show() 
```

```python
 sns.pairplot(var,
             hue='sex',
             hue_order=['Female','Male'], 
           kind='reg',
            diag_kind='hist' # changes In Diagonally
            )
plt.show()
```

```python
sns.pairplot(var,
             hue='sex',
             hue_order=['Female','Male'], 
             markers=['*','>']
            )
plt.show() 
```

```python
g = sns.pairplot(
    data=var,
    vars=['total_bill','tip','size'],
    hue='sex',
    diag_kind='kde',
    palette='dark',
    height=3
)

g.fig.suptitle(
    'Customer Spending Analysis',
    fontsize=16,
    fontweight='bold',
    color='navy'
)

g.fig.subplots_adjust(top=0.92)

plt.show()
```

### corner=True
    Displays only the lower triangle of the pair plot, reducing duplicate plots and making the visualization cleaner.

```python
sns.pairplot(
    data=var,
    vars=['total_bill','tip','size'],
    corner=True
)
plt.show()
```

### Important Parameters
    Parameter	           Purpose
    data	       Dataset
    hue            Group observations by category
    vars	       Select variables to plot
    x_vars	       Variables only on x-axis
    y_vars	       Variables only on y-axis
    diag_kind      Type of diagonal plot (hist, kde)
    palette	       Color palette
    markers	       Marker style for each hue category
    height	       Size of each subplot
    corner	       Show only the lower triangle of the pair plot
