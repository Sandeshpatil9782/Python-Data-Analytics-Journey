```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

np.random.seed(42)
```

```python
sales_data = pd.DataFrame({
    "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Sales":[12000,15000,17000,18000,21000,23000,25000,24500,22000,26000,28000,30000],
    "Profit":[2500,3200,3500,3800,4500,4800,5200,5100,4900,5600,6100,6500],
    "Advertising":[1000,1200,1500,1600,1700,1900,2000,2100,2200,2300,2500,2700],
    "Region":["North","South","East","West","North","South","East","West","North","South","East","West"],
    "Category":["Electronics","Furniture","Clothing","Electronics","Furniture","Clothing",
                "Electronics","Furniture","Clothing","Electronics","Furniture","Clothing"]
})

sales_data.head()
```

### Q1. Which month recorded the highest sales?
    Line Plot

```python

plt.figure(figsize=(8,5))

plt.plot(
    sales_data["Month"],
    sales_data["Sales"],
    marker="o",
    color="blue",
    linewidth=2
)

plt.title("Monthly Sales Trend",fontsize=14,fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)

plt.show()
```

### How are monthly profits distributed?
    Bar Chart

```python

plt.figure(figsize=(8,5))

plt.bar(
    sales_data["Month"],
    sales_data["Profit"],
    color="green"
)

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")

plt.show()
```

### What percentage of sales comes from each region?
    Pie Chart

```python

region_sales = sales_data.groupby("Region")["Sales"].sum()

plt.figure(figsize=(6,6))

plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Sales Contribution by Region")

plt.show()
```

### Is there a relationship between advertising and sales?
    Scatter Plot

```python

plt.figure(figsize=(8,5))

plt.scatter(
    sales_data["Advertising"],
    sales_data["Sales"],
    color="red",
    marker='D',
    s=100,
    edgecolor='black',
    alpha=0.8,
    
)

plt.title("Advertising vs Sales")
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")

plt.show()
```

### How does monthly sales compare?
    Horizontal Bar Chart

```python

plt.figure(figsize=(8,5))

colors = plt.cm.viridis(np.linspace(0, 1, len(sales_data)))

plt.barh(
    sales_data["Month"],
    sales_data["Sales"],
    color=colors
    
)

plt.title("Monthly Sales Comparison")
plt.xlabel("Sales")

plt.show()
```

### Which category has the highest average sales?
Bar Plot

```python

plt.figure(figsize=(8,5))

ax=sns.barplot(
    data=sales_data,
    x="Category",
    y="Sales",
    hue='Category',
    estimator="mean",
    palette="viridis",
    errorbar=None
)
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")


plt.title("Average Sales by Category")
plt.show()
```

### How are sales distributed?
Histogram

```python

plt.figure(figsize=(8,5))

sns.histplot(
    data=sales_data,
    x="Sales",
    bins=8,
    kde=True,
    color="blue"
)

plt.title("Sales Distribution")

plt.show()
```

### Is there a relationship between sales and profit?
Scatter Plot

```python

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=sales_data,
    x="Sales",
    y="Profit",
    hue="Category",
    palette="viridis",
    s=100
)

plt.title("Sales vs Profit")

plt.show()
```

### Which category has greater sales variation?
Box Plot

```python

plt.figure(figsize=(8,5))

sns.boxplot(
    data=sales_data,
    x="Category",
    y="Sales",
    hue='Category',
    palette="dark:r"
)

plt.title("Sales Distribution by Category")

plt.show()
```

### How does advertising relate to sales?
Regression Plot

```python

plt.figure(figsize=(8,5))

sns.regplot(
    data=sales_data,
    x="Advertising",
    y="Sales",
    scatter_kws={"s":80,"color":"blue"},
    line_kws={"color":"red"}
)

plt.title("Advertising vs Sales Trend")

plt.show()
```

### Conclusion

This analysis shows that sales and profits generally increase throughout the year, with the highest sales recorded in December. 
Advertising expenditure has a positive relationship with sales, indicating that increased marketing investment may contribute to higher revenue. 
Category- and region-level comparisons help identify high-performing business segments, supporting better inventory planning and marketing decisions.
