## Seaborn Scatter Plot
    A scatter plot is used to visualize the relationship between two numerical variables and identify patterns, trends, clusters, or outliers in the data.

    Real-World Use Cases
        Analyze the relationship between advertising spend and sales.
        Study customer income vs spending score.
        Examine experience vs employee salary.
        Analyze website visits vs conversion rate.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# The Dataset Cn fine In https://github.com/WalterHua/seaborn-datasets.git

tips = sns.load_dataset('tips').head(70)

tips.head()
```

### Is there a relationship between total bill and tip amount?

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip'
)

plt.title('Total Bill vs Tip')
plt.show()
```

### Differentiate Data by Gender

```python
plt.figure(figsize=(8,5))

m={"Male":'D','Female':'o'}
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    style='sex',
    size='sex',
    sizes=(100,80),  # (Famale,Male)
    palette='cool',
    markers=m
)

plt.title('Total Bill vs Tip by Gender',fontweight='bold',color='navy',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,5))

m={"Yes":'^','No':'s'}
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker',
    markers=m,
    s=120
)

plt.title('Tip Analysis by Gender and Smoking Status',fontweight='bold',color='navy',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,5), facecolor='whitesmoke')

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    size='size',
    style='smoker',
    sizes=(80,300),
    palette='viridis'
)

plt.title('Restaurant Bill Analysis',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')

plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

### Do larger groups spend more?

```python
sns.scatterplot(
    data=tips,
    x='size',
    y='total_bill',
    hue='sex'
)
plt.show()
```

### Does increasing advertisement spending increase revenue?
    
    Insight:
        Higher advertisement spending leads to higher revenue.

```python
marketing = pd.DataFrame({
    'Ad_Spend':[10,15,20,25,30,35,40,45,50],
    'Revenue':[100,120,150,170,200,230,250,280,310]
})
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=marketing,
    x='Ad_Spend',
    y='Revenue',
    s=150,
    color='magenta'
)

plt.title('Advertisement Spend vs Revenue',
          fontsize=15,
          fontweight='bold')

plt.xlabel('Advertisement Spend ($ Thousand)')
plt.ylabel('Revenue ($ Thousand)')


plt.show()
```
