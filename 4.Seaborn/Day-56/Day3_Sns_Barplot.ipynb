## barplot()
    Bar Plot shows the aggregated value (mean by default) of a numerical variable across different categories.

    Real-World Use
        Compare average sales, profit, ratings, or revenue across categories.

```python
plt.figure(figsize=(6,6),facecolor='lightgrey')
sns.barplot(
    data=df,
    x="Class",
    y="Fare",
    color='magenta'
)

plt.title("Average Fare by Passenger Class",fontsize=12,fontweight='bold',color='navy')
plt.show()
```

```python
plt.figure(figsize=(6,6),facecolor='lightgrey')
ords=['Dream','Torgersen','Biscoe']
sns.barplot(data=tips,
            x='island',
            y='bill_length_mm',
            hue='sex',
            order=ords,
            hue_order=['Female','Male'],
            errorbar=('ci',90) , # certificate Interval Range Between 0 to 100 OR ci=90
            n_boot=8,           # Same As CI
           )

plt.show()
```

```python
plt.figure(figsize=(8, 8), facecolor='lightgrey')

# Desired order
ords = ['Dream', 'Torgersen', 'Biscoe']

# Plot
sns.barplot(
    data=tips,
    x='island',
    y='bill_length_mm',
    hue='sex',
    order=ords,
    hue_order=['Female', 'Male'],
    errorbar=('ci', 93),
    err_kws={
        "color": "navy",
        "linewidth": 7
    },
    palette='plasma',
    saturation=1,
    edgecolor='black'
)

# Main Title
plt.title(
    'Average Bill Length by Island and Gender',
    fontsize=18,
    fontweight='bold',
    color='navy',
    pad=20
)

# Axis Labels
plt.xlabel(
    'Island',
    fontsize=13,
    fontweight='bold',
    color='darkred'
)

plt.ylabel(
    'Average Bill Length (mm)',
    fontsize=13,
    fontweight='bold',
    color='darkred'
)

# Tick Customization
plt.xticks(
    fontsize=13,
    rotation=0,
    fontweight='black'
)

plt.yticks(
    fontsize=13,
    fontweight='black'
)

# Grid
plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.6
)

# Legend
plt.legend(
    facecolor='lightgrey',
    title='Gender',
    title_fontsize=12,
    fontsize=11,
    loc=9,
    frameon=True,
    shadow=True
)

# Remove top and right spines for cleaner look
sns.despine()


# Adjust layout
plt.tight_layout()

plt.savefig('Sea_BarPlot.svg', bbox_inches='tight')

plt.show()
```
