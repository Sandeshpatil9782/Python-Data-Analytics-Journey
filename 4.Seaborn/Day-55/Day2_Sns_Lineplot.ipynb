```python
import matplotlib.pyplot as plt

import seaborn as sns
```

## Line Plot
    A line plot is used to visualize trends or changes in numerical data over a continuous variable, typically time.

```python
import pandas as pd

sales_data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [45000, 52000, 48000, 61000, 70000, 75000]
})
```

```python
plt.figure(figsize=(8, 5),facecolor='lightgrey')

sns.lineplot(
    data=sales_data,
    x='Month',
    y='Sales',
    marker='o',
    mfc='red',
    color='navy',
    markersize=13,
    linewidth=3
)

for x, y in zip(sales_data['Month'], sales_data['Sales']):
    plt.annotate(
        text=f'{y}',
        xy=(x, y),
        xytext=(-5,11),
        textcoords='offset points',
        ha='center'
    )

plt.title('Monthly Sales Trend Analysis',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Month', 
           fontsize=12
          )

plt.ylabel('Sales Revenue ($)', 
           fontsize=12
          )



plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

```python
sales_region = pd.DataFrame({
    'Month':['Jan','Feb','Mar','Jan','Feb','Mar'],
    'Sales':[50000,55000,60000,45000,52000,58000],
    'Region':['North','North','North',
              'South','South','South']
})
sns.lineplot(
    data=sales_region,
    x='Month',
    y='Sales',
    hue='Region',
    marker='o',
    ms=10,
    palette='dark',
    linewidth=2
)
plt.show()
```

```python


# Sample Data
loan_data = pd.DataFrame({
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Default_Rate': [2.1, 2.5, 3.0, 2.8]
})

# Set professional theme
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(10, 6), facecolor='whitesmoke')

# Line Plot
ax = sns.lineplot(
    data=loan_data,
    x='Quarter',
    y='Default_Rate',
    marker='o',
    linewidth=3,
    markersize=10,
    color='#1f77b4'
)

# Add data labels
for x, y in zip(loan_data['Quarter'], loan_data['Default_Rate']):
    plt.annotate(
        text=f'{y}%',
        xy=(x, y),
        xytext=(0,10),
        textcoords='offset points',
        ha='center',
        fontsize=10,
        fontweight='bold',
        color='darkred'
    )

# Title and subtitle
plt.title(
    'Quarterly Loan Default Rate Trend',
    fontsize=18,
    fontweight='bold',
    color='navy',
    pad=20
)

#plt.suptitle(
   # 'Monitoring Credit Risk Across Business Quarters',
    #fontsize=11,
   # color='dimgray',
    #y=0.93)

# Axis labels
plt.xlabel(
    'Quarter',
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    'Default Rate (%)',
    fontsize=13,
    fontweight='bold'
)

# Highlight highest default quarter
max_rate = loan_data['Default_Rate'].max()
max_quarter = loan_data.loc[
    loan_data['Default_Rate'].idxmax(), 'Quarter'
]



plt.annotate(
    'Highest Default',
    xy=(max_quarter, max_rate),
    xytext=(40, 20),
    textcoords='offset points',
    arrowprops=dict(
        arrowstyle='->',
        color='red',
        lw=2
    ),
    fontsize=10,
    fontweight='bold',
    color='red'
)

# Grid
plt.grid(linestyle='--', alpha=0.7)

# Remove extra borders
sns.despine()

# Tight layout
plt.tight_layout()


plt.show()
```
