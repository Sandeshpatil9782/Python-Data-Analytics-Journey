## Seaborn

    Seaborn is a Python statistical data visualization library built on top of Matplotlib that provides a high-level interface for creating attractive and informative visualizations.

    Real-World Use
    Data Analysts use Seaborn for:
        Exploratory Data Analysis (EDA)
        Business reporting
        Trend analysis
        Customer analysis

### Built-in Datasets
    Built-in datasets are sample datasets provided by Seaborn for learning, testing, and practicing visualizations.

    Examples:
    sns.load_dataset("tips")
    sns.load_dataset("titanic")
    sns.load_dataset("iris")

### Themes (set_theme())

    Themes are predefined styles that control the overall appearance of plots, including colors, grids, and backgrounds.
    sns.set_theme()

### Styles (set_style())
    Styles define the background and grid appearance of Seaborn plots.

    Available styles:white,dark,whitegrid,darkgrid,ticks

### Context (set_context())
    Context controls the scaling of plot elements such as labels, titles, and lines for different presentation environments.

    Options:paper,notebook,talk,poster

### Color Palette
    A color palette is a collection of colors used to represent categories or numerical values in a visualization.

```python
plt.figure(figsize=(5,5),facecolor='lightgrey')

sns.set_theme(style="white")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue='sex',
    s=200
)

plt.title("Total Bill vs Tip by Gender",fontsize=15,color='blue',fontweight='bold')
plt.xlabel("Total Bill ($)",fontsize=15,color='navy',fontweight='bold')
plt.ylabel("Tip ($)",fontsize=12,color='navy',fontweight='bold')

plt.show()
```

```python
plt.figure(figsize=(5,5),facecolor='lightgrey')

sns.set_theme(style="ticks")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue='sex',
    style='smoker',
    s=200
)

plt.title("Customer Spending by Smoking Status",fontsize=15,color='blue',fontweight='bold')
plt.xlabel("Total Bill ($)",fontsize=15,color='navy',fontweight='bold')
plt.ylabel("Tip ($)",fontsize=12,color='navy',fontweight='bold')

plt.show()
```

```python
plt.figure(figsize=(8, 5),facecolor='lightgrey')

sns.lineplot(
    data=tips,
    x="size",
    y="tip",
    color='navy',
    marker="o",
    ms=15,
    mfc='green',
    mec='black'
    linewidth=3
)

plt.title("Average Tip by Group Size",fontsize=15,color='blue',fontweight='bold')
plt.xlabel("Group Size",fontsize=12,color='navy',fontweight='bold')
plt.ylabel("Average Tip",fontsize=12,color='navy',fontweight='bold')

plt.show()
```
