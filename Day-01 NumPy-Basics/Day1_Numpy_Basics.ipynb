### Importing NumPy Library

```python
import numpy as np
```

###  Creating Array

```python
marks=np.array([45,67,89,90,34,58,90])
```

```python
print("Marks:",marks)
```

### Average Marks

```python
Avg=np.mean(marks)
```

```python
print("Average Marks:",Avg)
```

###  Maximum Value

```python
M=np.max(marks)
```

```python
print("Highest Marks:",M)
```

#### Python Basic max using list

```python
max([89,70,23])
```

### Minimum Value

```python
m=np.min(marks)
```

```python
print("Lowest Marks:",m)
```

#### Python Basic min using list

```python
min([20,4,6])
```

### Shape Value 
 Count the Number of Elements

```python
print("Shape:",marks.shape)
```

### Dimensions 
 Shows dimensions of array

```python
print("Dimensions:",marks.ndim)
```

### Data Type
 Shows types of data stored in Array


```python
print("Data Type:",marks.dtype)
```

## Extra Tasks

### SUM 
 Calculate total sum of array values

```python
Total=np.sum(marks)
```

```python
print("Total Sum:",Total)
```

### Length
 count number of elements Or same as Shape
 

```python
L=len(marks)
```

```python
print("Length:",L)
```
