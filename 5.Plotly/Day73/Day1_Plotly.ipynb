# PLOTLY

## Plotly is an interactive Python visualization library used to create dynamic, zoomable, hover-enabled, and presentation-ready charts for Exploratory Data Analysis (EDA), dashboards, and business reporting.

     Real-World Use Cases
        Interactive business dashboards
        Sales and revenue analysis
        Financial reporting
        Customer behavior analysis
        Time-series visualization

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
sales_data = pd.DataFrame({
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[45000,52000,61000,58000,67000,72000],
    "Profit":[7000,8500,10500,9800,12000,13500],
    "Orders":[120,145,170,160,190,210],
    "Region":["North","South","East","West","North","South"]
})

sales_data
```

### Line Chart

```python
fig = px.line(
    sales_data,
    x="Month",
    y="Sales",
    title="Monthly Sales Trend",
   
)

fig.show()
```

### Bar Chart

```python
fig = px.bar(
    sales_data,
    x="Month",
    y="Sales",
    color="Month"
)

fig.show()
```

### Scatter Plot

```python
fig = px.scatter(
    sales_data,
    x="Sales",
    y="Profit",
    size="Orders",
    color="Region"
)

fig.show()
```

### Pie Chart

```python
fig = px.pie(
    sales_data,
    names="Region",
    values="Sales"
)

fig.show()
```

### Donut Chart

```python
fig = px.pie(
    sales_data,
    names="Region",
    values="Sales",
    hole=0.5
)

fig.show()
```

### Area Chart

```python
fig = px.area(
    sales_data,
    x="Month",
    y="Sales"
)

fig.show()
```

### Histogram

```python
fig = px.histogram(
    sales_data,
    x="Sales"
)

fig.show()
```

### Box Plot

```python
fig = px.box(
    sales_data,
    y="Sales",
    points="all"
)

fig.show()
```

### Violin Plot

```python
fig = px.violin(
    sales_data,
    y="Sales",
    box=True,
    points="all"
)

fig.show()
```

### Bubble Chart

```python
fig = px.scatter(
    sales_data,
    x="Sales",
    y="Profit",
    size="Orders",
    color="Month"
)

fig.show()
```

## Parameter	    ->            Purpose
    title	           Chart title
    color	           Color grouping
    size	           Bubble size
    hover_data	       Extra hover information
    labels	           Rename axis labels
    template	       Apply chart theme
    width	           Chart width
    height	           Chart height

### Hover Customization

    Definition: Customize information displayed when hovering over data points.

    hover_name="Region"

    hover_data=[
        "Sales",
        "Profit",
        "Orders"
    ]

    hover_data={
        "Sales":":,.0f",
        "Profit":":,.0f"
    }

## Facet Charts

    Definition: Split one chart into multiple small charts.

    facet_row="Region"

    facet_col="Category"

    facet_col_wrap=2

### Animation
    animation_frame="Year"
    animation_group="Region"

### Marginal Plots
    marginal_x="histogram"
    marginal_y="box"
    marginal_x="violin"
    marginal_y="rug"
    
## Trendline
    trendline="ols"
    trendline_color_override="red"

## Error Bars
    error_y="Profit"
    error_x="Sales"

### Color Scale
    color="Profit"
    color_continuous_scale="Viridis"
    color_continuous_scale="RdBu"
    color_discrete_sequence=px.colors.qualitative.Set2

### Bubble Chart
    size="Orders"
    size_max=60

### Symbols
    symbol="Category"

### Category Ordering
    category_orders={
         "Month":["Jan","Feb","Mar","Apr"]
        }

### Figure Size
    width=1000
    height=600

### Axis Customization
    fig.update_xaxes()
    fig.update_yaxes()
    
    Example
        fig.update_xaxes(title="Month")
        fig.update_yaxes(title="Sales")

### Layout Customization
    fig.update_layout(
        title="Sales Dashboard",
        title_x=0.5,
        font=dict(size=16),
        legend_title="Region"
    )

### Add Annotation
    fig.add_annotation(
        text="Highest Sales",
        x="Jun",
        y=72000,
        showarrow=True
    )
### Shapes
    fig.add_hline()
    fig.add_vline()
    fig.add_hrect()
    fig.add_vrect()

### Export Chart
    fig.write_html("dashboard.html")
    fig.write_image("sales.png")

```python
fig = px.scatter(
    sales_data,
    x="Sales",
    y="Profit",
    color="Region",
    size="Orders",
    hover_name="Month",
    hover_data=["Sales"],
    trendline="ols",
    template="ggplot2",
    title="Sales vs Profit Analysis"
)

fig.update_layout(
    title_x=0.5,
    width=1000,
    height=600
)

fig.show()
```
