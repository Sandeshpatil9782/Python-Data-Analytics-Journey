# 📊 Statistics for Data Analysis

## What is Statistics?

**Statistics** is the science of collecting, organizing, analyzing, interpreting, and presenting data to make informed decisions.

In simple words, statistics helps us turn **raw data into meaningful insights**.

---

## Why is Statistics Important?

Statistics helps us to:

- Understand data
- Identify patterns and trends
- Compare different groups
- Make data-driven decisions
- Predict future outcomes

---

## Real-World Example

Imagine an e-commerce company wants to know the **average purchase amount** of its customers.

This is a practical application of **Statistics**.

# Population vs Sample
---
## Population :-The complete collection of all individuals, observations, or items of interest in a study.

## Sample :- A Subset of the population selected for analysis
            Which is more commonly analyzed in real companies?
                 Samples, because analyzing the entire population is often impractical.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Population Data

### This dataset contains the purchase amounts of **10,000 customers** from an e-commerce company. It represents the **entire population** and is used to understand the concept of **Population** and **Sample** in statistics.

```python
np.random.seed(42)

population = np.random.normal(
    loc=5000,
    scale=1200,
    size=10000
)

population = pd.DataFrame({
    "Purchase_Amount": population
})

population.head()
```

### Business Insight
    - The dataset represents all 10,000 customers, making it the population.
    - Customer purchase amounts are different because every customer has unique buying behavior.
    - Analyzing the entire population gives the most accurate insights, but it may not always be practical for very large datasets.
    - This population dataset will be used to draw a random sample in the next step.

### Understand The Population

```python
population.shape
```

### Calculate Statistic

```python
population.describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(population["Purchase_Amount"], bins=30)

plt.title("Population Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Customers")

plt.show()
```

### Take Random Sample from Population
    Take 300 Customers

```python
sample = population.sample(
    n=300,
    random_state=42
)
pd.set_option('display.max_rows', 20)
sample
```

```python
sample.shape
```

```python
sample.describe()
```

### Compare population Mean Vs Sample Mean

```python
population_mean = population["Purchase_Amount"].mean()

sample_mean = sample["Purchase_Amount"].mean()

print("Population Mean :", population_mean)

print("Sample Mean :", sample_mean)


print('\nObservation:\n           The sample mean should be very close to the population mean.')
```

## Compare Using Bar Chart

```python
plt.figure(figsize=(6,6))

plt.bar(
    ["Population","Sample"],
    [population_mean,sample_mean],
    color=['red','green'],
    #label=['Popuation','Sample']

)

plt.ylabel("Average Purchase")

plt.title("Population vs Sample Mean")

#plt.legend()

plt.show()
```

```python
print(f"Population Size : {len(population)}")

print(f"Sample Size : {len(sample)}")

print(f"Population Mean : {population_mean:.2f}")

print(f"Sample Mean : {sample_mean:.2f}")
```

### Conclusion
    In this step, we created a population dataset of 10,000 customers. It represents the complete group of customers under study.
