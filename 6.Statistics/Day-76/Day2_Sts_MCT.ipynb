## What is Statistics?

Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data to make informed decisions.

### Real-World Example

An e-commerce company wants to understand how much customers spend on average. Instead of looking at thousands of transactions individually, the company uses statistics to summarize and analyze the data.

### Types of Statistics

Statistics is divided into two main types:

1. Descriptive Statistics
2. Inferential Statistics

### 📊 Business Insight

Descriptive Statistics helps businesses understand past and current performance, while Inferential Statistics helps predict future outcomes and supports data-driven decision-making.

### 1️⃣ Descriptive Statistics

### Definition

Descriptive Statistics summarizes and describes the characteristics of a dataset. It helps us understand what the data looks like without making predictions.

### Business Example

A retail company analyzes the sales data of the last month and calculates:

- Average Sales
- Highest Sales
- Lowest Sales

These values describe the existing data.

### Examples

- Mean
- Median
- Mode
- Range
- Standard Deviation

### 2️⃣ Inferential Statistics

### Definition

Inferential Statistics uses a sample of data to make predictions or draw conclusions about the entire population.

### Business Example

A company has 50,000 customers. Instead of surveying every customer, it collects feedback from 1,000 randomly selected customers and estimates customer satisfaction for the entire population.

### Examples

- Hypothesis Testing
- Confidence Interval
- Regression
- ANOVA

### Difference Between Descriptive and Inferential Statistics

```python
import pandas as pd

comparison = pd.DataFrame({
    "Descriptive Statistics": [
        "Summarizes Data",
        "Uses Existing Data",
        "No Prediction",
        "Example: Average Sales"
    ],
    "Inferential Statistics": [
        "Makes Predictions",
        "Uses Sample Data",
        "Predicts Population",
        "Example: Customer Survey"
    ]
})

comparison
```

### Measures of Central Tendency

### Definition

Measures of Central Tendency are statistical measures that represent the center or typical value of a dataset.

Instead of analyzing every value individually, they provide a single value that best represents the entire dataset.

### Types

- Mean
- Median
- Mode

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
purchase = pd.DataFrame({
    "Customer_ID": range(1,16),
    "Purchase_Amount":[1200,1500,1800,1500,2200,
                       2500,1800,3000,1500,2700,
                       3500,1800,1500,4000,2500]
})
```

```python
purchase.shape
```

```python
purchase.head()
```

```python
purchase.describe()
```

```python
purchase.info()
```

### Mean

## Definition

Mean is the arithmetic average of all observations.

Formula:

Mean = Sum of all values / Number of observations

```python
me=purchase["Purchase_Amount"].mean()

print(me)
```

### Median

## Definition

Median is the middle value after arranging the data in ascending order.

It is useful when the dataset contains outliers.

```python
med=purchase["Purchase_Amount"].median()

print(med)
```

### Mode

## Definition

Mode is the value that occurs most frequently in a dataset.

It is useful for identifying the most common observation.

```python
mod=purchase["Purchase_Amount"].mode()[0]

print(mod)
```

```python
comparison = pd.DataFrame({
    "Measure":["Mean","Median","Mode"],
    "Value":[
        purchase["Purchase_Amount"].mean(),
        purchase["Purchase_Amount"].median(),
        purchase["Purchase_Amount"].mode()[0]
    ]
})

comparison
```

```python
plt.figure(figsize=(6,4))

plt.hist(purchase["Purchase_Amount"],
         bins=7,
        color='g',
        edgecolor='black'
        )

plt.axvline(m, color='red', linestyle='--', linewidth=2, label=f"Mean = {me:.0f}")
plt.axvline(med, color='blue', linestyle='-.', linewidth=2, label=f"Median = {med:.0f}")
plt.axvline(mod, color='orange', linestyle=':', linewidth=3, label=f"Mode = {mod:.0f}")

plt.title("Distribution of Purchase Amount")

plt.xlabel("Purchase Amount")

plt.ylabel("Frequency")

plt.legend()

plt.show()
```
