## HANLING MISSING DATAS
      Deal with missing values

```python
import pandas as pd

import numpy as np
```

```python
# Adding Null Value np.nan
data = {
    'Name':['Ram','Sham','Amit'],
    'Marks':[85,np.nan,70],          
    'Age':[21,22,np.nan]
       }                       

df = pd.DataFrame(data)

print(df)
```

### isnull()
    Checks missing values

```python
print(df.isnull())
```

### Count missing value

```python
print(df.isnull().sum())
```

### notnull()
    Checks available values

```python
print(df.notnull())
```

### dropna()
    Removes missing values rows

```python
print(df.dropna())
```

### fillna()
    Fills missing values

```python
print(df.fillna(0))
```

### Fill Using mean

```python
df['Marks'] = df['Marks'].fillna(
    df['Marks'].mean()
)

print(df)
```

### replace()
    Replace Values

```python
print(df.replace(70,75)) # Change 70 To 80
```

### interpolate()
    Missing Values Fills Approximately

```python
df = df.infer_objects(copy=False) # It Works Mainly On Numeric Data

print(df.interpolate())
```
