## PANDAS-DATAFRAME
    DataFrame is a 2-dimensional labeled table.

```python
import pandas as pd
```

### Create Dataframe

```python
import pandas as pd

data = {
    'Name':['Ram','Sham','Amit'],
    'Age':[21,22,20],
    'Marks':[85,90,88]
}

df = pd.DataFrame(data)

print(df)
```

### Access Column

```python
print(df['Name'])
```

### Access Multiple Column

```python
print(df[['Name','Marks']])
```

### Access Row

```python
print(df.loc[1]) #Index Row 1 displayed
```

```python
print(df.iloc[-1]) # Access Last Row
```

### head() Function
    Show First Rows

```python
df.head(2)
```

### tail() Function
    Shows Last Rows

```python
df.tail(2)
```

### Shape() Function
    Returns rows and columns count

```python
df.shape
```

### Column Function
Gives Column Names

```python
print(df.columns)
```

### info() function
    Gives Summury of dataframe


```python
print(df.info())
```

### describe Function
    Gives statistics of numerical column

```python
print(df.describe())
```

```python
data = {
    'Name':['Ram','Sham','Amit','Riya'],
    'Age':[21,22,20,23],
    'Marks':[85,90,70,95]
}

df1 = pd.DataFrame(data)

print(df1)
```

### loc[]
    Access data using labels/index names

```python
print(df1.loc[1])
```

```python
print(df1.loc[3,'Name'])  # Row + Column Labeled
```

### iloc[]
    Access data using integer position
    Syntax:-iloc[row_position , column_position]

```python
print(df1.iloc[0]) # Gives 2 Index Data
```

```python
print(df1.iloc[2,1]) # Row + Column Position
```

### Filtering Data

```python
print(df1[df1['Marks'] > 80]) # Displaying Values Where Marks Greater Than 80
```

#### Multiple Condition
    AND

```python
print(
    df1[
        (df1['Marks'] > 80) &
        (df1['Age'] > 21)
    ]
)
```

#### OR Condition

```python
print(
    df1[
        (df1['Marks'] > 90) |
        (df1['Age'] < 21)
    ]
)
```

#### Filtering Specific Column

```python
print(
    df1.loc[
        df1['Marks'] > 70,
        ['Name','Age','Marks']
    ]
)
```

```python
data = {
    'Name':['Ram','Sham','Amit','Raj'],
    'Marks':[85,90,70,80]
}

df = pd.DataFrame(data)

print(df)
```

### Add New Column

```python
df['Grade'] = ['A','A','B',"A"]

print(df)
```

### Update Column

```python
df['Marks'] = df['Marks'] + 5 # Marks 5 Added To Each Student
print(df)
```

#### Update Specific Value

```python
df.loc[2,'Marks'] = 60 # Marks Added To Index 2 Amit

print(df)
```

### Delete Column
    drop() Function

```python
df.drop('Grade', axis=1, inplace=True) # Grade Column Deleted Permenet From Dataframe

print(df)
```

### Delete Row


```python
df=df.drop(2) 
df
```

### Delete Multiple Rows


```python
df = df.drop([0,1])
df
```

### Rename Column
    rename() Function

```python
df.rename(
    columns={'Marks':'Score'},
    inplace=True
)

print(df)
```

### Insert Column
    insert(position, column_name, values)

```python
df.insert(1,'Age',[21,22,20,25])

print(df)
```
