### Handling Missing Data

#### Filling NAN Values

##### Purchase_Amount Column

```python
df["Purchase_Amount"] = df["Purchase_Amount"].fillna(
    df["Purchase_Amount"].median()
)
```

```python
df["Purchase_Amount"].unique()
```

##### Email Column

```python
df["Email"] = df["Email"].fillna("Not Available")
```

```python
df["Email"].unique()
```

#### Join_Date Column

```python
df["Join_Date"].unique()
```

##### Fill With Most frequent Date

```python
df["Join_Date"] = df["Join_Date"].fillna(
    df["Join_Date"].mode()[0]
)
```

```python
df["Join_Date"].unique()
```

#### Phone Column

```python
df["Phone"].unique()
```

```python
df["Phone"] = df["Phone"].replace(
    "98765-43210","9876543210"
)
```

```python
df["Phone"] = df["Phone"].fillna("0")
```

```python
df["Phone"].unique()
```

### Fixing Inconsistent Formating

#### Convert Email Lowercase

```python
df["Email"].unique()
```

```python
df["Email"] = df["Email"].str.lower()
```

#### Removing Spaces

```python
df["Purchase_Amount"] = df["Purchase_Amount"].astype(str).str.strip()
```

#### Replacing Values

```python
df["Purchase_Amount"] = df["Purchase_Amount"].replace(
    "five thousand","5000"
)
```

#### Checking Unique Values

```python
df["Purchase_Amount"].unique()
```

```python
df[df.duplicated(subset=["Email"], keep=False)]

df.head(10)
```

```python
df["Customer_Name"] = df["Customer_Name"].str.title()
df.head(5)
```

```python
df["City"] = df["City"].str.title()
df.head(5)
```

```python
df["Customer_Segment"] = df["Customer_Segment"].str.title()
df.head(5)
```

```python
df.columns = df.columns.str.strip()
```

```python
df = df.apply(
                lambda x: x.str.strip() if x.dtype == "object" else x
            )
```
