## PANDAS STRING FUNCTION
    Handling Text Data

```python
import pandas as pd
```

```python
data = {
    'Name':['ram',' sham ','AMIT'],
    'City':['delhi','mumbai','bangalore']
}

df = pd.DataFrame(data)

print(df)
```

### Upper :- Converts into uppercase

```python
print(df['Name'].str.upper())
```

### Lower :- Converts into lowercase

```python
print(df['Name'].str.lower())
```

### Title :- Gives First letter Capital

```python
print(df['Name'].str.title())
```

### Strip :- Remove Extra Spaces

```python
print(df['Name'].str.strip())
```

### Contains :- Ckecks word/character present or not

```python
print("PRESENT:-\n",df['Name'].str.contains('Aa'))

print("NOT PRESENT:-\n",df['Name'].str.contains('del')) 
```

### Replace :- Replace the text/char

```python
print(df['City'].str.replace('delhi','Delhi'))
```

### len :- Counts Character of 
### Count :- count thhe number of given character are present

```python
df['Name'].str.count('a')
```

```python
print(df['Name'].str.count('ram ')) # using count()
```

### Splits :- Splits The word

```python
data = {
    'FullName':['Ram Patil','Sham Rao']
}

df = pd.DataFrame(data)

print(
    df['FullName'].str.split()
)
```

### startswith
    Which latter start

```python
df['Name'].str.startswith('r')
```

### endswith()
    Which latter ends

```python
df['Name'].str.endswith('T')
```

### Capitalize()
    First latter capital

```python
df['Name'].str.capitalize()
```

### cat()
    String Join

```python
data = {
    'Name':['ram',' sham ','AMIT'],
    'City':['delhi','mumbai','bangalore']
}

df = pd.DataFrame(data)

df['Name'].str.cat(sep =" ")
```

```python
# Join two columns
df = pd.DataFrame({
    "First": ["Sandesh", "Rahul"],
    "Last": ["Patil", "Sharma"]
})

print(df["First"].str.cat(df["Last"], sep=" "))
```

### find()
    Finds the position of elememt or word

```python
df = pd.DataFrame({
    "Text": ["Python", "12345", "Python123", "Hello@123"]
})
print(df['Text'].str.find('thon'))
```

### isnumeric()
    Finds number is there or not

```python
df = pd.DataFrame({
    "Text": ["Python", "12345", "Python123", "Hello@123"]
})
print(df['Text'].str.isnumeric())
```

### isalpha()
    Finds alphabet is there or not

```python
df = pd.DataFrame({
    "Text": ["Python", "12345", "Python123", "Hello@123"]
})
print(df['Text'].str.isalpha())
```

### isalnum()
    Checks for both alphabet and numbers

```python
df = pd.DataFrame({
    "Text": ["Python", "12345", "Python123", "Hello@123"]
})
print(df['Text'].str.isalnum())
```

### lstrip()
    Remove left side spaces of string

```python
df = pd.DataFrame({
    "Text": ["   Python", "   Pandas", "   Data"]
})

print(df["Text"].str.lstrip())
```

### rstrip()
    Remove right side spaces of string

```python

df = pd.DataFrame({
    "Text": ["Python     ", "Pandas     ", "Data     "]
})

print(df["Text"].str.rstrip())
```

### Q1 Clean Extra Spaces from DataFrame

```python
df = pd.DataFrame({
    "Names": ["  Sandesh  ", "  Rahul", "Priya  "]
})

print(df["Names"].str.strip())
```

### Q2 Count Total Characters

```python
df = pd.DataFrame({
    "Text": ["Python", "Pandas", "DataScience"]
})

print(df["Text"].str.len())
```

### Q3 Find Invalid Usernames

```python
df = pd.DataFrame({
    "Username": ["Sandesh123", "Rahul@", "Python_99", "Admin123"]
})

print(df["Username"].str.isalnum())
```

### Q4 Remove Special Characters

```python
df = pd.DataFrame({
    "Text": ["Python@123", "Data#Science", "AI!ML"]
})

print(df["Text"].str.replace(r'[^A-Za-z0-9 ]', '', regex=True))
```

### Q5 Create Professional Email IDs

```python
df = pd.DataFrame({
    "Name": ["sandesh", "rahul", "priya"]
})

print(df["Name"].str.cat(["@gmail.com"] * len(df)))
```

### Q6 Extract Mobile Number Last 4 Digits

```python
df = pd.DataFrame({
    "Mobile": ["9876543210", "9123456780", "9988776655"]
})

print(df["Mobile"].str[-4:])
```
