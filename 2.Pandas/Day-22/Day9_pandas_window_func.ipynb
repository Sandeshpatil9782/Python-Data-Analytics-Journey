## Pandas Window Function
    - Moving average & analysis
    - Perform calculation on group of rows

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Day": [1,2,3,4,5,6,7,8],
    
    "Sales": [100,120,140,130,150,170,160,180],
    
    "Profit": [20,25,30,28,35,40,38,45]
}

df = pd.DataFrame(data)

df
```

### rolling()
    Creates Moving Window Calculation

### Rolling Mean

```python
df["Rolling_Mean"] = df["Sales"].rolling(3).mean()

df
```

### Rolling Sum

```python
df["Rolling_Sum"] = df["Sales"].rolling(3).sum()

df
```

### Rolling Max

```python
df["Rolling_Max"] = df["Sales"].rolling(3).max()

df
```

### Rolling Min

```python
df["Rolling_Min"] = df["Sales"].rolling(3).min()

df
```

### Rolling Count

```python
df["Rolling_Count"] = df["Sales"].rolling(3).count()

df
```

### Rolling STD

```python
df["Rolling_STD"] = df["Sales"].rolling(3).std()

df
```

### Rolling VAR

```python
df["Rolling_Var"] = df["Sales"].rolling(3).var()

df
```

### min_periods
    Starts calculation before full window size

```python
df["Rolling_Mean_2"] = df["Sales"].rolling(
    window=3,
    min_periods=1
).mean()

df
```

### center=TRUE
    Places result in center row

```python
df["Center_Mean"] = df["Sales"].rolling(
    3,
    center=True
).mean()

df
```

### expanding()
    Used for cumulative calculation

```python
data = {
    "Day": [1,2,3,4,5,6,7,8],
    
    "Sales": [100,120,140,130,150,170,160,180],
    
    "Profit": [20,25,30,28,35,40,38,45]
}

df1 = pd.DataFrame(data)

df1
```

### Expanding Sum

```python
df1["Expanding_Sum"] = df1["Sales"].expanding().sum()

df1
```

### Expanding Mean

```python
df1["Expanding_Mean"] = df1["Sales"].expanding().mean()

df1
```

### Expanding Max

```python
df1["Expanding_Max"] = df1["Sales"].expanding().max()

df1
```

### Expanding Min

```python
df1["Expanding_Min"] = df1["Sales"].expanding().min()

df1
```

### ewm()
    -Exponentially weighted function
    -Recent values get more importance

```python
EWM MEAN
```

```python
data = {
    "Day": [1,2,3,4,5,6,7,8],
    
    "Sales": [100,120,140,130,150,170,160,180],
    
    "Profit": [20,25,30,28,35,40,38,45]
}

df2 = pd.DataFrame(data)

df2
```

```python
df2["EWM_Mean"] = df2["Sales"].ewm(span=3).mean()

df2
```

### EWM STD

```python
df2["EWM_STD"] = df2["Sales"].ewm(span=3).std()

df2
```

### SHIFT()
    Moves values up or down

### SHIFT DOWN

```python
df2["Shift_Down"] = df2["Sales"].shift(1)

df2
```

### SHIFT UP

```python
df2["Shift_Up "]= df2["Sales"].shift(-1)

df2
```

### Difference Using diff()
    Finds difference between rows

```python
df2["Difference"] = df2["Sales"].diff()

df2
```

### Percent Change

```python
df2["Percent_Change"] = df2["Sales"].pct_change()

df2
```

### CUMULATIVE FUNCTIONS

### cumsum()

```python
data = {
    "Day": [1,2,3,4,5,6,7,8],
    
    "Sales": [100,120,140,130,150,170,160,180],
    
    "Profit": [20,25,30,28,35,40,38,45]
}

df3 = pd.DataFrame(data)

df3
```

### cumsum()

```python
df3["Cumulative_Sum"] = df3["Sales"].cumsum()

df3
```

### cummax()

```python
df3["Cumulative_Max"] = df3["Sales"].cummax()

df3
```

### cummin()

```python
df3["Cumulative_Min"] = df3["Sales"].cummin()

df3
```

### cumprod()

```python
df3["Cumulative_Product"] = df3["Sales"].cumprod()

df3
```

### GROUPBY + WINDOW FUNCTION

```python
data2 = {
    "Department": ["IT","IT","HR","HR","Finance","Finance"],
    
    "Salary": [50000,60000,45000,52000,70000,65000]
}

df2 = pd.DataFrame(data2)

df2
```

### Group Rolling Mean

```python
df2["Rolling_Mean"] = df2.groupby(
    "Department"
)["Salary"].rolling(2).mean().reset_index(level=0, drop=True)

df2
```
