## METHOD CHAINING & VECTORIZATION

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Aman", "Rahul", "Sneha", "Priya"],
    
    "Department": ["IT", "HR", "Finance", "IT"],
    
    "Salary": [50000, 60000, 70000, 80000],
    
    "Bonus": [5000, 6000, 7000, 8000]
}

df = pd.DataFrame(data)

df
```

### CHAINING 
    Means Connecting Multiple Methods Together.

### Without Chaining

```python
df1 = df.sort_values(by="Salary")

df2 = df1.head(2)

df2
```

### With Chaining

```python
df.sort_values(
    by="Salary"
).head(2)
```

### Multiple Chaining

```python
df.query(
    "Salary > 55000"
).sort_values(
    by="Salary"
).head(2)
```

### Chaining With assign()
    create new column during chaining

```python
df.assign(
    Tax = df["Salary"] * 0.1
)
```

### Chaining With rename()


```python
df.rename(
    columns={"Salary":"Income"}
).head()
```

### Chaining With drop()

```python
df.drop(
    columns=["Bonus"]
).head()
```

### Chaining with pipe()

```python
def add_bonus(data):
    
    data["Salary"] = data["Salary"] + 5000
    
    return data
```

```python
df.pipe(add_bonus)
```

### Advance Chaining

```python
(
    df
    .query("Salary > 50000")
    .assign(Tax = lambda x: x["Salary"] * 0.1)
    .sort_values(by="Salary")
)
```

## VECTORIZATION
    Performs operations on entire column at once.

### Simple Addition

```python
df["Salary"] + 1000
```

### Vector Multiplecation

```python
df["Salary"] * 2
```

### Creation Of New Column

```python
df["Total"] = df["Salary"] + df["Bonus"]

df
```

### Vector Comparison

```python
df["Salary"] > 60000

```

### Vector filtering

```python
df[df["Salary"] > 60000]
```

### Vector String Operation
    str :- String Vector

```python
df["Name"].str.upper()
```

### String length

```python
df["Name"].str.len()
```

### Vector Date Operations
    dt :- Date Vector

```python
date_df = pd.DataFrame({
    "Date": pd.to_datetime([
        "2025-01-01",
        "2025-02-01"
    ])
})

date_df
```

### Extract Year

```python
date_df["Date"].dt.year
```

### Extract Month

```python
date_df["Date"].dt.month
```

### Numpy Vectorization

```python
arr = np.array([1,2,3,4])

arr * 10
```

### where()

```python
df["Status"] = np.where(
    df["Salary"] > 60000,
    
    "High",
    
    "Low"
)

df
```

### Vector Conditional

```python
df["Bonus_10%"] = df["Salary"] * 0.10

df
```

### apply() VS Vectorization

```python
print("Using Apply:-\n",df["Salary"].apply(lambda x: x + 1000))


print("Using Vectorization:-\n",df["Salary"] + 1000) # Vectorization,its Faster
```

### Loop VS Vectorization

```python
result = []

for i in df["Salary"]:
    
    result.append(i + 1000)

print("Using Loop:-",result,"\n")

print("Using Vectorization:-\n",df["Salary"] + 1000) # Vectorization less Code + High Performance
```

### Performance Example

### eval()
    Fast Chaining Operations

```python
df.eval(
    "Total = Salary + Bonus"
)
```

### query()

```python
(
    df
    .query("Salary > 50000")
    .sort_values(by="Salary")
)
```

### np.select()

```python
conditions = [
    df["Salary"] > 70000,
    df["Salary"] > 50000
]

choices = ["High", "Medium"]

df["Level"] = np.select(
    conditions,
    choices,
    default="Low"
)

df
```

### clip()
    Limits values between min AND mav Boundary

```python
df["Salary"].clip(
    lower=40000,
    
    upper=70000
)
```

### Chaining With GroupBy

```python
(
    df
    .groupby("Department")
    ["Salary"]
    .mean()
    .sort_values()
)
```
