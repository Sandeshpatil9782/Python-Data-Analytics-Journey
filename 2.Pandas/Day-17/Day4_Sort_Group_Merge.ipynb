### Pandas Sorting & Duplicate Handling

```python
import pandas as pd
```

```python
data = {
    'Name':['Ram','Sham','Ram','Amit'],
    'Marks':[85,90,85,70]
}

df = pd.DataFrame(data)

print(df)
```

### sort_values()
    Sort values In Ascending

```python
print(
    df.sort_values('Marks')
)
```

### Descending

```python
print(
    df.sort_values('Marks',ascending=False)
)
```

### sort_index()
    Sorts the Index

```python
print(df.sort_index())
```

### duplicated()
    Checks duplicate rows

```python
print(df.duplicated()) # 2 Index Row is duplicate
```

### drop_duplicates()
    Removes duplicate rows

```python
print(df.drop_duplicates())
```

### unique()
    Shows unique values

```python
print(df['Name'].unique())
```

### value_counts()
    Counts occurrences

```python
print(df['Name'].value_counts())
```

## Aggregating Data
    Summarize your Data

```python
import pandas as pd

data = {
    'Department':['IT','HR','IT','Sales','HR'],
    'Employee':['Ram','Sham','Amit','Riya','Kiran'],
    'Salary':[50000,40000,60000,45000,42000]
}

df = pd.DataFrame(data)

print(df)
```

### groupby()
    Groups data based on common values

#### Group by  Salary Sum

```python
print(
    df.groupby('Department')['Salary'].sum()
)
# department wise salary added
```

### Group by Mean

```python
print(
    df.groupby('Department')['Salary'].mean()
) 
# department wise average salary
```

### Group by Count

```python
print(
    df.groupby('Department')['Employee'].count()
)
# department wise employee count
```

### Multiple Aggregation

```python
print(
    df.groupby('Department')['Salary']
    .agg(['sum','mean','max'])
)
```

### Group by multiple column

```python
print(
    df.groupby(['Department','Employee'])
    ['Salary'] 
    .agg(['sum'])
)
```

### Reset Index
    Convert grouped index into columns

```python
result = df.groupby('Department')['Salary'].sum()
print(result,"\n")

print(result.reset_index())
```

## Pandas Merge & Join
    Combine dataframes using Comman Columns

### First Dataframe

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID':[1,2,3],
    'Name':['Ram','Sham','Amit']
})

print(df1)
```

### Second Dataframe

```python
df2 = pd.DataFrame({
    'ID':[1,2,4],
    'Salary':[50000,60000,70000]
})

print(df2)
```

### merge() OR Inner Join

```python
pd.merge(df1, df2, on='ID')
```

```python
pd.merge(df1, df2, on='ID',how ='inner')
```

### Left Join
    Left table gets priority
    Match not found in second table then put it as null

```python
pd.merge(df1, df2, on='ID',how="left")
```

### Right Join
    Right table gets priority
    Match not found in first table puts null

```python
pd.merge(df1, df2, on='ID',how ='right')
```

### Outer OR Full Join
    Priority goes for both tables
    Gets Everything from both tables
    Match not found then puts null values

```python
pd.merge(df1, df2, on='ID',how='outer')
```
