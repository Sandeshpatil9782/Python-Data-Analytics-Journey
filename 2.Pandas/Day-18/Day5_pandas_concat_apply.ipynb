## Pandas Concatenation
    To join two or more DataFrames.

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Name':['Ram','Sham'],
    'Marks':[85,90]
})


print(df1) # 1st DataFrame
```

```python
df2 = pd.DataFrame({
    'Name':['Amit','Riya'],
    'Marks':[70,95]
})

print(df2) # 2nd Dataframe
```

### concat()
    Combine Dataframes

```python
result = pd.concat([df1,df2])

print(result)
```

### ignore_index = True
    New contunious Index Formed

```python
result = pd.concat(
    [df1,df2],
    ignore_index=True
)

print(result)
```

### Column Wise Concat

```python
df3 = pd.DataFrame({
    'City':['Delhi','Mumbai']
})

print(
    pd.concat(
        [df1,df3],
        axis=1
    )
)
```

### Keys
    Adds labels to dataframes

```python
print(
    pd.concat(
        [df1,df2],
        keys=['Class1','Class2']
    )
)
```

## Pandas Apply Function
    Apply custom or built in functions.

```python
data = {
    'Name':['Ram','Sham','Amit'],
    'Marks':[85,90,70]
}

df = pd.DataFrame(data)

print(df)
```

### apply()
    Applies function on data

```python
df['Marks'] = df['Marks'].apply(
    lambda x: x + 5
)

print(df)
```

### apply on multiple column

```python
df['Total'] = df.apply(
    lambda row: row['Marks'] + 10,
    axis=1
)

print(df)
```

### lambda Function
    small anonymous function
    Ex:-lambda x:x * 5 # Multiply by 5 to each data

### Custom Function

```python
def grade(x):

    if x >= 90:
        return 'A'

    else:
        return 'B'

df['Grade'] = df['Marks'].apply(grade)

print(df)
```

### map()
    Maps Values

```python
df['Name'] = df['Name'].map(
    {
        'Ram':'RAM',
        'Sham':'SHAM',
        'Amit':'AMIT'
    }
)

print(df)
```

### Student Perforemence

```python
df1 = pd.DataFrame({
    "Name": ["Ram", "Sam"],
    "Python": [78, 90]
})

df2 = pd.DataFrame({
    "Name": ["Amit", "Neha"],
    "Python": [67, 95]
})

df = pd.concat([df1, df2],ignore_index=True)

def performance(x):
    if x >= 80:
        return "Excellent"
    else:
        return "Average"

df["Performance"] = df["Python"].apply(performance)

df
```

### Employee Bonus

```python
df = pd.DataFrame({
    "Employee": ["A", "B", "C"],
    "Salary": [25000, 70000, 50000],
    "Experience": [2, 8, 5]
})

def bonus(exp, salary):
    if exp > 5:
        return salary * 0.20
    else:
        return salary * 0.10

df["Bonus"] = df.apply(
    lambda x: bonus(x["Experience"], x["Salary"]),
    axis=1
)

print(df)
```

### Product Category

```python
df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Mobile"],
    "Price": [50000, 500, 20000],
    "Quantity": [2, 5, 1]
})

df["Total"] = df["Price"] * df["Quantity"]

def category(x):
    if x > 50000:
        return "Premium"
    else:
        return "Normal"

df["Category"] = df["Total"].apply(category)

print(df)
```
