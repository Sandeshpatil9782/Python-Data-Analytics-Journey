### ALL PLOTS IN ONE FRAME

```python
import matplotlib.pyplot as plt
```

## 1. LINE PLOT

```python

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 120, 90, 150, 180, 200]

plt.figure(figsize=(8,5), facecolor='lightgrey')

plt.plot(
    months,
    sales,

    color='green',
    marker='o',
    mfc='blue',
    mec='black',
    linestyle='-',
    linewidth=2,
    markersize=8,

    label='Sales'
)

plt.title(
    "Monthly Sales Trend",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel(
    "Months",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Sales",
    fontsize=12,
    fontweight='bold'
)

plt.legend(
    title="Report",
    fontsize=10
)

plt.grid(
    linestyle='--',
    alpha=0.5
)

plt.legend(facecolor='lightgrey')
plt.tight_layout()
plt.show()
```

## 2. BAR PLOT

```python

products = ['Laptop', 'Mobile', 'Tablet', 'Accessories']
sales = [50, 80, 40, 30]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.bar(
    products,
    sales,

    color=['skyblue', 'orange', 'lightgreen', 'pink'],
    edgecolor='black',
    linewidth=1.5
)

plt.title(
    "Product Sales Analysis",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Products",fontweight='black',fontsize=12)
plt.ylabel("Units Sold",fontweight='black',fontsize=12)

plt.grid(axis='y', alpha=0.3)

plt.show()
```

## 3. HISTOGRAM

```python

salary = [
    20000,25000,25000,
    30000,30000,30000,
    35000,35000,35000,35000,
    40000,40000,40000,40000,40000,
    45000,45000,45000,
    50000,50000,
    55000,
    60000
]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.hist(
    salary,

    bins=5,

    color='gold',

    edgecolor='black'
)

plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Salary",fontweight='black',fontsize=12)
plt.ylabel("Frequency",fontweight='black',fontsize=12)

plt.xticks( [25000,35000,45000,55000],
    ['20K-30K','30K-40K','40K-50K','50K-60K'])

plt.show()
```

## 4. SCATTER PLOT

```python


hours = [1,2,3,4,5,6,7,8]

marks = [40,70,60,85,45,80,67,90]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.scatter(
    hours,
    marks,

    s=120,

    color='green',

    marker='D',
    label='Students'
)

plt.title(
    "Study Hours vs Marks",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.legend(facecolor='lightgrey')

plt.grid(alpha=0.4)

plt.show()
```

## 5. PIE CHART

```python

products = ['Laptop','Mobile','Tablet','Accessories']

sales = [40,30,20,10]

plt.figure(figsize=(8,8),facecolor='lightgrey')

plt.pie(
    sales,

    labels=products,

    autopct='%1.1f%%',

    explode=[0.1,0,0,0],

    shadow=True,

    textprops={
        'fontsize': 12,
        'color': 'darkblue',
        'fontstyle': 'italic',
        'fontweight': 'bold'
    }

)

plt.title(
    "Sales Contribution",
    fontsize=15,
    fontweight='bold'
)

plt.legend(loc=2,)
plt.show()
```

## 6. BOX PLOT

```python


data = [
    [10,12,15,18,20,25,30],
    [15,20,22,24,28,32,35]
]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.boxplot(
    data,
    
   boxprops=dict(
        color='black',
        linewidth=1
   ),
    
    whiskerprops=dict(
        color='green',
        linewidth=2
    ),

    capprops=dict(
        color='red',
        linewidth=2
    ),

    medianprops=dict(
        color='orange',
        linewidth=3
    ),

    meanprops=dict(
        marker='o',
        markerfacecolor='purple',
        markeredgecolor='black',
        markersize=10
    ),


    patch_artist=True,

    showmeans=True
)


plt.title(
    "Salary Distribution Analysis",
    fontsize=15,
    fontweight='bold'
)
plt.grid(alpha=0.7)
plt.show()
```

## 7. STEM PLOT

```python

x = [1,2,3,4,5]

y = [5,9,7,10,12]

plt.figure(figsize=(8,5),facecolor='lightgrey')

markerline, stemlines, baseline = plt.stem(
    x,
    y
)

plt.setp(markerline,
         marker='D',
         markersize=20,
        mec='black',
        mfc='green'
        ),

plt.setp(
    stemlines,
    color='blue',
    linewidth=2,
    linestyle='--'
),

plt.setp(
    stemlines,
    color='blue',
    linewidth=2,
    linestyle='--'
),

plt.title(
    "Stem Plot",
    fontsize=15,
    fontweight='bold'
)
plt.grid(alpha=0.7)
plt.show()
```

## 8. STEP PLOT

```python

x = [1,2,3,4,5]

y = [10,20,15,25,30]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.step(
    x,
    y,
    color='blue',
    where='mid',
    marker='o',
    mfc='green',
    ms=12,
    linestyle='--',
    linewidth=3,
    
)

plt.title(
    "Inventory Change Analysis",
    fontsize=15,
    fontweight='bold'
)
plt.grid(alpha=0.7)
plt.show()
```

## 9. AREA PLOT

```python

months = [1,2,3,4,5]

sales = [100,90,140,130,170]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.plot(months, sales, color='navy')

plt.fill_between(
    months,
    sales,

    color='magenta',
    
    alpha=0.7
)

plt.title(
    "Monthly Sales Area Chart",
    fontsize=15,
    fontweight='bold'
)
plt.grid(alpha=0.5)
plt.show()
```

## 10. STACK PLOT

```python

months = [1,2,3,4,5]

laptop = [10,17,20,35,30]

mobile = [20,25,30,45,40]

tablet = [5,10,18,23,20]

plt.figure(figsize=(8,5),facecolor='lightgrey')

plt.stackplot(
    months,

    laptop,
    mobile,
    tablet,

    labels=['Laptop','Mobile','Tablet'],
    edgecolor='black',
    alpha=0.8
)

plt.legend()

plt.title(
    "Product Contribution Analysis",
    fontsize=15,
    fontweight='bold'
)

plt.show()
```

## 11. DONUT CHART

```python

products = ['Laptop', 'Mobile', 'Tablet', 'Accessories']

sales = [40, 30, 20, 10]

colors = ['skyblue', 'orange', 'lightgreen', 'pink']

plt.figure(
    figsize=(8,7),
    facecolor='lightgrey'
)

plt.pie(
    sales,

    labels=products,

    colors=colors,

    autopct='%1.1f%%',

    startangle=90,

    explode=[0.05, 0, 0, 0],

    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 1
    }
)

# Create Donut Hole
centre_circle = plt.Circle(
    (0, 0),
    0.45,

    fc='white',
    edgecolor='black'
)

plt.gca().add_artist(centre_circle)

plt.title(
    "Product Sales Contribution",
    fontsize=15,
    fontweight='bold',
    color='darkblue'
)

plt.legend(
    title="Products",
    loc='upper left',
    fontsize=10,

     facecolor='lightgrey'
)

plt.tight_layout()

plt.show()
```

## BUSINESS DASHBOARD(SubPlots)

```python

fig, ax = plt.subplots(
    2,2,

    figsize=(10,8),
    facecolor='lightgrey'
)

fig.suptitle(
    "Business Performance Dashboard",
    fontsize=18,
    fontweight='bold'
)

ax[0,0].plot([1,2,3],[10,20,30],marker='>',color='blue',mfc='green',ms=12)
ax[0,0].set_title("Sales Trend",fontweight='bold')

ax[0,0].set_xticks([1,2,3])
ax[0,0].set_yticks([10,20,30])

col=['magenta','purple','brown']
ax[0,1].bar(['A','B','C'],[20,30,15],color=col,edgecolor='black',linewidth=2)
ax[0,1].set_title("Category Sales",fontweight='bold')

ax[1,0].scatter([1,2,3,4,5,6,7,8,9],[2,3,6,23,13,36,5,12,9],
               s=150,
                color='green',
                
               
               )
ax[1,0].set_title("Correlation",fontweight='bold')

ax[1,0].set_xticks([1,2,3,4,5,6,7,8,9])

data = [10,20,20,30,40]

labels = ['A','B','C','D','E']

ax[1,1].pie(
    data,
   
    autopct='%1.1f%%',
    shadow=True,
    startangle=90,
    radius=1.3,
    radiuscolor='black',
    linewidth=2,
    
    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 2
    }
)

ax[1,1].set_title(
    "Distribution",
    fontsize=12,
    fontweight='bold'
)

plt.tight_layout(rect=[0,0,1,0.95])

plt.show()
```
