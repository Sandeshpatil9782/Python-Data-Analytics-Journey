## Box Plot
    Box Plot (Box-and-Whisker Plot) is used to visualize:

    Data Distribution
        Median
        Quartiles (Q1, Q2, Q3)
        Spread of Data
        Outliers

    Real-World Uses
        Employee Salary Analysis
        Student Marks Analysis
        Sales Distribution
        Customer Spending Analysis
        
    Box Plot Structure
    Outlier   Min    Q1   Median   Q3    Max
       ●       |------|----|----|------|
              Whisker    Box     Whisker
    Components:-
    Term                Meaning
    Min             Minimum Value
    Q1              25th Percentile
    Median(Q2)  	50th Percentile
    Q3              75th Percentile
    Max             Maximum Value
    Outlier         Unusual Data Point
    Whiskers        Lines extending from box

```python
import matplotlib.pyplot as plt
```

```python
salary = [
    25000,28000,30000,
    32000,35000,38000,
    40000,42000,45000,
    50000,55000,60000,
    80000
]

plt.boxplot(salary)

plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.boxplot(salary,
           vert=False,
           sym='g+'  # Change outlier Style
            
           )   # Vert=True And Vert=False

plt.show()
```

### ATTRIBUTES 
    patch_artist=True -> Allows Box Coloring,Without This Color Won't Be Fill
    Vert=True         -> For Vertical Shown
    Vert=False        -> For Horizontal Shown
    showmeans=True    -> Shows Averages(Mean)
    showfliers=False  -> Hide Outliers
    notch=True        -> Shows Notch Around Mediun
    whis=1.5          -> Control Whisker Length  Default Is 1.5
    

```python
plt.figure(figsize=(5,4))
plt.boxplot(salary,
           vert=True,
           patch_artist=True,
           widths=0.4,
           tick_labels=['Employees'],
           showmeans=True,
           showfliers=False,
           notch=True,
           whis=1,
           
           )   
plt.show()
```

### Stylling Attributes
    

```python
plt.figure(figsize=(6,6))
plt.boxplot(salary,
           vert=True,
           patch_artist=True,
           widths=0.4,
           tick_labels=['Employees'],
           showmeans=True,
           #notch=True,
           whis=1.6,

            boxprops=dict(
                facecolor='Magenta',
                color='navy',
                linewidth=2   
                    ),
            # Whiskers Line
            whiskerprops=dict(
                color='green',
                linewidth=2
                    ),
            # Min And Max Line
            capprops=dict(
                color='red',
                linewidth=2
                     ),
            # Median Line
            medianprops=dict(
                color='yellow',
                linewidth=3
                    ),
            # Outliers Style   Also Using sym=" + "
            flierprops=dict(
                marker='o',
                markerfacecolor='lightgreen',
                markeredgecolor='black',
                linewidth=2,
                markersize=15
                    ),
            # showmean Style 
            meanprops=dict(
                marker='D',
                markerfacecolor='gold',
                markeredgecolor='black',
                markersize=10
    )

           ),




plt.show()
```

### Multiple Box Plot

```python

data_analyst = [
    30000,35000,40000,
    45000,50000,55000
]

data_scientist = [
    50000,60000,70000,
    80000,90000,100000
]

data_engineer = [
    45000,55000,65000,
    75000,85000,95000
]

business_analyst = [
    28000,32000,38000,
    42000,48000,52000
]

ml_engineer = [
    60000,70000,80000,
    90000,100000,120000
]

plt.figure(
    figsize=(10,6),
    facecolor='lightgrey'
)

box = plt.boxplot(
    [
        data_analyst,
        data_scientist,
        data_engineer,
        business_analyst,
        ml_engineer
    ],

    tick_labels=[
        'DA',
        'DS',
        'DE',
        'BA',
        'MLE'
    ],

    patch_artist=True,

    showmeans=True,
    meanprops=dict(
        marker='D',
        markerfacecolor='magenta',
        markeredgecolor='black',
        markersize=10
    ),
    widths=0.6
)

colors = [
    'skyblue',
    'lightgreen',
    'orange',
    'plum',
    'gold'
]

for patch, color in zip( box['boxes'],colors):
    patch.set_facecolor(color)

plt.title(
    "Salary Distribution By Role",
    fontsize=15,
    fontweight='bold'
)

plt.ylabel(
    "Salary",
    fontsize=12,
    fontweight='bold'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.7
)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6,6),facecolor='lightgrey')
plt.boxplot(salary,
           vert=True,
           patch_artist=True,
           widths=0.4,
           tick_labels=['Employees'],
           showmeans=True,
           notch=True,
           whis=1.6,

            boxprops=dict(
                facecolor='Magenta',
                color='navy',
                linewidth=2   
                    ),
            # Whiskers Line
            whiskerprops=dict(
                color='green',
                linewidth=2
                    ),
            # Min And Max Line
            capprops=dict(
                color='red',
                linewidth=2
                     ),
            # Median Line
            medianprops=dict(
                color='yellow',
                linewidth=3
                    ),
            # Outliers Style   Also Using sym=" + "
            flierprops=dict(
                marker='o',
                markerfacecolor='lightgreen',
                markeredgecolor='black',
                linewidth=2,
                markersize=15
                    ),
            # showmean Style 
            meanprops=dict(
                marker='D',
                markerfacecolor='gold',
                markeredgecolor='black',
                markersize=10
    )

           ),


plt.title(
    "Employee Salary Distribution",
    color='navy',
    fontsize=15,
    fontweight='bold'
)

plt.ylabel("Salary",
           color='blue',
          fontsize=11,
    fontweight='bold'
          )
plt.yticks(color='black',fontweight='bold')

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()

plt.savefig("BoxPlot.png",dpi=150)

plt.show()
```
