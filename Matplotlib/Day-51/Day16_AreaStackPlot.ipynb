## Area Plot 

    It is a Line Plot but below area became filled.

    Real-World Use
    Monthly Sales Trend
    Website Visitors Growth
    Revenue Trend
    Stock Price Trend

```python
import matplotlib.pyplot as plt
```

```python
months = [1,2,3,4,5,6]

sales = [100,120,140,130,160,180]

plt.fill_between(
    months,
    sales
)

plt.show()
```

## Stackplot
    Multiple area plots stacked on top of each other to show the contribution of different categories over time.

    Real-World Use
    Product-wise Sales Contribution
    Department Expenses
    Traffic Source Analysis
    Revenue Breakdown


```python
months = [1,2,3,4,5,6]

sales = [100,110,130,150,115,160]

plt.stackplot(
    months,
    sales,
    color='lightgreen',
    edgecolor='black',
    linewidth=2,
    alpha=0.7
)

plt.show()
```

### Multiple Stack Plot
    basline='zero'  
    zero	          Starts stacking from 0
    sym               Symmetric around center
    wiggle	          Wave-like layout
    weighted_wiggle	  Streamgraph layout

```python
months = [1,2,3,4,5]

laptop = [10,15,12,20,25]

mobile = [20,25,30,35,40]

l=['Laptop','mobile']
cl=['lightgreen','lightblue']

plt.stackplot(
    months,
    laptop,
    mobile,
    colors=cl,
    labels=l,
    edgecolor='black',
    linewidth=2,
    alpha=1,
    baseline='weighted_wiggle',   
    
    
)
plt.legend()

plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']

laptop = [20,25,15,30,32,38]

mobile = [15,20,25,28,42,37]

tablet = [10,15,20,25,40,35]

plt.figure(
    figsize=(8,5),
    facecolor='lightgrey'
)

plt.stackplot(
    months,

    laptop,
    mobile,
    tablet,

    labels=[
        'Laptop',
        'Mobile',
        'Tablet'
    ],

    colors=[
        'skyblue',
        'orange',
        'lightgreen'
    ],

    edgecolor='black',
    linewidth=1,
    alpha=0.9
)

plt.title(
    "Product Contribution Analysis",
    fontsize=15,
    color='blue',
    fontweight='bold'
)

plt.xlabel("Months",
          fontsize=12,
          color='navy',
          fontweight='bold'
          )

plt.ylabel("Sales",
          fontsize=12,
          color='navy',
          fontweight='bold')

legend=plt.legend(
    title="Products",
    facecolor='lightgrey',
    fontsize=13,
   
)
legend.get_title().set_color('navy')
legend.get_title().set_fontweight('bold')

plt.grid(
    linestyle='--',
    alpha=0.6
)

plt.savefig('StackPlot.png',dpi=150)
plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']

sales = [100,120,140,130,160,180]

plt.figure(
    figsize=(6,5),
    facecolor='lightgrey'
)

plt.fill_between(
    months,
    sales,

    color='orange',

    alpha=0.8,

    edgecolor='navy',

    linewidth=2,

    label='Sales'
)

plt.title(
    "Monthly Sales Trend",
    color='blue',
    fontsize=15,
    fontweight='bold'
)
plt.xlabel(
    "Months",
    fontsize=12,
    color='navy',
    fontweight='bold'
)

plt.ylabel(
    "Sales ",
    fontsize=11,
    color='navy',
    fontweight='bold'
)

plt.legend(fontsize=15,facecolor='lightgrey')

plt.grid(
    linestyle='--',
    alpha=0.7
)

plt.savefig('AreaPlot.png',dpi=150)
plt.show()
```

## fill_between Plot
    An area plot that fills the space between a line and the x-axis, or between two lines.

    Real-World Uses
    Profit vs Target Profit
    Temperature Range
    Revenue vs Cost
    Stock Price Movement

```python

months = [1,2,3,4,5]

sales = [100,120,140,160,180]

target = [90,110,130,150,170]

plt.fill_between(
    months,
    sales,
    target,

    label='Sales',
    
    color='green',

    
)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

months = [1,2,3,4,5]

sales = [100,125,145,170,200]

target = [90,115,135,140,180]

plt.figure(
    figsize=(6,5),
    facecolor='lightgrey'
)
plt.fill_between(
    months,
    sales,
    target,

    color='green',
    alpha=0.7,

    edgecolor='yellow',
    linewidth=2,

    label='Sales'
)

plt.plot(
    months,
    sales,
    color='yellow',
    linewidth=2
)

plt.plot(
    months,
    target,
    color='yellow',
    linewidth=2
)

plt.title(
    "Sales vs Target Performance",
    fontsize=15,
    fontweight='bold',
    color='darkblue'
)

plt.xlabel(
    "Months",
    fontsize=12,
    fontweight='bold',
    color='purple'
)

plt.ylabel(
    "Sales Amount",
    fontsize=12,
    fontweight='bold',
    color='purple'
)

plt.legend(
    title="Performance",
    fontsize=11,
    title_fontsize=13
)

plt.grid(
    linestyle='--',
    alpha=0.5
)

plt.legend()

plt.savefig('Fill_Between_Plot.png',dpi=150)

plt.show()
```

### Condition Based filling

```python
import numpy as np


months = np.array([1,2,3,4,5,6])

sales = np.array([100,120,140,130,160,180])

plt.fill_between(
    months,
    sales,

     label='Sales',
   
    where=((sales > 120) & (sales < 180)),

    color='magenta',

    alpha=1
)
plt.legend()
plt.show()
```
