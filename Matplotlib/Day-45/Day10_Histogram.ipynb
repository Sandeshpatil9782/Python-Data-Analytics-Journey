# Histogram Plot
    -A histogram shows the distribution of numerical data by grouping values into intervals called bins.
    =Continuous Pictorial Data Representation

    Customer Age Distribution
    Employee Salary Distribution
    Product Price Distribution
    Check Data Distribution
    Detect Skewness
    Find Outliers

    Ex:- A company wants to analyze employee salaries.

```python
import matplotlib.pyplot as plt 
```

```python
salary = [
    25000,26000,27000,28000,29000,
    30000,30000,31000,32000,32000,
    33000,34000,35000,35000,36000,
    37000,38000,38000,39000,40000,
    40000,41000,42000,42000,43000,
    44000,45000,45000,46000,47000,
    48000,49000,50000,50000,51000,
    52000,53000,54000,55000,56000,
    57000,58000,59000,60000,61000,
    62000,63000,64000,65000,66000
]

plt.hist(salary)  # Bins are groups


plt.title(
    "Employee Salary Distribution",
    
)

plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python

plt.hist(salary,
         color='magenta',
         edgecolor='g',
         bins=7
         
        )  


plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

### Hist Types:-  bar,barstacked,step,stepfilled,

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='magenta',
         edgecolor='g',
         linewidth=2,
         bins=b,
         histtype='step' 
         
        )  


plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
          color='g')

plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')

plt.show()
```

### Density:-Shows probability distribution instead of frequency count.

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='magenta',
         edgecolor='g',
         alpha=0.8,
         density=True, 
         linewidth=1,
         bins=b,
         histtype='bar',
         
         
        )  


plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
            color='g')

plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')

plt.show()
```

### Bottom:- Gives Starting Range For Bars

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='magenta',
         edgecolor='g',
         alpha=0.8,
         linewidth=1,
         bins=b,
         histtype='bar',
         bottom=20000 #Starting Range
         
        )  


plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
            color='g')

plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')

plt.show()
```

### cumulative=True
    makes the histogram show the cumulative frequency instead of the normal frequency.
    
    cumulative=-1 Shows cumulative counts from right to left.
    
    Normal histogram:How many values are in each bin?
    Cumulative histogram:How many values are up to this bin?

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='skyblue',
         edgecolor='g',
         alpha=0.9,
         linewidth=1,
         bins=b,
         histtype='bar',
         cumulative=True,
         align='mid'  # Gives us Position Set
         
        )  

# Axis Vertical Line To set a reference
plt.axvline(
    42000,
    color='darkred',
    linestyle='--',
    linewidth=2,
    alpha=0.8,
    label='Average Salary'
)
# Textual Representaion
plt.text(
    42000,
    40,
    "<-Partition "
)
plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'

)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
            color='g'
          )


plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')
plt.legend()
plt.show()
```

### Log Values :- To set Y-axis logarithmic scale

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='skyblue',
         edgecolor='g',
         alpha=0.9,
         linewidth=1,
         bins=b,
         histtype='bar',
         log=True  
        )  



plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'

)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
            color='g'
          )


plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')

plt.show()
```

```python

b=[20000,30000,40000,50000,60000]
plt.hist(salary,
         color='darkblue',
         edgecolor='black',
         alpha=0.9,
         linewidth=2,
         bins=b,
         histtype='bar',
         align='mid',
         orientation='horizontal'  # Deafault Vertical
        )  


plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold'

)

plt.xlabel("Salary",
           fontsize=15,
            fontweight='bold',
            color='g'
          )


plt.ylabel("Frequency",
          fontsize=10,
          fontweight='bold',
          color='g')

plt.show()
```

```python

bins = [20000,30000,40000,50000,60000]

plt.figure(
    figsize=(8,5),
    facecolor='lightgrey'
)

plt.hist(
    salary,
    bins=bins,
    color='#4E79A7',
    edgecolor='black',
    alpha=0.9,
    linewidth=1,
    histtype='bar',
    align='mid',
    
    rwidth=0.9, #To set bar width
    
    label='Salary Distribution'
)



plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold',
    color='blue'
)

plt.xlabel(
    "Salary Range",
    fontsize=12,
    fontweight='bold',
    color='green'
)

plt.ylabel(
    "Frequency",
    fontsize=12,
    fontweight='bold',
    color='green'
)

# Only these ticks will appear
plt.xticks(
    [20000,30000,40000,50000,60000],
    ['20K','30K','40K','50K','60K'],
    fontsize=10,
    fontweight='bold'
)

plt.yticks(
    fontsize=10,
    fontweight='bold'
)

plt.legend(
    title='HR Report',
    loc='best',
    facecolor='yellow',
    edgecolor='black'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.7
)

plt.tight_layout()




plt.show()
```

```python

bins = [20000,30000,40000,50000,60000]

plt.figure(
    figsize=(8,5),
    facecolor='lightgrey'
)

plt.hist(
    salary,
    bins=bins,
    color='darkgreen',
    edgecolor='black',
    alpha=0.9,
    linewidth=1,
    histtype='bar',
    align='mid',
    label='Salary Distribution'
)

plt.title(
    "Employee Salary Distribution",
    fontsize=15,
    fontweight='bold',
    color='blue'
)

plt.xlabel(
    "Salary Range",
    fontsize=12,
    fontweight='bold',
    color='green'
)

plt.ylabel(
    "Frequency",
    fontsize=12,
    fontweight='bold',
    color='green'
)

# Only these ticks will appear
plt.xticks(
    [20000,30000,40000,50000,60000],
    ['20K','30K','40K','50K','60K'],
    fontsize=10,
    fontweight='bold'
)

plt.yticks(
    fontsize=10,
    fontweight='bold'
)

plt.legend(
    title='HR Report',
    loc='best',
    facecolor='yellow',
    edgecolor='black'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.7
)

plt.tight_layout()


plt.savefig(Histplot.png",
    dpi=300,
    bbox_inches='tight')

plt.show()
```
