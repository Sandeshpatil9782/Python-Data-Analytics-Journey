## Subplots:
    Subplots allow multiple plots to be displayed in a single figure.

    Real-World Uses
    Sales Dashboard
    EDA Reports
    KPI Comparison
    Data Analyst dashboards

    plt.subplot(rows, columns, position)

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))

# First Plot
plt.subplot(2,2,1)

plt.plot([1,2,3],[10,20,30])

plt.title("Line Plot")


# Second Plot
plt.subplot(2,2,2)

plt.bar(['A','B','C'],[10,20,15])

plt.title("Bar Chart")

plt.show()
```

```python

fig, ax = plt.subplots(
    2,
    2,

    figsize=(10,8)
)

ax[0,0].plot(
    [1,2,3],
    [10,30,20],color='navy',linewidth=5,
)

ax[0,0].text(2,20,"Center",fontsize=12,style='italic', bbox={'facecolor':'lightgrey'})

ax[0,0].set_xticks([1,2,3])

ax[0,0].set_yticks([10,20,30])

ax[0,0].set_title(
    "Line Plot"
)

colors=['r','g','magenta']

ax[0,1].bar(
    ['A','B','C'],
    [10,20,15],color=colors,edgecolor='black'
)

ax[0,1].set_title(
    "Bar Plot"
)


ax[1,0].scatter(
    [1,2,3,4,5,6,7],
    [5,9,15,20,17,23,12],s=150,color='purple'
)

ax[1,0].annotate("Middle Most",xy=(4.2,20),xytext=(6,19.6),arrowprops=dict(facecolor='black',shrink=100) )

ax[1,0].set_title(
    "Scatter Plot"
)

salary = [
    20000,
    25000, 25000,
    30000, 30000, 30000,
    35000, 35000, 35000, 35000,
    40000, 40000, 40000, 40000, 40000,
    45000, 45000, 45000, 45000,
    50000, 50000, 50000,
    55000, 55000,
    60000
]
b = [20000,30000,40000,50000,60000]
ax[1,1].hist(
    salary,bins=b,color='yellow',edgecolor='black'
)
ax[1,1].set_xticks(
    [20000, 30000, 40000, 50000, 60000]
)

ax[1,1].set_title(
    "Histogram"
)

plt.tight_layout()

plt.show()
```

```python

fig, ax = plt.subplots(
    2, 2,
    figsize=(10, 8),
    facecolor='lightgrey'
)
fig.suptitle(
    "Sales Dashboard Analysis",
    fontsize=18,
    fontweight='bold',
    color='darkblue'
)

# 1. Line Plot
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 120, 140, 70, 180]

ax[0,0].plot(
    months,
    sales,
    color='blue',
    marker='o'
)

ax[0,0].set_title("Line Plot")


# 2. Pie Chart
products = ['Laptop', 'Mobile', 'Tablet']

product_sales = [40, 35, 25]

ax[0,1].pie(
    product_sales,
    labels=products,
    autopct='%1.1f%%'
)

ax[0,1].set_title("Pie Chart")


# 3. Area Plot
ax[1,0].fill_between(
    months,
    sales,
    color='skyblue',
    alpha=0.7
)

ax[1,0].plot(
    months,
    sales,
    color='navy'
)

ax[1,0].set_title("Area Plot")


# 4. Histogram
salary = [
    20000,25000,25000,30000,30000,
    35000,35000,40000,40000,40000,
    45000,45000,50000,55000,60000
]

ax[1,1].hist(
    salary,
    bins=5,
    color='yellow',
    edgecolor='black'
)

ax[1,1].set_title("Histogram")


plt.tight_layout()

plt.savefig('SubPlots.png',dpi=150,bbox_inches='tight')

plt.show()
```
