## Matplotlib + NumPy + Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Graph Background Color
    facecolor=' '
    plt.axes().set_facecolor(" ")

### Numpy Array Visualization

```python
x = np.arange(1,11)
y = x**2
plt.axes().set_facecolor("lightgreen")
plt.plot(x,y,color="purple")
plt.show()
```

### Marker 
    marker=
    markersize=
    markerfacecolor=
    markeredgecolor=

### Pandas DataFrame Visualization

```python
df = pd.DataFrame({
    "Month":["Jan","Feb","Mar","Apr"],
    "Sales":[100,150,120,180]
})

plt.axes().set_facecolor("lightblue")
plt.plot(df["Month"],df["Sales"],marker="o", markerfacecolor="blue",markeredgecolor="black",color="red")
plt.show()
```

### Line Styling
    linewidth=
    linestyle=
    alpha=

#### Multiple Numpy lines

```python
x = np.arange(1,11)

plt.plot(x,x,linewidth=2,label="x")
plt.plot(x,x**2,linestyle="--",alpha=0.9,label="x²")

plt.legend()
plt.show()
```

#### Legend Position
    upper left
    upper right
    lower left
    lower right
    best

```python
plt.plot(
    [1,2,3],
    [10,20,25],
    label="Sales"
)

plt.legend(loc="upper left")

plt.show()
```

### Tick Customization

```python
plt.plot([1,2,3,4],[10,20,15,30])

plt.xticks(
    [1,2,3,4],
    ["Jan","Feb","Mar","Apr"],
    rotation=45
)

plt.show()
```

### Annotation

```python
plt.plot([1,2,3],[10,20,30])

plt.annotate(
    "Highest",
    xy=(3,30),
    xytext=(2.5,25),
    arrowprops=dict(arrowstyle="->")
)

plt.show()
```

### Text Add On Color

```python
plt.plot([1,2,3],[10,20,30])

plt.text(
    2,
    20,
    "Center Point"
)

plt.show()
```

### Style Theme

```python

plt.style.use("ggplot")

plt.plot([1,2,3],[10,20,30])

plt.show()
plt.style.use("default")
```

### Subplot

```python
plt.subplot(2,1,1)
plt.plot([1,2,3],[10,20,30])

plt.subplot(2,1,2)
plt.bar([1,2,3],[5,6,7])

plt.tight_layout()

plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,15,30,25]

plt.figure(figsize=(10,8))

plt.suptitle(
    "Matplotlib Dashboard",
    fontsize=16,
    fontweight='bold',
    y=0.98
)

# Line
plt.subplot(2,2,1)
plt.plot(x,y,color="purple")
plt.title("Line Graph")

# Bar
plt.subplot(2,2,2)
plt.bar(x,y,color="purple")
plt.title("Bar Graph")

# Scatter
plt.subplot(2,2,3)
plt.scatter(x,y,color="purple")
plt.title("Scatter Plot")

# Histogram
plt.subplot(2,2,4)
plt.hist(
    [10,20,15,30,25,18,22,28,16,14,35,27,19],
    bins=5,
    color="purple"
)
plt.title("Histogram")


plt.tight_layout()

plt.show()
```
