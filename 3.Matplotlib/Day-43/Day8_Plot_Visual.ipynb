```python
from matplotlib import pyplot as plt
```

### LINE PLOT
    A line plot is used to visualize trends and changes over time. Data points are connected using a line.

    Real-World Uses
    Monthly Sales Analysis
    Website Traffic Monitoring
    Stock Price Trends
    Temperature Changes
    Revenue Growth Analysis

    Ex:-A company wants to track monthly sales performance throughout the year.

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [100,120,90,150,180,200]

plt.figure(figsize=(6,4))
plt.plot(months,sales)

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [100,120,90,150,180,200]

plt.figure(figsize=(7,4))
plt.plot(months,sales,
        color='green',
        marker='o',
        linestyle='solid',
        linewidth=2,
        markersize=10,
        markerfacecolor='yellow',
        markeredgewidth=2,
        markeredgecolor='blue'
      
    
        )

plt.xlabel("Months",fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.title("Monthly Sales Trend",fontsize=15,
          fontweight="bold",color='b')


plt.show()

```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [100,120,90,150,180,200]

plt.figure(figsize=(7,4))
plt.plot(months,sales,'b*--',        #'b*--'-> b=Blue,*=Marker,--=Linestyle

        linewidth=3,
        markersize=15,
        markerfacecolor='yellow',
        markeredgewidth=1,
        markeredgecolor='g'
      
        )

plt.xlabel("Months",fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.title("Monthly Sales Trend",fontsize=15,
          fontweight="black",color='purple')


plt.show()

```

### Legend Attributes
    loc for position 'upper left/right','lower left/right','center','best'
    fontsize for text size
    facecolor for background color
    edgecolor for border color
    labelcolor for legend text color
    shadow for visual emphasis
    frameon for Shows/Hide Border
    ncol for Useful when many labels exist.

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [100,120,90,150,180,200]

plt.figure(figsize=(7,4))
plt.plot(months,sales,'bs-',     

        linewidth=2,
        markersize=7,
        markerfacecolor='yellow',
        markeredgewidth=1,
        markeredgecolor='black',
        label='Sales'
        )

plt.xlabel("Months",fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.title("Monthly Sales Trend",fontsize=15,
          fontweight="bold",color='b')

plt.legend(title="Monthly Report",
    loc="upper left",
    fontsize=12,
    facecolor="lightgrey",
    edgecolor="pink",
    labelcolor="blue",
    shadow=True,
    frameon=True
          
    )

plt.show()

```

### xticks() And yticks()
    rotation	Rotate labels
    fontsize	Text size
    color	    Text color
    fontweight	Bold text

```python
months = ["January","February","March","April","May","June"]
sales = [100,120,90,150,180,200]

plt.figure(figsize=(7,4))
plt.plot(months,sales,     

        color='green',
        marker='^',
        linestyle=':',
        linewidth=2,
        markersize=8,
        markerfacecolor='yellow',
        markeredgewidth=1,
        markeredgecolor='black',
        label='Sales'
        
        )


plt.xlabel("Months",fontsize=14,color='purple')
plt.ylabel("Sales",fontsize=14,color='purple')
plt.title("Monthly Sales Trend",fontsize=15,
          fontweight="bold",color='b')

plt.legend(title="Monthly Report",  
    loc="best",
    fontsize=14,
    facecolor="darkgrey",
    edgecolor="black",
    labelcolor="red",
    shadow=False,
    frameon=True  
          )

plt.xticks(rotation=75,
           fontsize=10,
           fontweight='bold',
           color='black'
          
          )
plt.yticks(
           fontsize=10,
           fontweight='bold',
           color='black'
          
          )

plt.tight_layout()
plt.show()

```

### Multiple Plots

```python

months = ["January","February","March","April","May","June"]

sales = [100,120,90,150,180,200]
profit = [20,25,15,35,50,100]

plt.figure(figsize=(8,5), facecolor='lightgrey')

# First Line - Sales
plt.plot(
    months,
    sales,
    color='green',
    marker='^',
    linestyle='solid',
    linewidth=2,
    markersize=8,
    markerfacecolor='yellow',
    markeredgecolor='black',
    label='Sales'
)

# Second Line - Profit
plt.plot(
    months,
    profit,
    color='blue',
    marker='o',
    linestyle='--',
    linewidth=2,
    markersize=8,
    markerfacecolor='red',
    markeredgecolor='black',
    label='Profit'
)

plt.xlabel("Months", fontsize=14, color='purple')
plt.ylabel("Amount", fontsize=14, color='purple')

plt.title(
    "Monthly Sales vs Profit Trend",
    fontsize=15,
    fontweight="bold",
    color='black'
)

plt.legend(title="Monthly Report",  
    loc="best",
    fontsize=14,
    facecolor="darkgrey",
    edgecolor="black",
    labelcolor="purple",
    shadow=False,
    frameon=True  
          )

plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()


plt.show()
```
