## Step Plot:-

    Step Plot is used when values change at specific intervals rather than continuously.

    Real-World Uses
        Electricity Tariff Slabs
        Product Pricing Tiers
        Inventory Levels
        Machine States (ON/OFF)
        Stock Levels
        Employee Shift Changes
    

```python
import matplotlib.pyplot as plt
```

```python

hours = [1,2,3,4,5,6]
sales = [10,10,20,20,30,40]

plt.step(hours, sales)

plt.show()
```

## ATTRIBUTES:-
    WHERE
        pre  -> Step changes before x-value.
        post -> Step changes after x-value.
        mid  -> Step changes in middle.


```python
plt.step(
    hours,
    sales,
    where='pre'
)
```

```python
plt.step(
    hours,
    sales,
    where='post'
)
```

```python
plt.step(
    hours,
    sales,
    where='mid'
)
```

```python
plt.step(
    hours,
    sales,
    color='g',
    linewidth=3,
    linestyle='--',
    marker='D',
    markersize=10,
    alpha=0.8,
    label='sales',


)

plt.legend()
```

### Multiple Plots

```python

hours = [1,2,3,4,5,6]

sales = [100,100,150,150,200,250]

profit = [30,30,50,50,80,100]

plt.step(
    hours,
    sales,

    linewidth=2,
    
    where='pre',

    color='green',

    marker='o',
    mfc='magenta',
    mec='black',
    ms=10,

    label='Sales'
)

plt.step(
    hours,
    profit,
    linewidth=3,
    
    where='post',

    color='red',

    marker='D',
    mfc='blue',
    mec='black',
    ms=10,
    label='Profit'
)

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6]

sales = [100,100,150,150,200,250]

plt.figure(
    figsize=(8,5),
    facecolor='lightgrey'
)

plt.step(
    hours,
    sales,

    where='post',

    color='green',

    linewidth=3,

    linestyle='-',

    marker='D',
    mfc='red',
    mec='black',


    markersize=8,

    label='Sales'
)

plt.title(
    "Hourly Sales Trend",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel(
    "Hours",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Sales",
    fontsize=12,
    fontweight='bold'
)

plt.xticks(
    fontsize=10,
    fontweight='bold'
)

plt.yticks(
    fontsize=10,
    fontweight='bold'
)

plt.grid(
    linestyle='--',
    alpha=0.7
)

plt.legend(
    fontsize=15,
    facecolor='lightgrey',
    labelcolor='red',
    fancybox=True,
    frameon=True,
    borderpad=1,
    edgecolor='black',
    markerscale=1.5
    

  )

plt.tight_layout()

plt.savefig("StepPlot.png",dpi=150,  bbox_inches='tight',)

plt.show()
```
