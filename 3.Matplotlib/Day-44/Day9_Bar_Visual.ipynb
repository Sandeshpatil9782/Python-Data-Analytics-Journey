### Bar PLOT
    A bar chart is used to compare values across different categories.

    Real-World Uses
    Product Sales Comparison
    Region-wise Revenue
    Employee Performance
    Category-wise Profit
    Department Budget Analysis
    Business Example

    Ex:-A company wants to compare sales of different products.

```python
import matplotlib.pyplot as plt
```

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]

plt.bar(products, sales)

plt.show()
```

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]


plt.bar(products, sales,
       color='purple',
       edgecolor="black",
       width=0.5,
       alpha=0.7,
       
       )

plt.show()
```

### Different Color

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]

colors=['red','blue','green','orange']
plt.bar(products, sales,
       color=colors,
       edgecolor="black",
       width=0.4
    
       
       )

plt.show()
```

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]


plt.bar(products, sales,
       color='orange',
       edgecolor="black",
       linewidth=1,
       width=0.5,
       label='Sales'
       
       )

plt.title(
    "Product Sales Comparison",
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.xlabel(
    "Products",
    fontsize=12,
    color='navy',
    fontweight='black'
)

plt.ylabel(
    "Sales",
    fontsize=12,
    color='navy',
    fontweight='black'
    
)
plt.legend()
plt.show()
```

```python

import numpy as np

products = ["Laptop","Mobile","Tablet","Watch"]

sales = [500,700,300,450]
profit = [120,180,80,100]

x = np.arange(len(products))

plt.figure(figsize=(6,4))

plt.bar(
     x,
    sales,
    width=0.4,
    color='green',
    label='Sales',
    linewidth=1,
    edgecolor='black'
)

plt.bar(
    x ,
    profit,
    width=0.4,
    color='orange',
    label='Profit',
     linewidth=1,
    edgecolor='black'
)

plt.xticks(x, products)

plt.title(
    "Sales vs Profit Comparison",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Products",fontweight='bold')
plt.ylabel("Amount",fontweight='bold')

plt.legend(
    title="Monthly Report",
    loc="best",
    fontsize=12,
    facecolor="lightyellow",
    edgecolor="black"
)



plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

products = ["Laptop","Mobile","Tablet","Watch"]

sales = [500,700,300,450]
profit = [120,180,80,100]

x = np.arange(len(products))

plt.figure(figsize=(8,5))

plt.bar(
    x - 0.2,
    sales,
    width=0.4,
    color='green',
    label='Sales',
    linewidth=1,
    edgecolor='black'
)

plt.bar(
    x + 0.2,
    profit,
    width=0.4,
    color='orange',
    label='Profit',
    linewidth=1,
    edgecolor='black'
)

plt.xticks(x, products)

plt.title(
    "Sales vs Profit Comparison",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Products",fontweight='black')
plt.ylabel("Amount",fontweight='black')

plt.legend(
    title="Monthly Report",
    loc="best",
    fontsize=12,
    facecolor="grey",
    edgecolor="black"
)



plt.tight_layout()
plt.show()
```

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]

plt.figure(
    figsize=(6,4),
    facecolor='lightgrey'
)

plt.bar(products, sales,
       color='grey',
       edgecolor="blue",
       linewidth=1,
       linestyle='dashed',
       width=0.4,
       label='Sales'
       
       )

plt.title(
    "Product Sales Comparison",
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.xlabel(
    "Products",
    fontsize=12,
    color='navy',
    fontweight='black'
)

plt.ylabel(
    "Sales",
    fontsize=12,
    color='navy',
    fontweight='black'
    
)

plt.xticks(
    rotation=45,
    fontsize=10,
    fontweight='bold',
    color='blue'
)

plt.yticks(
    fontsize=10,
    fontweight='bold',
    color='blue'
)




plt.legend(
    title="Monthly Report",
    loc="best",
    fontsize=12,
    facecolor="yellow",
    edgecolor="blue",
    labelcolor="blue",
    shadow=True,
    frameon=True
)

plt.tight_layout()
plt.show()
```

```python
products = ["Laptop","Mobile","Tablet","Watch"]
sales = [500,700,300,450]

plt.figure(
    figsize=(6,4),
    facecolor='lightgrey'
)

plt.bar(products, sales,
       color='grey',
       edgecolor="blue",
       linewidth=1,
       width=0.4,
       label='Sales'
       
       )

plt.title(
    "Product Sales Comparison",
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.xlabel(
    "Products",
    fontsize=12,
    color='navy',
    fontweight='black'
)

plt.ylabel(
    "Sales",
    fontsize=12,
    color='navy',
    fontweight='black'
    
)

plt.xticks(
    rotation=45,
    fontsize=10,
    fontweight='bold',
    color='black'
)

plt.yticks(
    fontsize=10,
    fontweight='bold',
    color='black'
)




plt.legend(
    title="Monthly Report",
    loc="best",
    fontsize=12,
    facecolor="yellow",
    edgecolor="black",
    labelcolor="blue",
    shadow=True,
    frameon=True
)


plt.tight_layout()
plt.show()
plt.savefig(
    "BarChart.png",
    dpi=300
)

print("Fighure Saved Successfully")
```
