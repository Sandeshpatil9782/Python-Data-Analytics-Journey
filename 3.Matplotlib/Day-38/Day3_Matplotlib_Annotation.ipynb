```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Fill Between Graph
    Used to fill color between line and x-axis

```python
x = [1,2,3,4,5]
y = [2,4,3,5,4]
                                            
plt.plot(x,y,color="black")
plt.fill_between(x,y,alpha=0.5,color="green")    # alpha=0.5 -> Transparency level and Range: 0 to 1

plt.show()
```

### Stack Plot
    -Shows Cumulative Data
    -Creates a stacked area chart for multiple datasets

```python
days = [1,2,3,4,5]

study = [2,3,4,3,2]
sleep = [8,7,6,8,7]

plt.stackplot(
    days,
    study,
    sleep,
    labels=["Study","Sleep"],edgecolor="purple",)

plt.legend()
plt.show()
```

### Stem Plot
    Useful for descrete signal
    linefmt="y-"   ->   Stem line color = yellow
    markerfmt="go" ->   Marker color = green-g
                        Marker shape = Circle-o
    basefmt="k-"   ->   Base line color = Black

```python
x = [1,2,3,4,5]
y = [5,3,8,2,7]

plt.stem(
    x,
    y,
    linefmt="y-",
    markerfmt="go",
    basefmt="k-"
)


plt.show()
```

### Error Bar Plot
    Shows error/ in data

```python
x = [1,2,3,4]
y = [10,20,15,25]

error = [1,2,1,3]

plt.errorbar(
    x,
    y,
    yerr=error,
    fmt="o"
)

plt.show()
```

### Annotation
    Adds labels to specific points on a graph
    arrowprops=dict(arrowstyle="->") Gives arrow for address

    Text on Graph Using plt.text() ->Adds text anywhere on a graph

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y,color="green")

plt.annotate(
    "Lowest",
    xy=(1,10)
)

plt.text(
    2.25,
    17,
    "Center Point"
)

plt.annotate(
    "Highest",
    xy=(4,25)
)

plt.show()
```

### Axis Customization
    plt.xlim()	Sets the x-axis range	
    plt.ylim()	Sets the y-axis range

```python
plt.plot([1,2,3,4],[10,20,15,30])

plt.xlim(0,5)
plt.ylim(0,40)

plt.show()
```

### Tick Customization
    plt.xticks()	Customizes x-axis tick positions and labels	
    plt.yticks()	Customizes y-axis tick positions and labels

```python
plt.plot([1,2,3,4],[10,20,15,30])

plt.xticks(
    [1,2,3,4,5],
    ["Mon","Tue","Wed","Thu","Fri"]
)
plt.yticks([0,10,20,30,40])

plt.show()
```

### Multiple Figures

```python
plt.figure(1)
plt.style.use("fivethirtyeight")
plt.plot([1,2,3],[4,5,6],color="green")

plt.figure(2)
plt.style.use('classic')
plt.bar([1,2,3],[5,6,7],color='purple',width=0.4)

plt.show()
```

### Logarithmic Scale
    plt.xscale()	Changes the x-axis scale	#Attributes:- "linear", "log"
    plt.yscale()	Changes the y-axis scale

```python
x = [1,10,100,1000]
y = [1,2,3,4]

plt.plot(x,y)

plt.xscale("log")

plt.show()
```
