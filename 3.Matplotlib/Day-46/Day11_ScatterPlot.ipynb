```python
hours = [1,2,3,4,5,6,7,8]
marks = [35,40,55,60,50,65,75,92]


sizes = [50,80,100,120,150,180,200,220]

colors = [29,35,39,56,63,70,89,79]

plt.scatter(
           hours, marks,
           edgecolor='black',
           marker="o",
           s=sizes, 
           c=colors,
           cmap='BrBG' # Shows Distrubuted colors
             
         )

t=plt.colorbar()

t.set_label("ColorBar",fontsize=15)

plt.show()
```

```python
hours = [1,2,3,4,5]

boys_marks = [40,78,60,80,87]
girls_marks = [35,49,69,47,95]

plt.figure(figsize=(6,4),facecolor='lightgreen')

plt.scatter(
    hours,
    boys_marks,
    color='blue',
    edgecolor='black',
    marker='o',
    s=200,
    label='Boys'
)

plt.scatter(
    hours,
    girls_marks,
    color='red',
    edgecolor='black',
    marker='^',
    s=100,
    label='Girls'
)

plt.title(
    "Study Hours vs Marks",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Study Hours",fontsize=12,fontweight='bold')
plt.ylabel("Marks",fontsize=12,fontweight='bold')

plt.legend(
    title='Student Data',
    facecolor='lightgrey',
    edgecolor='black'
)

plt.xticks(
    [1,2,3,4,5],
    ['1 Hr','2 Hr','3 Hr','4 Hr','5 Hr']
)

plt.grid(True,alpha=0.2)
plt.show()
```

```python

hours = [1,2,3,4,5,6,7,8]
marks = [35,40,55,60,50,65,75,92]

plt.figure(
    figsize=(6,4),
    facecolor='lightgreen'
)


plt.scatter(
    hours,
    marks,
    color='blue',
    edgecolor='black',
    linewidth=1,
    s=250,
    marker='o',
    alpha=0.8,
    label='Students'
)

plt.title(
    "Study Hours vs Marks",
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.xlabel(
    "Study Hours",
    fontsize=12,
    fontweight='bold'
)

plt.ylabel(
    "Marks",
    fontsize=12,
    fontweight='bold'
)

legend=plt.legend(
    title='Student Data',

    facecolor='lightgrey',
    edgecolor='black',
    markerscale=1,
    labelcolor="g",
    borderpad=1,
    labelspacing=1
)
legend.get_title().set_color('red')

plt.xticks(
    [1,2,3,4,5,6,7,8],
    ['1 Hr','2 Hr','3 Hr','4 Hr','5 Hr','6 Hr','7 Hr','8 Hr']
)

plt.grid(
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()


plt.savefig("ScatterPlot.png",dpi=150)
plt.show()
```
