## Stem Plot:-

    Stem Plot is used to show discrete data points.

    Every Value Has:
           One vertical line (stem)
           One marker (head)
    Can be displayed.
    
    Real-World Uses
        Daily Defect Counts
        Sensor Readings
        Signal Processing
        Inventory Counts
        Website Errors Per Day

```python
from matplotlib import pyplot as plt
```

```python
days = [1,2,3,4,5]
errors = [2,5,3,8,6]

plt.stem(
    days,
    errors
)

plt.show()
```

### ATTRIBUTES:-
    linefmt='g-'     -> Controls Line Style(Stem)
                        'r-'  -> Red solid
                        'g--' -> Green dashed
                        'b:'  -> Blue dotted
                        'k-.' -> Black dash-dot
                        
    markerfmt='ro'   ->  Marker Style
                        'ro',   'gs',   'b^', 'kD',  'm*'
                        
    basefmt='black'  -> Control Baseline

    bottom=0         -> Moves Baseline
    

    

```python
plt.stem(
    days,
    errors,
    linefmt='g--',
    markerfmt='ro',
    basefmt='black',
    label='Errors',
    bottom=1,
    orientation='vertical',
)
```

```python
plt.stem(
    days,
    errors,
    linefmt='g-',
    markerfmt='bD',
    basefmt='magenta',
    label='Errors',
    bottom=0,
    orientation='horizontal',
)
```

### Advance Styling

```python
markerline, stemlines, baseline = plt.stem(
    days,
    errors
)
stemlines.set_linewidth(3)
stemlines.set_color('blue')
stemlines.set_linestyle(':-')

markerline.set_color('yellow')
markerline.set_marker('^')
markerline.set_markersize(15)
markerline.set_markeredgecolor('black')
markerline.set_markeredgewidth(1)


baseline.set_color('magenta')
baseline.set_linewidth(4)
```

```python
# Negative Data

days = [1,2,3,4,5,6,7]
errors = [2,-5,3,8,-6,4,-7]

markerline, stemlines, baseline = plt.stem(
    days,
    errors,
    bottom=0
)

markerline.set_marker('D')
markerline.set_markersize(10)
markerline.set_markerfacecolor('gold')
markerline.set_markeredgecolor('black')
markerline.set_markeredgewidth(2)

plt.setp(
    stemlines,
    color='green',
    linewidth=3
)

baseline.set_color('red')
baseline.set_linewidth(2)

plt.show()
```

#### Multiple Stem

```python
import matplotlib.pyplot as plt

months = [1,2,3,4,5]

sales = [10,15,12,20,18]
profit = [4,7,5,10,8]

# Sales Stem Plot
markerline1, stemlines1, baseline1 = plt.stem(
    months,
    sales,
    linefmt='g-',
    markerfmt='g^',
    basefmt='black',
    label='Sales'
)
markerline1.set_markersize(12)
markerline1.set_markeredgecolor('black')

stemlines1.set_color('blue')

# Profit Stem Plot
markerline2, stemlines2, baseline2 = plt.stem(
    months,
    profit,
    linefmt='--',
    markerfmt='rD',
    basefmt='black',
    label='Profit'
)

markerline2.set_markersize(10)
markerline2.set_markeredgecolor('black')

stemlines2.set_color('brown')
stemlines2.set_linewidth(3)

plt.legend(
    fontsize=15,
    facecolor="yellow")

plt.show()
```

```python
days = [1,2,3,4,5,6,7]

errors = [2,5,3,8,6,4,7]

plt.figure(
    figsize=(8,5),
    facecolor='lightgrey'
)

markerline,stemlines,baseline=plt.stem(
                                        days,
                                        errors,
                                        label='Website Errors',
                                        )

markerline.set_marker('D')
markerline.set_markersize(10)
markerline.set_markerfacecolor('gold')
markerline.set_markeredgecolor('black')
markerline.set_markeredgewidth(2)

plt.setp(
    stemlines,
    color='green',
    linewidth=3
)

baseline.set_color('red')
baseline.set_linewidth(2)


plt.title(
    "Daily Website Errors",
    fontsize=15,
     color='navy',
    fontweight='bold'
)

plt.xlabel(
    "Days",
    fontsize=12,
     color='blue',
    fontweight='bold'
)

plt.ylabel(
    "Error Count",
    fontsize=12,
    color='blue',
    fontweight='bold'
)

plt.xticks(
    fontsize=10,
    fontweight='bold'
)

plt.yticks(
    fontsize=10,
    fontweight='bold'
)

plt.grid(
    linestyle='--',
    alpha=0.6
)

plt.legend(facecolor='lightgrey')

plt.tight_layout()

plt.savefig('StemPlot.png',dpi=150)

plt.show()
```
