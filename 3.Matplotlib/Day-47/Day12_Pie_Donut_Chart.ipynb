### PIE CHART
    Pie chart is used to show part-to-whole relationships.

        Company Revenue by Product
        Market Share by Brand
        Department Budget Allocation
        Student Distribution by Course

        Ex:-
        Revenue Share by Product
        Contribution of each product to total revenue.

```python
import matplotlib.pyplot as plt
```

### Dot Pie chart

```python
plt.pie([1])

plt.show()
```

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

plt.pie(
    sales
    
)

plt.show()
```

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

plt.pie(
    sales,
    colors=['skyblue','orange','magenta','purple'],
    labels=products,
    
    autopct='%1.1f%%',   # Shows Percentage

    startangle=90        # Rotate Chart
)

plt.show()
```

### Highlight Common Slice

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

explode = [0,0.1,0,0] 

plt.pie(
    sales,
    colors=['skyblue','orange','green','purple'],
    labels=products,
    autopct='%1.2f%%',
    explode=explode
    
)


plt.show()
```

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

plt.pie(
    sales,
    colors=['skyblue','orange','magenta','purple'],
    labels=products,
    autopct='%1.1f%%',   
    startangle=90,
    shadow=True,
    radius=1.5,    # Default/Normal 1
    counterclock=False,
    labeldistance=1.06      
    
    
)

plt.show()
```

### Slice Border Style
    wedgeprops={   }
    
    edgecolor
    linewidth
    linestyle
    center

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

plt.pie(
    sales,
    colors=['skyblue','orange','magenta','purple'],
    labels=products,
    autopct='%1.1f%%',   
    startangle=0,
    radius=1.5,
    counterclock=True,
   
    # Slice Border Style
    wedgeprops={
        'edgecolor':'black',
        'linewidth':2,
        'linestyle':'--'
        
    }
    
    
)

plt.show()

```

### Label Styling
     textprops={ }
        fontsize
        fontweight
        color

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

plt.pie(
    sales,
    colors=['skyblue','orange','magenta','purple'],
    labels=products,
    autopct='%1.1f%%',   
    startangle=0,
    radius=1.5,
    counterclock=True,
    
    rotatelabels=True,  # Rotate The Labels

     wedgeprops={
        'edgecolor':'black',
         
     },
    
     textprops={
        'fontsize':12,
        'fontweight':'bold',
        'color':'black'
         
    }
    
    
)

plt.show()

```

### Multiple OR Donut Chart

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]

sales2=[45,56,23,12]

plt.pie(
    sales,
    labels=products,
    radius=1.5
    
)
c=['brown','y','m','purple']

plt.pie(
    sales,
    radius=1,
    colors=c
    
)


plt.show()
```

```python
products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]


plt.pie(
    sales,
    labels=products,
    radius=1.9
    
)

plt.pie([1],colors='w')

plt.title("Donut Chart Using Another pie",fontsize=15,fontweight='bold',pad=100)
plt.show()
```

```python
# Using Circle Function

products = ['Laptop','Mobile','Tablet','Accessories']
sales = [40,30,20,10]


c=['brown','y','m','purple']

plt.pie(
    sales,
    labels=products,
    radius=1.5,
    colors=c
    
)

cr=plt.Circle(xy=(0,0),facecolor='w',radius=1)


plt.gca().add_artist(cr)

plt.title('Donut Chart Using Circle funtion',fontsize=15,pad=50,fontweight="bold")

plt.show()
```

```python
import matplotlib.pyplot as plt

products = [
    'Laptop',
    'Mobile',
    'Tablet',
    'Accessories'
]

sales = [40,30,20,10]

colors = [
    'cornflowerblue',
    'orange',
    'lightgreen',
    'plum'
]

plt.figure(
    figsize=(10,8),
    facecolor='lightgrey'
)

plt.pie(
    sales,
    labels=products,
    colors=colors,
    autopct='%1.1f%%',
    startangle=0,
    wedgeprops={
        'width':0.6,
        'edgecolor':'black',
        'linewidth':1
    }
)

plt.title(
    "Product Sales Distribution With Donut",
    fontsize=15,
    fontweight='bold',
    color='navy'
)
plt.legend(
    loc='upper left',
    facecolor="lightgrey"
)
plt.savefig("Donut Chart.png", dpi=150)

plt.show()
```

```python

products = [
    'Laptop',
    'Mobile',
    'Tablet',
    'Accessories'
]

sales = [40,30,20,10]


colors = [
    'cornflowerblue',
    'orange',
    'lightgreen',
    'plum'
]

plt.figure(
    figsize=(8,6),
    facecolor='lightgrey'
)

plt.pie(
    sales,
    labels=products,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    #shadow=True,
    
    wedgeprops={
        'edgecolor':'black',
        'linewidth':1.5
    },
    textprops={
        'fontsize':11,
        'fontweight':'bold'
    }
)

plt.title(
    "Product Sales Distribution With Pie",
    fontsize=15,
    fontweight='bold',
    color='navy'
)

plt.tight_layout()



plt.legend(loc='best',
          facecolor='lightgrey')

plt.show()
```
