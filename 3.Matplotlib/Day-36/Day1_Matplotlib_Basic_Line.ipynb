## PYTHON MATPLOTLIB LIBRARY
    Data Visualization And Plotting Library

### Importing 
    Library :- matplotlib
    Module  :- pyplot
    Alias   :- plt

```python
import matplotlib.pyplot as plt
```

### Line Plot
    Show Trends Over Time
    plot() :- To Draw The Graph

    show() :- To Display The Plot

```python
x = [1,2,3,4]  # Horizontal X-axis Values
y = [10,20,15,25] # Vertical Y-axis Values

plt.plot(x, y)

plt.show()
```

### LABELS
    Title :- Monthly Sales
    XLABEL :- Months
    YLABEL :- Sales

```python
x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x, y)

plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Add color
    Common Color :- Red,Green,Blue,Black,Orange,Purple

```python
x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x, y, color="Red")

plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Add Marker
    Important Markers:-
        o = Circle
        * = Star
        + = Plus
        s = Square
        ^ = Triangle
        

```python
x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x, y, color="Red",marker="o")

plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Line Style
    Styles:-
        -    = Solid
        --   = Dashed
        :    = Dotted
        -.   = Dash Dot

```python
x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x, y, color="Green", marker="o", linestyle="--" )



plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Add Grid 
    - grid() is a Matplotlib function used to display grid lines on a graph for better readability and easier data analysis.
    - Grid lines helps user to read graph more clearly.
    Grid Types:-
        1.Default :- plt.grid()
        2.Enable  :- plt.grid(True)
        3.Disable :- plt.grid(False)
    
    Style Grid
        -Grid Color :- plt.grid(color="red")
        -Grid Line  :- plt.grid(linestyle="--")
        -Grid Size  :- plt.grid(linewidth=2)

    Axis Example
        Only X-axis grid :- plt.grid(axis="x")
        Only Y-axis grid :- plt.grid(axis="y")
        Both:            :- plt.grid(axis="both")


```python
x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x, y, color="Green", marker="o", linestyle="--" )


plt.grid()
plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Legend
    - legend() is used to show labels for plotted data, making graphs easier to understand and analyze.
        
    - A label is used to identify plotted data in a graph and is displayed using the legend() function.
    - A = [1,2,3,4]  
      A single X-axis can be used to compare multiple datasets like sales and profit on the same graph.

```python
A = [1,2,3,4]

sales = [10,20,15,25]
profit = [5,15,10,20]

plt.plot(
    A,
    sales,
    marker="s",
    color="blue",
    label="Sales"
)

plt.plot(
    A,
    profit,
    marker="^",
    color="red",
    label="Profit"
)
plt.title("Monthly Sales")
plt.legend()

plt.show()
```

### Figure Size
    figure(figsize=()) is used to set the width and height of a graph in Matplotlib.

```python
x = [1,2,3,4]
y = [10,20,15,20]

plt.figure(figsize=(10,5)) # Width = 10 AND Height = 5

plt.plot(x, y,color="Green", marker="o")

plt.title("Monthly Sales")

plt.xlabel("Months")
plt.ylabel("Sales")


plt.show()
```

### Save Fugure
    savefig() is used to export and save visualizations in image formats such as PNG, JPG, or PDF.
    DPI-Dot Per Inch
    72  - Lower
    150 - Medium
    300 - Higher Quality

```python
A = [1,2,3,4]

sales = [10,20,15,25]
profit = [5,15,10,20]

plt.plot(
    A,
    sales,
    marker="s",
    color="blue",
    label="Sales"
)

plt.plot(
    A,
    profit,
    marker="^",
    color="red",
    label="Profit"
)
plt.title("Monthly Sales")
plt.legend()

plt.savefig("LineGraph.png",dpi=300)
plt.show()
```

### Example

```python
months = ["Jan","Feb","Mar","Apr","May"]

sales = [12000,15000,18000,17000,22000]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="blue",
    marker="o",
    linestyle="--",
    linewidth=2,
    label="Monthly Sales"
)

plt.title("Company Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.grid()
plt.legend()

plt.show()
```
