### PANDAS-Series
    Pandas is a Python library used for data analysis and data manipulation.

### Importing Pandas
    Library:-pandas
    Shortcut Alias:-pd

```python
import pandas as pd
print("Pandas Version:",pd.__version__)
```

### SERIES
    Series is a one-dimensional labeled array.

### Creating Simple Series

```python
a = pd.Series([10,20,30])

print(a)
```

### Access Value

```python
print(a[0]) #Gives 0 index value 10
```

```python
print(a[2])
```

### Custom Index
    Puting Index label for each values

```python
a = pd.Series(
    [100,200,300],
    index=['a','b','c']
)

print(a)
```

### Accesing Using Index Label

```python
print(a['a']) 
```

```python
print(a['c'])
```

### Series From Dictionary
    Dictionary is a KEY VALUE pair structure.

```python
data = {
    'Math':90,
    'Science':85,
    'English':88
}

s = pd.Series(data)

print(s)
```

### Series Atributes

```python
print(a.index) # Gives  Series Indexes

print(a.values) # Gives Series Values

print(a.dtype) # Gives Series Datatype used
```

###  Create Series using list

```python
import pandas as pd

data = [100, 200, 300]

s = pd.Series(data, index=['A', 'B', 'C'])

print(s)
```

### Create Series using tuple

```python
data = (1, 2, 3, 4)
s2 = pd.Series(data)

print(s2)
```

### Access last element

```python
import numpy as np
data = (100, 200, 300, 400)
s2 = pd.Series(data)

print(s2.iloc[-1])
```

### Slicing The Serie

```python
import numpy as np
data = (100, 200, 300, 400)
s1 = pd.Series(data)

print(s1[1:3])
```

### Create Series of strings

```python
names = ["Ram", "Shyam", "Mohan"]
s10 = pd.Series(names)

print(s10)
```

## Create Series using NumPy array

```python
arr = np.array([5, 10, 15,20,25,30,35])
print("Array:",arr,"\n")
s11 = pd.Series(arr)

print(s11)
```

### Find Size Of Series

```python
print(s11.size)
```

### Find Shape

```python
print(s11.shape)
```

### Find Dimension

```python
print(s11.ndim)
```

### Count The Values

```python
print(s11.count())
```

### Find Sum Of Series

```python
print(s11.sum())
```

### Maximum Value

```python
print(s11.max())
```

### Minimum Value

```python
print(s11.min())
```

### Mean Of Series Value

```python
print(s11.mean())
```

### Sorting The Series in Ascending Order

```python
s20 = pd.Series([40, 10, 30, 20])

print(s20.sort_values())
```

### The Series in Descending Order

```python
print(s20.sort_values()[: : -1])
```
