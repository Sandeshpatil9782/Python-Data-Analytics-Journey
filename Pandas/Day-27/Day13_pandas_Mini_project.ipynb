```python
import numpy as np

import pandas as pd

print(np.__version__)
print(pd.__version__)
```

### Creation Of Data

```python
data = {
    'Date': [
        '2025-01-01','2025-01-02','2025-01-03',
        '2025-01-04','2025-01-05','2025-01-06',
        '2025-01-07','2025-01-08','2025-01-09',
        '2025-01-10'
    ],

    'Product': [
        'Laptop','Mouse','Laptop',
        'Keyboard','Mouse','Monitor',
        'Laptop','Keyboard','Mouse',
        'Monitor'
    ],

    'Sales': [
        50000,1500,52000,None,1700,
        12000,55000,3000,None,
        14000
    ],

    'City': [
        'Pune','Mumbai','Delhi',
        'Bangalore','Chennai','Hyderabad',
        'Pune','Mumbai','Delhi',
        'Chennai'
    ],

    'Quantity': [
        2,5,1,3,4,
        2,1,2,6,
        1
    ],

    'Profit': [
        5000,200,6000,None,250,
        1500,7000,400,None,
        1800
    ]
}

df = pd.DataFrame(data)
```

### To SAVE CSV

```python
df.to_csv('Mini.csv',index=False) # Without Index Saved
```

### Convert Date Column

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

### Check Missing Values

```python
df.isnull().sum()
```

### Fill Missing 
    Filled With Mean

```python
df['Sales'] = df['Sales'].fillna(
    df['Sales'].mean()
)

df['Profit'] = df['Profit'].fillna(
    df['Profit'].mean()
)

df
```

### Total Sales

```python
Total=df['Sales'].sum()
print(Total)
```

### Average Sales

```python
print(df['Sales'].mean())
```

### Product-Wise Sales

```python
df.groupby('Product')['Sales'].sum()
```

### City-Wise Sales

```python
print(df.groupby('City')['Sales'].sum())
```

#### Highest Selling Product

```python
df.groupby('Product')['Sales'].sum().idxmax()
```

### Lowest Selling Product

```python
df.groupby('Product')['Sales'].sum().idxmin()
```

### Daily Sales Trend

```python
df.groupby('Date')['Sales'].sum()
```

### Sales Category

```python
df['Category'] = pd.cut(
    df['Sales'],
    bins=[0,5000,50000,100000],
    labels=['Low','Medium','High']
)

df
```

### Highest Sales

### Filter Data

```python
print(df[df['Sales'] > 10000])
```

### Multiple Conditions

### GST Column Added

```python
df['GST'] = df['Sales'] * 0.18
df
```

### Discount Column Added

```python
df['Discount'] = df['Sales'] * 0.05
df
```

### Profit Percentage

```python
df['Profit_Percentage'] = np.round(((df['Profit']/df['Sales'])*100),2)
df
```

### Extract Year

```python
df['Year'] = df['Date'].dt.year
df
```

### Extract Month

```python
df['Month'] = df['Date'].dt.month
df
```

### Extract Day

```python
df['Day'] = df['Date'].dt.day
df
```

### Pivot Table

```python
print(pd.pivot_table(df,
                     values='Sales',
                     index='Product',
                     columns='City',
                     aggfunc='sum'))

print("\n\nMising Values Fill with Zero:-\n\n",pd.pivot_table(df,
                     values='Sales',
                     index='Product',
                     columns='City',
                     aggfunc='sum',
                     fill_value=0
                    ))
```
