### Fill Missing Values

```python
df["Sales"] = df["Sales"].fillna("0") # Filled O
```

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())  #Fill Mean Value
```

### Count Duplicate

### Round Sales Column

```python
df['Sales']=df["Sales"].round(2)
df.head()
```

### Delete Duplicatese

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

### Clean Text Column

```python
df["City"] = df["City"].str.strip()
df.head()
```

```python
df["Product"] = df["Product"].str.lower()
df.head()
```

```python
df["Product"] = df["Product"].str.upper()
df.head()
```

```python
df["Product"] = df["Product"].str.title()
df.head()
```

### Remove Extra Space

```python
df=df.apply(lambda x:x.str.strip() if x.dtype=="object" else x)
df.head(10)
```

### Inconsistent Cleaning

```python
df['City']=df['City'].replace(
    {
        "pune":"Pune"
    }
)
df.head(6)
```

### Check Unique Values

```python
print(df["City"].unique())

print("\n",df.nunique())
```

```python
print(df["Product"].unique())
```

### To Change Date Column

```python
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df["Year"] = df["Date"].dt.year
df.head()
```

```python
df["Month"] = df["Date"].dt.month
df.head()
```

### Summarize Data

```python
pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    columns="City",
    aggfunc="sum",
    fill_value=0
)
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)

print(df.duplicated().sum())
```

```python
df.duplicated().any()
```

### Top Products

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
```

### Top Sales By City

```python
df.groupby("City")["Sales"].sum().sort_values(ascending=False)
```
