### Fix Data Types

### Inches

```python
df["Inches"].unique()
```

```python
df["Inches"] = pd.to_numeric(
    df["Inches"],
    errors="coerce"
)                        # ? == nan convert
```

```python
df["Inches"].unique()
```

```python
df["Inches"] = df["Inches"].astype(float)
```

### Formating Values

#### Ram Column

```python
df["Ram"] = df["Ram"].str.replace("GB"," GB")
```

#### Weight Column

```python
df["Weight"] = df["Weight"].str.replace("kg"," KG")
```

### Cpu Column
     r'(\d+\.\d+)GHz' is a Regular Expression (Regex) pattern used to extract

```python
df["Cpu_Brand"] = df["Cpu"].apply(
    lambda x: x.split()[0] if pd.notna(x) else None
)
```

```python
df["Cpu_Speed"] = df["Cpu"].str.extract(r'(\d+\.\d)')[0] + " GHz"

df["Cpu_Type"] = df["Cpu"].str.extract(r'(Core i[3579])')
```

```python
df["Cpu_Speed"] = df["Cpu_Speed"].fillna(0)

df["Cpu_Type"] = df["Cpu_Type"].fillna(df["Cpu_Type"].mode()[0])
```

```python
df.drop(["Cpu"], axis=1, inplace=True)
```

### Screen Resolution

```python
df["Display_Type"] = df["ScreenResolution"].str.extract(r'^(.*)\s(\d+x\d+)$')[0]
```

```python
df["Resolution"] = df["ScreenResolution"].str.extract(r'(\d+x\d+)')
```

```python
df["X_res"] = df["Resolution"].apply(
    lambda x: x.split("x")[0] if pd.notna(x) else None
)

df["Y_res"] = df["Resolution"].apply(
    lambda x: x.split("x")[1] if pd.notna(x) else None
)
```

#### Calculate PPI (Professional Feature Engineering)
    Pixel Per Inch

```python
 df["X_res"] = pd.to_numeric(df["X_res"], errors="coerce")
df["Y_res"] = pd.to_numeric(df["Y_res"], errors="coerce")
df["Inches"] = pd.to_numeric(df["Inches"], errors="coerce")
```

```python
import numpy as np

df["PPI"] = (
    np.sqrt(df["X_res"]**2 + df["Y_res"]**2)
    / df["Inches"]
)
df["PPI"] = df["PPI"].round(3)
```

```python
df["Display_Type"] = df["ScreenResolution"].apply(
    lambda x: "Unknown" if pd.isna(x)
    else "Retina" if "Retina Display" in x
    else "IPS" if "IPS Panel" in x
    else "Full HD" if "Full HD" in x
    else "Quad HD+" if "Quad HD+" in x
    else "Standard"
)
```

```python
df.drop(["ScreenResolution"], axis=1, inplace=True)
```

### Clean Memory Column

```python
df["Storage_Type"] = df["Memory"].str.extract(
    r'(SSD|HDD|Flash Storage|Hybrid)'
)
```

```python
df["Storage_Size"] = df["Memory"].str.extract(r'(\d+\s?(?:GB|TB))')[0]

df["Storage_Size"] = (
    df["Storage_Size"]
    .str.replace("GB", " GB")
    .str.replace("TB", " TB")
    .str.strip()
)
```

```python
df.fillna({
    "Storage_Type": df["Storage_Type"].mode()[0],
    "Storage_Size": df["Storage_Size"].mode()[0],
  
}, inplace=True)
```

```python
df.drop(["Memory"], axis=1, inplace=True)
```

### Gpu column

```python
df["Gpu_Brand"] = df["Gpu"].str.extract(r'^(Intel|Nvidia|AMD)')
```

```python
df["Gpu_Model"] = df["Gpu"].str.replace(
    r'^(Intel|Nvidia|AMD)\s+',
    '',
    regex=True
)
```

```python
df["Gpu_Brand"] = df["Gpu_Brand"].fillna(df["Gpu_Brand"].mode()[0])
```

```python
df.drop(["Gpu"], axis=1, inplace=True)
```

### Clean Operating System Column

#### Group Categories

```python
def os_category(x):
    if pd.isna(x):
        return "Unknown"
    elif "Windows" in x:
        return "Windows"
    elif "Mac" in x or "macOS" in x:
        return "Mac"
    elif "Linux" in x:
        return "Linux"
    else:
        return "Other"

df["OS"] = df["OpSys"].apply(os_category)
```

```python
df.drop(["OpSys"], axis=1, inplace=True)
```

### Price Column

```python
df["Price"] = df["Price"].astype(float)
df["Price"] = df["Price"].round(2)
```
