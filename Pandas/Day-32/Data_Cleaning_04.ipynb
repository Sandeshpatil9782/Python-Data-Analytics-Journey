```python
import pandas as pd
import random

random.seed(42)

names = [
    "rahul sharma", "RAHUL SHARMA", "Rahul  Sharma",
    "pooja patil", "AKSHAY kumar", "Sanjay Patil"
]

genders = ["M", "F", "Male", "Female", "male", "female", None]

departments = [
    "Data Analyst",
    "data analyst",
    "DATA ANALYST",
    "Busines Analyst",
    "Business Analyst",
    None
]

cities = [
    "Mumbai",
    "mumbai",
    " MUMBAI ",
    "Pune",
    "pune",
    "Bangalore",
    "belgaum",
    None
]

emails = [
    "USER@GMAIL.COM",
    " test@mail.com ",
    "employee@company.com",
    "invalid_email",
    None
]

dates = [
    "2024-01-15",
    "15/01/2024",
    "Jan 15 2024",
    None
]

rows = []

for i in range(2000):

    emp_id = 10000 + i

    if i % 150 == 0 and i > 0:
        emp_id = 10000 + i - 1   # duplicate IDs

    rows.append({
        "Employee_ID": emp_id,
        "Name": random.choice(names),
        "Age": random.choice([22, 25, 30, 35, 45, None, -5, 150]),
        "Gender": random.choice(genders),
        "Department": random.choice(departments),
        "City": random.choice(cities),
        "Salary": random.choice([25000, 30000, 45000, 60000, None, 5000000]),
        "Experience_Years": random.choice([1, 2, 5, 8, None, -3, 40]),
        "Email": random.choice(emails),
        "Joining_Date": random.choice(dates)
    })

df = pd.DataFrame(rows)

df.to_csv("employee_dirty_2000_rows.csv", index=False)

print("CSV Created Successfully")
```

### Formating Inconsistent Values

```python
df.columns = df.columns.str.strip()
```

```python
df["Name"] = df["Name"].str.strip().str.title()
df.head()
```

```python
df["Department"] = df["Department"].str.strip().str.title()
```

```python
df["City"] = df["City"].str.strip().str.title()
```

```python
df["Email"] = df["Email"].str.strip().str.lower()
```

```python
df["Email"].unique()
```

```python
df["Gender"] = df["Gender"].str.strip().str.title()
```

### Removing Invalid Age

```python
df = df[
((df["Age"] >= 18) &
(df["Age"] <= 60))
]
```

```python
df['Age'].unique()
```

```python
# Invalid Experience Years
df = df[
df["Experience_Years"] >= 0
]
```

### Spliting And Removing

```python
Name_split = df["Name"].str.split(" ", expand=True)
```

```python
df.insert(1, "First_Name", Name_split[0])

df.insert(2, "Last_Name", Name_split[1])
df.head()
```

#### Drop Full Name Column

```python
df.drop("Name",axis=1,inplace=True)
df.head()
```

#### Removing Duplicates

```python
df[df.duplicated(subset=["Employee_ID"], keep=False)]
```

```python
df["Employee_ID"].duplicated().sum()
```

```python
df = df.drop_duplicates(
    subset=["Employee_ID"],
    keep="first"
)
```

```python
df["Employee_ID"].duplicated().any()
```

### Replacing Values

```python
df["Gender"].unique()
```

```python
df["Gender"] = df["Gender"].replace(
{
"M":"Male",
"male":"Male",
"F":"Female",
"female":"Female"
}
)
```

```python
df["Gender"].unique()
```

```python
df["Department"].unique()
```

```python
df["Department"] = df["Department"].replace(
{
"Busines Analyst":"Business Analyst"
}
)
```

```python
df["Department"].unique()
```

### Filling NAN Values

```python
df["Salary"] = df["Salary"].fillna(
df["Salary"].median()
)
```

```python
df["Gender"] = df["Gender"].fillna(
    df["Gender"].mode()[0]
)
```

```python
df["Department"] = df["Department"].fillna(
    df["Department"].mode()[0]
)
```

```python
df["City"] = df["City"].fillna(
    df["City"].mode()[0]
)
```

```python
df["Email"] = df["Email"].fillna("Not Available")
```

```python
df["Email"] = df["Email"].fillna("Not Available")
```

### Convert Data Types

```python
df["Joining_Date"] = pd.to_datetime(
df["Joining_Date"],
errors="coerce"
)

df["Joining_Date"].dtypes  # NaT (Not a Time) is Pandas' standard representation for missing or invalid values in datetime columns, 
                           # similar to how NaN represents missing values in numeric columns.
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df["Gender"]= df["Gender"].astype("category")
```

```python
df["City"]= df["City"].astype("category")
```

```python
df["Department"]= df["Department"].astype("category")
```
