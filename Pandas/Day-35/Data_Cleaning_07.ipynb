### Missing Values

```python
df.isnull().sum()
```

#### Numrical Values Food_Bill 

```python
df["Food_Bill"] = df["Food_Bill"].fillna(
    df["Food_Bill"].median()
)
```

#### Categorical Values Booking source

```python
df["Booking_Source"] = df["Booking_Source"].fillna(
    df["Booking_Source"].mode()[0]
)
```

### Not Null Values

```python
df.notnull().sum()
```

### Duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Datatypes

```python
df["CheckIn_Date"] = pd.to_datetime(
    df["CheckIn_Date"]
)
```

```python
df["Rating"] = df["Rating"].astype(float)
```

```python
df.dtypes
```

### String Cleaning

```python
df["City"] = df["City"].str.strip().str.title()
```

### Features

```python
df["Total_Bill"] = (
    df["Room_Rate"] * df["Nights_Stayed"]
    + df["Food_Bill"]
)
```

#### Cost Per Night

```python
df["Cost_Per_Night"] = (
    df["Total_Bill"]
    / df["Nights_Stayed"]
)
```

### Numpy Functions

```python
df["Total_Bill"].agg(["min", "max", "mean", "median", "sum"]).round(2)
```

#### Group By Analysis

```python
df.groupby("City")["Rating"].mean() # Avarage Rating By City
```

```python
df.groupby("Hotel_Name")["Total_Bill"].sum() # Revenue By Hotel
```

### Sorting
    Top Customers

```python
df = df.sort_values(by="Total_Bill", ascending=False)
df.head()
```

#### Date Analysis

```python
df["Month"] = df["CheckIn_Date"].dt.year  # Year Extract
```

```python
df["Month"] = df["CheckIn_Date"].dt.month    # Month Extract
```

```python
df["Weekday"] = df["CheckIn_Date"].dt.day_name() # Weekday Extraction
```
