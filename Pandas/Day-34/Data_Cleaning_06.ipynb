```python
df.drop_duplicates(inplace=True)
```

### Missing Values

#### Matches

```python
df["Matches"] = df["Matches"].fillna(
    df["Matches"].median()
)
```

#### Strike Rate

```python
df["StrikeRate"] = df["StrikeRate"].fillna(
    df["StrikeRate"].mean()
)
```

#### Average

```python
df["Average"] = df["Average"].fillna(
    df["Average"].median()
)
```

#### Salary

```python
df["Salary"] = df["Salary"].fillna(
    df["Salary"].median()
)
```

#### Verification

```python
df.isnull().sum()
```

### Checking Datatypes

```python
df.dtypes
```

#### Convert

```python
df["Matches"] = df["Matches"].astype(int)

df["Salary"] = df["Salary"].astype(int)
```

### Salary in Lakh

```python
df["Salary_Lakh"] = np.round(
    df["Salary"]/100000,
    2
)
```

### Runs Per Match

```python
df["Runs_Per_Match"] = np.round(
    df["Runs"]/df["Matches"],
    2
)
```
