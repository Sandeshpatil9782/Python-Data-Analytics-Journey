### 7.Adding New Column

```python
df["Bonus"] = df["Spending Score"] * 0.10

print(df)
```

#### Applying Lambada

```python
df["Category"] = df["Spending Score"].apply(
    lambda x: "VIP" if x > 300 else "Normal"
)

df
```

#### Age Group

```python
df["Age_Group"] = df["Age"].apply(
    lambda x: "Teen" if x < 18
    else "Adult" if x < 60
    else "Senior"
)
df
```

### GST

```python
df["GST"] = df["Spending Score"] * 0.18

df
```

### Total Purchase

```python
df["Total"] = (
    df["Spending Score"] *
    df["Quantity"]
)

df
```

### Discount %

```python
df["Discount"] = df["Spending Score"].apply(
    lambda x: "20%" if x > 300 else "5%"
)

df
```

### 19.Map Gender

```python
df["Gender"] = df["Gender"].map({
    "M": "Male",
    "F": "Female"
})
df
```

### 20.Customer Type

```python
def customer_type(x):
    if x > 500:
        return "Premium"
    else:
        return "Regular"

df["Customer_Type"] = df["Quantity"].apply(customer_type)

df
```

### 21.Convert Uppercase

### 27.Age Bins

```python
df["Age_Category"] = pd.cut(
    df["Age"],
    bins=[0,18,60,100],
    labels=["Teen","Adult","Senior"]
)

df
```

### 28.Transform Score

```python
df["Double_Score"] = df[
    "Score"
].transform(
    lambda x: x * 2
)

df
```

### Apply Row Wise

```python
df["Final"] = df.apply(
    lambda x:
    x["Score"] +
    x["Quantity"],
    axis=1
)

df
```
