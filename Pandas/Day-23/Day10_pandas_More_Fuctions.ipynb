## IMPORTANT DATA ANALYSIS FUNCTION

```python
import pandas as pd

import numpy as np
```

```python
data = {
    "Name": ["Aman", "Rahul", "Sneha", "Priya"],
    
    "Department": ["IT", "HR", "Finance", "IT"],
    
    "Salary": [50000, 60000, 70000, 80000],
    
    "Skills": [
        ["Python", "SQL"],
        ["Excel", "Power BI"],
        ["Finance", "Excel"],
        ["Python", "ML"]
    ]
}

df = pd.DataFrame(data)

df
```

### NORMALIZE

### Normalize Using value_counts()
    Gives percentage distribution

```python
df["Department"].value_counts(normalize=True)
```

### Convert Into Percentage Distribution

```python
df["Department"].value_counts(normalize=True) * 100
```

### DATA TYPE CONVERSATION

#### Check Datatype

```python
df.dtypes
```

#### Convert To Float

```python
df["Salary"] = df["Salary"].astype(float)

df.dtypes
```

#### Convert To String

```python
df["Salary"] = df["Salary"].astype(str)

df.dtypes
```

#### to_numeric()

```python
df["Salary"] = pd.to_numeric(df["Salary"])

df.dtypes
```

#### to_datetime()

```python
date_df = pd.DataFrame({
    "Date": ["2025-01-01", "2025-02-10"]
})

date_df["Date"] = pd.to_datetime(date_df["Date"])

date_df.dtypes
```

## query()
    Used for filtering data

#### Filter Salary > 60000

```python
df.query("Salary > 60000")
```

#### Multiple Conditions

```python
df.query("Salary > 50000 and Department == 'IT'")
```

#### VARIABLE IN query()

```python
value = 60000

df.query("Salary >= @value")
```

### explode()
    Used to separate list values into rows

#### BEFORE explode()

```python
df
```

#### AFTER explode()

```python
df.explode("Skills")
```

### melt()
    Converts wide data into long data

```python
sales_df = pd.DataFrame({
    "Name": ["Aman", "Rahul"],
    
    "Jan": [100, 200],
    
    "Feb": [150, 250]
})

sales_df
```

```python
sales_df.melt(
    id_vars="Name",
    
    var_name="Month",
    
    value_name="Sales"
)
```

### stack()
    Column TO rows

```python
stack_df = pd.DataFrame({
    "A": [1,2],
    "B": [3,4]
})

stack_df
```

```python
stack_df.stack()
```

### unstack()
    Rows To Columns

```python
stack_df.stack().unstack()
```
