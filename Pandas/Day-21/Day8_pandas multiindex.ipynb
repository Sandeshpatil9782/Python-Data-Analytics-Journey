### PANDAS MULTIINDEX
    Access hierachical rows

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance", "Finance"],
    "Employee": ["Aman", "Rahul", "Sneha", "Priya", "Karan", "Rohit"],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", "Delhi", "Mumbai"],
    "Salary": [50000, 60000, 45000, 52000, 70000, 65000],
    "Bonus": [5000, 6000, 4000, 4500, 7000, 6500]
}

df = pd.DataFrame(data)

df
```

### Set Single index

```python
df1 = df.set_index("Department")

df1
```

### Set MultiIndex

```python
df2 = df.set_index(["Department", "Employee"])

df2
```

### Check Index

```python
df2.index
```

### Access data using loc[]

```python
df2.loc["IT"]
```

### Access particular Rows

```python
df2.loc[("IT", "Aman")]
```

### Access Multiple Rows

```python
df2.loc[[("IT", "Aman"), ("HR", "Sneha")]]
```

### Reset Index

```python
df2.reset_index()
```

### Sort Index

```python
df2.sort_index()
```

### Sort Index Using Level

```python
df2.sort_index(level="Department")
```

### Get Level Values

```python
df2.index.get_level_values(0)
```

### Get First Level Values

```python
df2.index.get_level_values(1)
```

### MultiIndex with Column

```python
pivot1 = pd.pivot_table(
    df,
    values=["Salary", "Bonus"],
    index=["Department", "Employee"],
    columns="City",
    aggfunc="sum",
    fill_value=0
)

pivot1
```

### Swap Level
    Used to interchange index levels.

```python
df2.swaplevel()
```

### Remove Level
    Used to remove one index level

```python
df2.droplevel(0)
```

### Groupby with multindex

```python
group1 = df.groupby(
    ["Department", "City"]
)["Salary"].mean()

group1
```

### Convert series to dataframe

```python
group1.to_frame()
```

### Access multiindex column

```python
pivot2 = pd.pivot_table(
    df,
    values=["Salary", "Bonus"],
    index="Department",
    aggfunc=["sum", "mean"]
)
pivot2
```

```python
pivot2['sum']
```

### Access particular column

```python
pivot2["sum"]["Salary"]
```

### Flatten multiindex column

```python
pivot2.columns = ['_'.join(col) for col in pivot2.columns]

pivot2
```

### Rename Index                                                                             

```python
df2.rename_axis(
    ["Dept", "Emp"]
)
```

### xs()
    xs = Cross Section
    Used to filter specific level data from MultiIndex.

```python
df2.xs("IT")
```

### Filter second level

```python
df2.xs("Aman", level="Employee")
```
