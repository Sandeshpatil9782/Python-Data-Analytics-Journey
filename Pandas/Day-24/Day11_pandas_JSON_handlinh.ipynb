## JSON Handling
    -Javascript Object Notation
    -Common in API'S And Web Data 

```python
import pandas as pd
import json
```

```python
data = {
    "Name": ["Aman", "Rahul", "Sneha"],
    
    "Age": [23, 25, 24],
    
    "City": ["Pune", "Mumbai", "Delhi"]
}

data
```

### Convert Dictionary To Dataframe

```python
df=pd.DataFrame(data)
df
```

### Create JSON File

```python
sample = {
    "Name": ["Aman", "Rahul","Ram","Akash","Sham"],
    "Salary": [50000, 60000,70000,80000,90000]
}

with open("employee.json", "w") as f:
    json.dump(sample, f)
```

#### Read JSON File

```python
pd.read_json("employee.json")
```

### Saved Dataframe as JSON

```python
df.to_json("output.json")
```

#### Read Saved JSON File

```python
pd.read_json("output.json")
```

### JSON Orients

#### records
    -Row wise JSON Dictionary
    -Each row becomes dictionary

```python
df.to_json(orient="records")
```

#### Columns

```python
df.to_json(orient="columns")
```

### index
    Index becomes key

```python
df.to_json(orient="index")
```

### split
    Separate Column Index Data

```python
df.to_json(orient="split")
```

#### Table
    JSON table schema format

```python
df.to_json(orient="table")
```

#### Values
    Only Values

```python
df.to_json(orient="values")
```

### Create JSON String

```python
json_data = '''
[
    {"Name":"Aman","Salary":50000},
    {"Name":"Rahul","Salary":60000}
]
'''
```

#### Load JSON 
    Reads JSON file using JSON Module

    loads :- Reads Strings
    load :- Reads File

```python
data = json.loads(json_data)

data
```

### Conversion To DataFrame

```python
df2 = pd.DataFrame(data)

df2
```

#### NORMALIZE 
    Used for Nested JSON

### Nested JSON

```python
nested_data = [
    {
        "id": 1,
        "Name": "Aman",
        "Education": {
            "Degree": "BCA",
            "Year": 2025
        }
    },

    {
        "id": 2,
        "Name": "Ram",
        "Education": {
            "Degree": "BSC",
            "Year": 2026
        }
    },

    {
        "id": 3,
        "Name": "Sham",
        "Education": {
            "Degree": "MCA",
            "Year": 2024
        }
    }
]
```

#### json_normalize
    -Converts nested JSON into Flat Table

```python
pd.json_normalize(nested_data)
```

#### API Style

```python
api_data = {
    "users": [
        {"id":1, "name":"Aman"},
        {"id":2, "name":"Rahul"}
    ]
}
```

#### Extract Users

```python
users_df = pd.DataFrame(api_data["users"])

users_df
```

### Write JSON Line

```python
df.to_json(
    "json_lines.json",
    
    orient="records",
    
    lines=True
)
```

#### Read JSON Line

```python
pd.read_json(
    "json_lines.json",
    
    lines=True
)
```

#### Print JSON
    -dumps :- Writes String
    -dump :- Writes File

```python
print(
    json.dumps(
        nested_data,
        indent=4
    )
)
```
