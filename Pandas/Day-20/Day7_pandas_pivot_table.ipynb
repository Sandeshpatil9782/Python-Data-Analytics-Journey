```python
import pandas as pd
import numpy as np

print(pd.__version__)
```

## PANDAS PIVOT TABLE
    It is used to summarize the data

```python
data = {
    "Name": ["Aman", "Aman", "Rahul", "Rahul",
             "Sneha", "Sneha", "Priya", "Priya"],

    "Department": ["IT", "HR", "IT", "Finance",
                   "HR", "Finance", "IT", "Finance"],

    "City": ["Pune", "Mumbai", "Pune", "Delhi",
             "Mumbai", "Delhi", "Pune", "Delhi"],

    "Salary": [50000, 45000, 60000, 70000,
               48000, 65000, 72000, 80000],

    "Bonus": [5000, 4000, 6000, 7000,
              4500, 6500, 7200, 8500]
}

df = pd.DataFrame(data)

df
```

### Basic pivot table

```python
pivot1 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)

pivot1
```

### Multiple Aggfunc

```python
pivot2 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc=["sum", "mean", "max", "min"]
)

pivot2
```

### Rows + Columns

```python
pivot3 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="City",
    aggfunc="sum"
)

pivot3
```

### Fill_value
    Fills missing values

```python
pivot4 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="City",
    aggfunc="sum",
    fill_value=0
)

pivot4
```

### margins=
    Adds total row/columns

```python
pivot5 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="City",
    aggfunc="sum",
    margins=True
)

pivot5
```

### Multiple Values

```python
pivot6 = pd.pivot_table(
    df,
    values=["Salary", "Bonus"],
    index="Department",
    aggfunc="sum"
)

pivot6
```

### Multiple Index

```python
pivot7 = pd.pivot_table(
    df,
    values="Salary",
    index=["Department", "City"],
    aggfunc="mean"
)

pivot7
```

### Multiple Column

```python
pivot8 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns=["City", "Name"],
    aggfunc="sum",
    fill_value=0
)

pivot8
```

## Crosstab()
    -To compare the categories
    -Create Frequency Table 
    

```python
cross1 = pd.crosstab(
    df["Department"],
    df["City"]
)

cross1
```

### Crosstab with values

```python
cross2 = pd.crosstab(
    df["Department"],
    df["City"],
    values=df["Salary"],
    aggfunc="sum"
)

cross2
```

### Crosstab With margins

```python
cross3 = pd.crosstab(
    df["Department"],
    df["City"],
    margins=True
)

cross3
```

### Normalize=True
    Shows percentage or proportion

```python
cross3 = pd.crosstab(
    df["Department"],
    df["City"],
    margins=True
)

cross3
```

### Row-Wise Percentage

```python
cross4 = pd.crosstab(
    df["Department"],
    df["City"],
    normalize="index"
)

cross4
```

### Column-Wise Percentage

```python
cross5 = pd.crosstab(
    df["Department"],
    df["City"],
    normalize="columns"
)

cross5
```

### Complete Table Percentage

```python
cross6 = pd.crosstab(
    df["Department"],
    df["City"],
    normalize=True
)

cross6
```

### fillna()

```python
pivot9 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="City",
    aggfunc="sum"
)

pivot9.fillna(0)
```

### DROPNA

```python
pivot10 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="City",
    aggfunc="sum",
    dropna=True
)

pivot10
```

### Count using pivot table

```python
pivot12 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="count"
)

pivot12
```

### Unique Value Count

```python
pivot13 = pd.pivot_table(
    df,
    values="City",
    index="Department",
    aggfunc="nunique"
)

pivot13
```

### Sort Values

```python
pivot14 = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="sum"
)

pivot14.sort_values(by="Salary", ascending=False)
```
