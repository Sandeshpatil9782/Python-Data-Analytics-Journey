### Change Fhone Column Data Type

```python
df["Phone"] = df["Phone"].fillna("Not Available")

df["Phone"] = df["Phone"].astype("string")
```

### Columns Striping

```python
df.columns = df.columns.str.strip()
df.head()
```

### Name Column Strip

```python
df["Name"] = df["Name"].str.strip()
df.head()
```

### Proper Name 

```python
df["Name"] = df["Name"].str.title()
df.head(10)
```

### Spliting Name

```python
Name_split=df[["First_Name", "Last_Name"]] = df["Name"].str.split(" ",n=1, expand=True)
print(Name_split)
```

```python
df.head()
```

### Insert At Postion

```python
print(df.columns.tolist())
```

```python
first = df.pop("First_Name")
last = df.pop("Last_Name")

df.insert(1, "First_Name", first)
df.insert(2, "Last_Name", last)

df.head()
```

### Drop Full Name

```python
df.drop("Name", axis=1, inplace=True)

df.head()
```

### Department Clean

### Unique Values

```python
print(df["Department"].unique())
```

```python
df["Department"] = df["Department"].str.strip()
df.head()
```

### Lower Case

```python
df["Department"] = df["Department"].str.lower()
df.head(10)
```

### Upper Case

```python
df["Department"] = df["Department"].str.upper()
df.head(10)
```

```python
df["Department"] = df["Department"].replace(
    {
      "data analyst":"Data Analyst"
    }
)
df.head(10)
```

### In Title

```python
df["Department"] = df["Department"].str.title()
df.head(10)
```

### City Column

### Checking Unique Values

```python
print(df["City"].unique())
print("\n")
```

### Convert As Title

```python
df["City"] = df["City"].str.title()
df.head(10)
```

### Email Column

```python
df["Email"] = df["Email"].str.lower()
df.head(10)
```

### Check Duplicates

```python
df.duplicated().any()
```

```python
print(df["Employee_ID"].duplicated().sum())
```

### Salary Fill Null Value

```python
df["Salary"] = df["Salary"].fillna(0)
df.head()
```

### Proper Phone Number

```python
df["Phone"] = df["Phone"].astype("Int64")
df.head()
```

```python
print(df['Phone'].isnull().sum)
```

### Filled Null Phone Number 0

```python
df["Phone"] = df["Phone"].fillna(0)
df.head()
```
