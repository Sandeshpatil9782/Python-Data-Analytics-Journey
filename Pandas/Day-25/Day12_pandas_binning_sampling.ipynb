## Pandas Binning,Sampling,Custom Function

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Aman", "Rahul", "Sneha", "Priya",
             "Karan", "Rohit", "Neha", "Pooja"],

    "Marks": [45, 67, 89, 90, 55, 72, 38, 81],

    "Salary": [25000, 40000, 60000, 80000,
               35000, 50000, 22000, 75000]
}

df = pd.DataFrame(data)

df
```

#### BINNING
    Divide Data In Groups

### pd.cut()
    Divide data into Fixed Intervals/Bins 

```python
df["Marks_Category"] = pd.cut(
    df["Marks"],
    
    bins=[0, 40, 60, 80, 100],
    
    labels=["Fail", "Average", "Good", "Excellent"]
)

df
```

### Right Parameter 
    - right decides which side interval is included.
    
        right = False Included Symbol { ] } [0,50) Left Value Included i.e 0

        right = True Excluded Symbol  { ( } (0,50] Right Value Included i.e 50

```python
pd.cut(
    df["Marks"],
    
    bins=[0, 50, 100],
    
    right=False
)
```

### Include Lowest :-To include minimum boundary value in first interval.
        include_lowest = True :- Includes first lowest values

        Include_lowest = False :- lowest value excluded

```python
pd.cut(
    df["Marks"],
    
    bins=[0, 50, 100],
    
    include_lowest=True
)
```

#### Value Count Of Bins
    Count Frequency Of Values

```python
df["Marks_Category"].value_counts()
```

### qcut()
    -Divides Data into Equal-Size Distribution Bins

```python
df["Salary_Group"] = pd.qcut(
    df["Salary"],
    
    q=4,
    
    labels=["Low", "Medium", "High", "Very High"]
)

df
```

### SAMPLING
    Used to get random sample
    Random Sample()

```python
df.sample(3)
```

### Fixed Random State
    Everytime gets Same output

```python
df.sample(
    3,
    
    random_state=1
)
```

### Fraction Sampling
    Fraction sampling selects percentage/fraction of rows.

```python
df.sample(frac=0.5) #Returns 50% rows randomly
```

### Sampling with Replacement
    Same row can repeat

```python
df.sample(
    5,
    replace=True
)
```

### Weigheted Sampling
    Weighted sampling gives higher chance to selected rows
    1 =Lower Chance
    5 =Higher Chance

```python
weights = [1,2,1,5,1,1,1,3]

df.sample(
    3,
    
    weights=weights
)
```

### Custom Function
    Function created by user

### Sample Function

```python
def bonus(salary):
    
    return salary + 5000
```

#### Apply custom Function

```python
df["Updated_Salary"] = df["Salary"].apply(bonus)

df
```

### Lambada function
    

```python
df["Double_Marks"] = df["Marks"].apply(
    lambda x: x * 2
)

df
```

#### With Condition Forms Grade

```python
def grade(marks):
    
    if marks >= 80:
        return "A"
    
    elif marks >= 60:
        return "B"
    
    elif marks >= 40:
        return "C"
    
    else:
        return "Fail"
```

#### Apply on multiple column

```python
def total(row):
    
    return row["Marks"] + row["Salary"]
```

```python
df["Total"] = df.apply(
    total,
    
    axis=1
)

df
```

### map() with Custom Function

```python
df["Marks"] = df["Marks"].map(
    lambda x: x + 5
)

df
```

### applymap() With custom function

```python
num_df = pd.DataFrame({
    "A": [1,2],
    "B": [3,4]
})

num_df.map(lambda x: x * 10)
```

### pipe() With custom function

```python
def add_bonus(data):
    
    data["Salary"] = data["Salary"] + 2000
    
    return data
```

```python
df.pipe(add_bonus)
```

#### Advance Binning

#### BIN LABELS
    Labels give custom names to bins/categories.
    

```python
bins = [0,40,60,80,100]

labels = ["Fail","Average","Good","Excellent"]

pd.cut(
    df["Marks"],
    
    bins=bins,
    
    labels=labels
)
```

#### WITHOUT LABELS

```python
pd.cut(
    df["Marks"],
    
    bins=4
)
```
