### Q41. Compute the **determinant** of a 3×3 matrix.

```python
A = np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]], dtype=float)
ans = np.linalg.det(A)
print(round(ans, 4))  
```

### Q42. Solve the linear system **Ax = b** where A = [[2,1],[5,7]] and b = [11, 13].

```python
A = np.array([[2, 1], [5, 7]], dtype=float)
b = np.array([11, 13], dtype=float)

x = np.linalg.solve(A, b)
a=np.round(x,3)
print(a)
```

### Q43. Compute the **eigenvalues** and **eigenvectors** of a 2×2 symmetric matrix.

```python
A = np.array([[4, 2], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors:\n', eigenvectors)
```

### Q44. Compute the **matrix inverse** of a 3×3 matrix and verify A @ A_inv ≈ I.

```python
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]], dtype=float)

A_inv = np.linalg.inv(A)
print(np.allclose(A @ A_inv, np.eye(3)))  
```

### Q45. Compute the **Frobenius norm** of a 3×3 matrix.

### Q49. Compute the **pairwise Euclidean distances** between every pair of points in a (5, 2) array **without a loop**.

```python
np.random.seed(3)
points = np.random.rand(5, 2)
print('Points shape:', points.shape)

diff = points[:, np.newaxis, :] - points[np.newaxis, :, :]
dist_matrix = np.sqrt((diff ** 2).sum(axis=-1))
print(dist_matrix.round(3))
```

### Q50. 🏆 BOSS LEVEL — Implement **1-step of K-Means update** using only NumPy:
Given data points and initial centroids, assign each point to the nearest centroid and recompute centroid positions.

```python
np.random.seed(0)
data = np.random.rand(10, 2)          # 10 points in 2D
centroids = data[[0, 5]]              # 2 initial centroids
print('Data shape:', data.shape)
print('Centroids:\n', centroids)

diffs = data[:, np.newaxis, :] - centroids[np.newaxis, :, :]  # (10, 2, 2)
distances = np.sqrt((diffs**2).sum(axis=2))                    # (10, 2)
labels = distances.argmin(axis=1)                              # (10,)
new_centroids = np.array([data[labels == k].mean(axis=0) for k in range(2)])
print('Labels:', labels)
print('New centroids:\n', new_centroids)
```
