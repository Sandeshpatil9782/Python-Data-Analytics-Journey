## Indexing And Slicing

```python
import numpy as np
```

```python
a=np.array([10,20,30,40,50,60])
```

```python
print(a)
```

### Indexing
    Used to access specific elements

#### Positive Indexing
    Starts from 0

```python
print(a[0])
```

```python
print(a[3])
```

#### Negative Indexing
    Starts from End OR -1
    

```python
print(a[-1])
```

```python
print(a[-2])
```

### Slicing 
    Used to access multiple elements

```python
print(a[1:4])
```

#### Step Slicing
    Return values with step size

```python
print(a[: :2])
```

```python
print(a[: : 3])
```

#### Negative Slicing

```python
print(a[-3:-1])
```

```python
print(a[-5:-1 :3])
```

### Array Operations

```python
b=np.array([2,3,4,6,4,2])
```

```python
print(b)
```

#### (b + 2) Added each element of b

```python
print(b + 2)
```

#### (b - 2) Subtract each element of b

```python
print(b - 2)
```

#### (b * 3) Multply each element of b

```python
print(b * 3)
```

#### (b / 2) Divide each element of b

```python
print(b / 2)
```

#### (b % 2)  Each element of b Gives Remainder(Modulo)

```python
print(b % 2)
```
