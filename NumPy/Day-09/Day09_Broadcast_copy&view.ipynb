## Broadcasting And Copy VS View 

```python
import numpy as np
```

### 1.Scalar Broadcasting

```python
a=np.array([1,2,3])
print(a + 5)
```

#### For 2D Arrays

```python
b=np.arange(11,17).reshape(2,3)
print("B Array:\n",b,"\n")
print("Broadcasting:\n",b + a)
```

### Copy()
    Creates completely separate array
    Original arrays does not change

```python
a=np.array([1,2,3,4])
b=a.copy()
b[0]=100
print(a)
print(b)
```

### View() 
    Share Arrays same Memory
    Original array also change

```python
a=np.array([1,2,3,4])
b=a.view()
b[0]=100
print(a)
print(b)
```

### Base Attribute
        Tells that whether array owns data or share memory

```python
a=np.array([1,2,3,4])
b=a.view()
print(b.base) # b is connected to a
```

```python
a=np.array([1,2,3,4])
c=a.copy()
print(c.base) # c is Independent
```

### Practice Task

```python
import numpy as np

arr = np.array([10,20,30])

copy_arr = arr.copy()

view_arr = arr.view()

copy_arr[0] = 999

view_arr[1] = 777

print("Original:", arr)

print("Copy:", copy_arr)

print("View:", view_arr)
```
