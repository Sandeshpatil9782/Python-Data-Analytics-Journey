## Functions 

```python
import numpy as np
```

```python
a=np.array([10,50,23,45,36])
a
```

### 1.sort()
    Arrange elements in ascending order

```python
print(np.sort(a))
```

```python
#Descending order Reverse the array
print(np.sort(a)[: : -1])
```

### 2.argsort()
    Returns indexes of sorted elements

```python
print(np.argsort(a))
```

### 3.searchsorted()
    Finds insertion position with ascending order

```python
c=np.array([10,50,23,45,36])
print(np.searchsorted(c,24))
```

### 4.nonzero()
    Leaves the 0 index and gives other indesx

```python
a=np.array([0,2,4,5,0,6])
print(np.nonzero(a))
```

### 5.extract()
    Condition matched elements are given

```python
a=np.array([0,2,4,5,0,6,7,8,9])
print(np.extract(a>4,a))
```

### 6.unique()
    Returns only unique values
    

```python
a=np.array([0,2,4,2,5,6,7,5,0,6])
print(np.unique(a))
```

### 7.any()
    Checks if atleast one condition is true

```python
a=np.array([0,0,0,0,6])
print("One non zero element:\n",np.any(a))

a=np.array([0,0,0,0])
print("All zeros:\n",np.any(a))
```

### 8.all()
    Checks if all values are true

```python
c=np.array([10,50,23,45,36])
print("All are non zero:\n",np.all(c))

c=np.array([10,0,23,0,36])
print("If atleast zero:\n",np.all(c))
```

### 9.append()
    Adds new element at ends

```python
c=np.array([10,50,23,45,36])
print("Original array:\n",c)

print("Added Element:\n",np.append(c,100))
```

### 10.insert()
    Adds value at specific positions


```python
c=np.array([10,50,23,45,36])
print("Original array:\n",c)

print("Added Element at postion 1:\n",np.insert(c,1,100))
```

### 11.delete()
    Remove value using index

```python
c=np.array([10,50,23,45,36])
print("Original array:\n",c)

print("Element deleted from index 1:\n",np.delete(c,1))
```

## Bitwisw Functions
    Operations work on binary numbers (0 And 1)

### 1.bitwise_and()
    Returns true only if both bits are 1

```python
print(np.bitwise_and(5,3)) #True case Binary bits(5 = 1 0 1 AND 3 = 0 1 1)
                                      

print(np.bitwise_and(2,1)) #False case
```

### 2.bitwise_or()
    Returns true if any one bit is 1

```python
print(np.bitwise_or(5,3)) #True case

print(np.bitwise_or(0,0)) #False case
```

### 3.bitwise_xor
    Returns true if bits are different

```python
print(np.bitwise_xor(5,3)) #True case

print(np.bitwise_xor(2,2)) #False case
```

### 4.invert()
    Flips the  bits ,Change 1 To 0 and 0 To 1

```python
print(np.invert(5)) # binary bits(5 = 0000101 ) Inverted To 1111010
```

### 5.left_shift()
    Shift bits to left 
    

```python
print(np.left_shift(5,1)) #Bit Postion shifted by 1

print(np.left_shift(5,2)) #Bit Postion shifted by 2
```

### 6.right_shift()
    Shift bits to  right

```python
print(np.right_shift(5,1)) #Bit Postion shifted by 1

print(np.right_shift(5,2)) #Bit Postion shifted by 2
```

```python
import numpy as np

print("AND:", np.bitwise_and(6,3))

print("OR:", np.bitwise_or(6,3))

print("XOR:", np.bitwise_xor(6,3))

print("LEFT SHIFT:", np.left_shift(6,1))

print("RIGHT SHIFT:", np.right_shift(6,1))
```
