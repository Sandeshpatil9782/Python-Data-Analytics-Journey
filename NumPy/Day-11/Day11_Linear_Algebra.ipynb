## Linear Algebra

```python
import numpy as np
```

### 1.dot product
    dot() Multiplies arrays mathematically

```python
a=np.array([1,2,3])
print(a)
b=np.array([4,5,6])
print(b)
print("Dot product:",np.dot(a,b))
```

### 2.Matrix Multiplecation 

```python
a=np.array(
    [
        [1,2],
        [3,4]
    ])
b=np.array(  [
        [5,6],
        [7,8]
    ]
)
print(a)
print("\n",b,"\n")
print("Matrix Multiplecation:\n",np.matmul(a,b))
```

### 3.Determinant-det()
    It gives matrix value

```python
a=np.array(
    [
        [1,2],
        [3,4]
    ])
print(a)
print("Determinant of a:",np.linalg.det(a))
```

### 4.Inverse matrix
    Returns opposite matrix mathematically
    Only square matrix can have inverse

### Practice Task

```python
import numpy as np

a = np.array([
    [2,3],
    [1,4]
])

b = np.array([
    [5,6],
    [7,8]
])

print("Dot Product:")
print(np.dot([1,2],[3,4]))

print("Matrix Multiplication:")
print(np.matmul(a,b))

print("Transpose:")
print(a.T)

print("Determinant:")
print(np.linalg.det(a))

print("Inverse:")
print(np.linalg.inv(a))
```

```python
a=np.array(
    [
        [1,2],
        [3,4]
    ])
print("Matrix A:\n",a)
print("Inverse Matrix of a:\n",np.linalg.inv(a))
```

### 5.trace()

    Returns sum of diagonal elements


```python
a=np.array(
    [
        [1,2],
        [3,4]
    ])
print(a)
print("Sum of diagonal:",np.trace(a))
```

### 6.diagonal()
    Returns diagonal elements

```python
a=np.array(
    [
        [1,2],
        [3,4]
    ])
print(a)
print("Diagonal Elements:",np.diagonal(a))
```

### 7.rank()
    Tells independent row/columns count

```python
a=np.array(
    [
        [1,2,3],
        [3,4,4],
        [5,6,8]
    ])
print(a)
print("Rank:",np.linalg.matrix_rank(a))
```

### 8.Eigenvalues - eig()
    Special values related to matrix transformation

```python
values,vectors=np.linalg.eig(a)
print(values)
```

### 9.Eigenvectors - eig()
    Directions associated with eigenvalues

```python
print(vectors) # function will give output at ones both
```

### 10.solve()
    Solve Linear equations
    Ex:-  2x + y = 5
          x + y =3

```python
a=np.array([
            [2,3],
            [1,1]
])
print("Array A:\n",a)
b=np.array([5,3])
print("Array B\n",b)
print("Results:",np.linalg.solve(a,b))
```

### Practice Task

```python
import numpy as np

a = np.array([
    [4,2],
    [1,3]
])

print("Trace:")
print(np.trace(a))

print("Diagonal:")
print(np.diagonal(a))

print("Rank:")
print(np.linalg.matrix_rank(a))

values, vectors = np.linalg.eig(a)

print("Eigenvalues:")
print(values)

print("Eigenvectors:")
print(vectors)
```

## Vectorization 
    Performing operations on entire array without loops

```python
a=np.array([1,2,3,4,5])
print(a * 2)
```

### Vectorization Arithmetic

```python
a = np.array([1,2,3])

b = np.array([4,5,6])

print(a + b)

print(a * b)

print(b - a)

print(b / a)
```

### Universal Functions
    Functions that operate element-wise

### 1.sqrt()
    Square Root of Elements

```python
a=np.array([2,4,16,121,169])
print(np.sqrt(a))
```

### 2.sin()
    Sin value of Elements

```python
a=np.array([2,4,16,20])
print(np.sin(a))
```

### 3.log()
    Log value of elements

```python
a=np.array([2,4,8,10])
print(np.log(a))
```

### 4.abs()
    Gives absulute value
    Negative value became a positive

```python
a=np.array([2,4,-16,-23,-4])
print(np.abs(a))
```

### 5.pow()
    Gives power of elements

```python
a=np.array([2,4,16,5,6,7])
print(np.pow(a,2))
```
