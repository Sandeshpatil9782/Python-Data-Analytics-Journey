### Practice Task

```python
import numpy as np

salary = np.array([25000,30000,35000,40000])

bonus = np.array([2000,3000,4000,5000])

print("Updated Salary:")
print(salary + bonus)

print("Double Salary:")
print(salary * 2)

print("Half Salary:")
print(salary / 2)

print("Salary Square:")
print(salary ** 2)
```

### Practice Task

```python
import numpy as np

marks = np.array([20,45,67,89,90,34,76,55])

print("Passed Students:")
print(marks[marks >= 40])

print("Top Scorers:")
print(marks[marks > 80])

print("Average Students:")
print(marks[(marks >= 50) & (marks <= 80)])
```

#### Print values between 20 and 55

```python
print(marks[(marks >= 20) & (marks <= 55)])
```
