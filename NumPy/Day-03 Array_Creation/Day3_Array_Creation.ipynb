## Array Creations Methods

```python
import numpy as np
```

### 1.Basic

```python
a=np.array([2,3,4,5,6])
```

```python
print("Array are:",a)
```

### 2.Using arange()
    Sequence of numbers
    arange(start,stop,step)

```python
a=np.arange(1,10)
```

```python
print("Sequence:",a)
```

#### Example for step

```python
a=np.arange(1,20,2)
```

```python
print("Sequence:",a)
```

### 3. Using zeros()
    Array filled with zero

```python
np.zeros(5)
```

```python
a=np.zeros(10)
```

```python
print("Zeros:",a)
```

### 4.Using ones()
    Array filled with ones

```python
np.ones(5)
```

```python
a=np.ones(10)
```

```python
print("Ones:",a)
```

### 5.Using linspace()
        Given range a specific elements are formed
        linspace(start,stop,number)
        here,difference depends on the given number

```python
a=np.linspace(1,10,5)
```

```python
print("Linspace:",a)
```

```python
b=np.linspace(1,10,4)
```

```python
print("Linspace:",b)
```

### Size
    Total number of elements

```python
print("Size:",a.size)
```
